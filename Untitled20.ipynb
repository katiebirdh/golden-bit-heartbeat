```python
# @title Stability panel: morphic/Sturmian bigits under ONE fixed normalization rule
# @markdown **Rule (pre-declared, single):** For a candidate binary word `A` and the Fibonacci word `F`,
# @markdown interpret both as **decimal bigits** (i.e., "0." + bits in base 10), then compute
# @markdown `scaled = ((A/F) / 729) * 1e4`. Compare `scaled` to the target constant `alpha_inv = 137.035999`.
# @markdown
# @markdown **Families swept:** Rabbit (morphic), two generic **balanced** Sturmian words (slopes √2−1 and 1/π),
# @markdown and Thue–Morse (as an automatic control, not balanced). Denominator is **always** Fibonacci.
# @markdown
# @markdown **Outputs:** a results DataFrame (by length & family), a summary table, and optional plots.

import math, itertools, statistics, sys, typing
from decimal import Decimal, getcontext
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------- Parameters (feel free to edit) ----------
alpha_inv = Decimal("137.035999")       # target constant for comparison (your value)
C_norm    = Decimal(729)                # fixed normalization constant = 9^3
SCALE     = Decimal(10_000)             # fixed post-scale
lengths   = [30, 60, 90, 120, 180, 240, 360, 480, 600]  # prefix lengths to sweep
precision_margin = 50                   # extra decimal precision beyond max length
show_plots = True                       # turn on/off plots

# ---------- Utilities ----------
def fib_word(n: int) -> str:
    """
    Fibonacci word via standard morphism: 0 -> 01, 1 -> 0, starting from '0'.
    Returns the first n bits as a string of '0'/'1'.
    """
    s = "0"
    # Grow by morphism until length >= n
    while len(s) < n:
        s = s.replace("0", "X").replace("1", "Y")
        s = s.replace("X", "01").replace("Y", "0")
    return s[:n]

def rabbit_word(n: int) -> str:
    """
    Rabbit (Knuth's 'rabbit' word) via Fibonacci concatenation:
    R1='1', R2='10', R_{k+1} = R_k + R_{k-1}.
    Returns first n bits as '0'/'1'.
    """
    a, b = "1", "10"
    s = a
    while len(s) < n:
        a, b = b, a + b
        s = b
    return s[:n]

def thue_morse(n: int) -> str:
    """
    Thue–Morse: t(n) = parity of popcount(n), starting at index 0.
    Returns first n bits as '0'/'1'.
    """
    out = []
    for i in range(n):
        out.append(str(bin(i).count("1") & 1))
    return "".join(out)

def sturmian_word(n: int, slope: float, intercept: float = 0.0) -> str:
    """
    Balanced mechanical word (Sturmian) with given irrational slope in (0,1).
    Bit k is floor((k+1)*slope + intercept) - floor(k*slope + intercept).
    Uses float arithmetic (adequate for modest n; avoids ties for typical irrational slopes).
    """
    bits = []
    f = math.floor
    s = slope
    b = intercept
    for k in range(n):
        x1 = (k + 1) * s + b
        x0 = k * s + b
        bits.append("1" if (f(x1) - f(x0)) == 1 else "0")
    return "".join(bits)

def bigit_decimal(bits: str) -> Decimal:
    """
    Interpret a 'bigit' as the Decimal value of the string "0." + bits (in base-10 digits 0/1).
    """
    # Set precision high enough for this conversion
    getcontext().prec = len(bits) + precision_margin
    return Decimal("0." + bits) if bits else Decimal(0)

def ones_density(bits: str) -> float:
    return bits.count("1") / max(1, len(bits))

# ---------- Fixed normalization rule ----------
# scaled(A,F) = ((A/F) / 729) * 1e4
def scaled_value(A_dec: Decimal, F_dec: Decimal) -> Decimal:
    if F_dec == 0:
        return Decimal("NaN")
    getcontext().prec = max(getcontext().prec,  max(len(str(A_dec)), len(str(F_dec))) + 20)
    return (A_dec / F_dec) / C_norm * SCALE

# ---------- Families to test ----------
# Reference denominator (fixed): Fibonacci word (balanced/Sturmian)
FAMILIES = {
    # name: function to generate bits of length n
    "Rabbit": lambda n: rabbit_word(n),
    "Sturmian_sqrt2-1": lambda n: sturmian_word(n, slope=math.sqrt(2)-1.0, intercept=0.0),
    "Sturmian_1/pi":   lambda n: sturmian_word(n, slope=1.0/math.pi, intercept=0.0),
    "Thue-Morse":      lambda n: thue_morse(n),  # control (automatic, not balanced)
    # You can add more families here if you like:
    # "Sturmian_phi^-2":  lambda n: sturmian_word(n, slope=(2 - (1+5**0.5)/2), intercept=0.0),
}

# ---------- Sweep and collect ----------
rows = []
for n in lengths:
    F_bits = fib_word(n)
    F_dec  = bigit_decimal(F_bits)
    F_den  = F_dec if F_dec != 0 else Decimal("NaN")

    for name, gen in FAMILIES.items():
        A_bits = gen(n)
        A_dec  = bigit_decimal(A_bits)
        scaled = scaled_value(A_dec, F_den)
        err    = scaled - alpha_inv
        # ppm relative to target
        ppm = (err / alpha_inv) * Decimal(1_000_000)

        rows.append({
            "length": n,
            "family": name,
            "ones_density_A": ones_density(A_bits),
            "ones_density_F": ones_density(F_bits),
            "A_bigit_decimal": str(A_dec),   # as string, to avoid scientific fmt
            "F_bigit_decimal": str(F_dec),
            "scaled": float(scaled),
            "error_abs": float(err),
            "error_ppm": float(ppm),
        })

df = pd.DataFrame(rows).sort_values(["family","length"]).reset_index(drop=True)

# ---------- Summaries ----------
def summarize_family(sub: pd.DataFrame) -> dict:
    errs = sub["error_abs"].to_numpy()
    ppms = sub["error_ppm"].to_numpy()
    lens = sub["length"].to_numpy()

    # stability proxy: correlation of |error| with length (negative is "good")
    with np.errstate(divide='ignore', invalid='ignore'):
        abs_err = np.abs(errs)
        # Spearman-like rank correlation (fallback if constant)
        try:
            rank_len  = pd.Series(lens).rank(method="average").to_numpy()
            rank_err  = pd.Series(abs_err).rank(method="average").to_numpy()
            corr_rank = np.corrcoef(rank_len, rank_err)[0,1]
        except Exception:
            corr_rank = np.nan

        # log-log slope of |error| ~ N^beta (more negative beta => faster decay)
        mask = (abs_err > 0) & (lens > 0)
        if mask.sum() >= 2:
            X = np.log(lens[mask]); y = np.log(abs_err[mask])
            beta = np.polyfit(X, y, 1)[0]
        else:
            beta = np.nan

    return {
        "family": sub["family"].iloc[0],
        "n_points": len(sub),
        "mean_abs_error": float(np.mean(np.abs(errs))),
        "median_abs_error": float(np.median(np.abs(errs))),
        "min_abs_error": float(np.min(np.abs(errs))),
        "mean_ppm": float(np.mean(ppms)),
        "median_ppm": float(np.median(ppms)),
        "min_ppm": float(np.min(np.abs(ppms))),
        "spearman_like_corr_len_vs_abs_err": float(corr_rank),
        "loglog_beta(|err|~N^beta)": float(beta),
    }

summary = (df
           .groupby("family", as_index=False)
           .apply(lambda g: pd.Series(summarize_family(g)))
           .reset_index(drop=True))

# ---------- Display ----------
print("\n=== Results (first 20 rows) ===")
print(df.head(20).to_string(index=False))

print("\n=== Summary by family (lower abs error & more negative beta are better) ===")
print(summary.sort_values(["median_abs_error"]).to_string(index=False))

# Save CSVs in Colab
df.to_csv("/content/stability_panel_rows.csv", index=False)
summary.to_csv("/content/stability_panel_summary.csv", index=False)
print("\nSaved CSVs:")
print(" - /content/stability_panel_rows.csv")
print(" - /content/stability_panel_summary.csv")

# ---------- Optional plots ----------
if show_plots:
    # Error (ppm) vs length
    plt.figure(figsize=(7,5))
    for fam, sub in df.groupby("family"):
        plt.plot(sub["length"], sub["error_ppm"], marker="o", label=fam)
    plt.axhline(0, linestyle="--")
    plt.xlabel("prefix length N")
    plt.ylabel("error (ppm) vs α⁻¹")
    plt.title("Stability: ((A/F)/729)*1e4 vs α⁻¹")
    plt.legend()
    plt.show()

    # |error| vs length (log-log)
    plt.figure(figsize=(7,5))
    for fam, sub in df.groupby("family"):
        y = np.abs(sub["error_abs"].to_numpy())
        plt.loglog(sub["length"], y, marker="o", label=fam)
    plt.xlabel("prefix length N (log)")
    plt.ylabel("|error| (log)")
    plt.title("Decay of |error| with N (log–log)")
    plt.legend()
    plt.show()
```
