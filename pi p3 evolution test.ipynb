```python
#!/usr/bin/env python3
"""
π EVOLUTION PERIOD-4 TEST SUITE
===============================
Critical test of Harmonic Resonance Theory:
Tests whether sequences EVOLVING toward π show period-4 signatures
while final π digits (already at "unity state") do not.

This directly tests the "finite approaching infinite" hypothesis
and could be the definitive proof of the theory.

Author: [Your Name]
Date: 2025
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.fft import fft, fftfreq
import pandas as pd
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
import mpmath

# High precision setup
mpmath.mp.dps = 100  # Very high precision for π computations
np.random.seed(42)
plt.style.use('default')

# Constants
PHI = float(mpmath.phi)
PI = float(mpmath.pi)
E = float(mpmath.e)

print("🔬 π EVOLUTION PERIOD-3 vs PERIOD-4 TEST SUITE")
print("=" * 60)
print("Testing: Sequences evolving toward π vs final π digits")
print("Hypothesis: Evolution sequences show period-3 (triangular) or period-4 (square) patterns")
print("Control: Final π/e digits should show neither pattern")
print("=" * 60)

class PiEvolutionAnalyzer:
    """Analyze period-4 signatures in π evolution vs final state sequences"""

    def __init__(self):
        self.phi = PHI

    def leibniz_pi_sequence(self, n_terms):
        """Generate Leibniz series convergence: π/4 = 1 - 1/3 + 1/5 - 1/7 + ..."""
        print("🔄 Computing Leibniz π/4 series...")

        partial_sums = []
        current_sum = 0

        for n in tqdm(range(n_terms), desc="Leibniz series"):
            term = (-1)**n / (2*n + 1)
            current_sum += term
            partial_sums.append(current_sum * 4)  # Scale to approach π

        return np.array(partial_sums)

    def machin_pi_sequence(self, n_terms):
        """Generate Machin formula convergence: π/4 = 4*arctan(1/5) - arctan(1/239)"""
        print("🔄 Computing Machin formula convergence...")

        def arctan_series(x, n_terms):
            """Compute arctan(x) using Taylor series"""
            result = 0
            x_power = x
            for n in range(n_terms):
                term = (-1)**n * x_power / (2*n + 1)
                result += term
                x_power *= x*x
            return result

        partial_sums = []

        for n in tqdm(range(1, n_terms + 1), desc="Machin formula"):
            arctan_1_5 = arctan_series(1/5, n)
            arctan_1_239 = arctan_series(1/239, n)
            pi_approx = 4 * (4 * arctan_1_5 - arctan_1_239)
            partial_sums.append(pi_approx)

        return np.array(partial_sums)

    def chudnovsky_pi_sequence(self, n_terms):
        """Generate Chudnovsky algorithm convergence (fastest known π algorithm)"""
        print("🔄 Computing Chudnovsky algorithm convergence...")

        # Use mpmath for high precision
        mpmath.mp.dps = 50

        partial_sums = []
        pi_sum = mpmath.mpf(0)

        for k in tqdm(range(n_terms), desc="Chudnovsky"):
            # Chudnovsky series term
            numerator = mpmath.factorial(6*k) * (545140134*k + 13591409)
            denominator = mpmath.factorial(3*k) * (mpmath.factorial(k)**3) * ((-262537412640768000)**k)

            term = numerator / denominator
            pi_sum += term

            # Convert to π approximation
            if pi_sum != 0:
                pi_approx = 1 / (12 * pi_sum) * mpmath.sqrt(10005)
                partial_sums.append(float(pi_approx))
            else:
                partial_sums.append(0)

        return np.array(partial_sums)

    def sqrt_pi_evolution(self, n_terms):
        """Generate √π through iterative computation"""
        print("🔄 Computing √π evolution...")

        # Use Newton's method: x_{n+1} = (x_n + π/x_n) / 2
        sequence = []
        x = 1.5  # Initial guess

        for n in tqdm(range(n_terms), desc="√π evolution"):
            x = (x + PI/x) / 2
            sequence.append(x)

        return np.array(sequence)

    def pi_fraction_digits(self, fraction, n_digits):
        """Generate digits of π/fraction"""
        print(f"🔄 Computing π/{fraction} digits...")

        mpmath.mp.dps = n_digits + 10
        pi_frac = mpmath.pi / fraction
        pi_str = str(pi_frac).replace('.', '')

        # Convert to digit sequence
        digits = [int(d) for d in pi_str[:n_digits] if d.isdigit()]
        return np.array(digits)

    def pi_final_digits(self, n_digits):
        """Generate final π digits (control - should NOT show period-4)"""
        print("🔄 Computing final π digits...")

        mpmath.mp.dps = n_digits + 10
        pi_str = str(mpmath.pi).replace('.', '')

        # Convert to digit sequence
        digits = [int(d) for d in pi_str[:n_digits] if d.isdigit()]
        return np.array(digits)

    def e_final_digits(self, n_digits):
        """Generate final e digits (control - should NOT show period-4)"""
        print("🔄 Computing final e digits...")

        mpmath.mp.dps = n_digits + 10
        e_str = str(mpmath.e).replace('.', '')

        # Convert to digit sequence
        digits = [int(d) for d in e_str[:n_digits] if d.isdigit()]
        return np.array(digits)

    def period_fft_analysis(self, sequence, sequence_name, target_period=3):
        """Analyze period-N content using FFT"""
        if len(sequence) < 16:
            return 0.0, 0.0, [], []

        # Normalize sequence
        seq_clean = sequence[np.isfinite(sequence)]  # Remove any NaN/inf
        if len(seq_clean) < 16:
            return 0.0, 0.0, [], []

        seq_norm = (seq_clean - np.mean(seq_clean))
        if np.std(seq_norm) > 0:
            seq_norm = seq_norm / np.std(seq_norm)

        # FFT analysis
        fft_vals = fft(seq_norm)
        power_spectrum = np.abs(fft_vals)**2
        freqs = fftfreq(len(seq_norm))

        # Find period-3 component (frequency = 1/3 ≈ 0.333)
        target_freq_3 = 1.0/3.0
        freq_tolerance = 0.03

        period3_indices = np.where(np.abs(freqs - target_freq_3) < freq_tolerance)[0]
        if len(period3_indices) > 0:
            period3_power = np.max(power_spectrum[period3_indices])
        else:
            period3_power = 0

        # Also check period-4 for comparison (frequency = 1/4 = 0.25)
        target_freq_4 = 0.25
        period4_indices = np.where(np.abs(freqs - target_freq_4) < freq_tolerance)[0]
        if len(period4_indices) > 0:
            period4_power = np.max(power_spectrum[period4_indices])
        else:
            period4_power = 0

        # Total power (excluding DC component)
        total_power = np.sum(power_spectrum[1:len(power_spectrum)//2])  # Only positive frequencies
        period3_ratio = period3_power / total_power if total_power > 0 else 0
        period4_ratio = period4_power / total_power if total_power > 0 else 0

        print(f"  {sequence_name}: Period-3 ratio = {period3_ratio:.6f}, Period-4 ratio = {period4_ratio:.6f}")

        return period3_ratio, period4_ratio, freqs, power_spectrum

    def statistical_uniformity_test(self, sequence, sequence_name, n_bins=10):
        """Test for 'too uniform' distribution"""
        if len(sequence) == 0:
            return 0, 1, 0

        # Convert to integers for binning
        if np.all(sequence >= 0) and np.all(sequence <= 1):
            # Probability-like sequence
            bins = (sequence * n_bins).astype(int)
            bins = np.clip(bins, 0, n_bins-1)
        else:
            # General sequence - use modulo
            bins = np.array(sequence).astype(int) % n_bins

        # Count occurrences
        observed = np.bincount(bins, minlength=n_bins)
        expected = len(sequence) / n_bins

        # Chi-square test
        chi2 = np.sum((observed - expected)**2 / expected)
        p_val = 1 - stats.chi2.cdf(chi2, df=n_bins-1)

        # Z-score for "too uniform" (negative means more uniform than expected)
        expected_chi2 = n_bins - 1
        var_chi2 = 2 * (n_bins - 1)
        z_score = (chi2 - expected_chi2) / np.sqrt(var_chi2)

        print(f"  {sequence_name}: Uniformity z-score = {z_score:.2f}σ")

        return chi2, p_val, z_score

    def convergence_quality_analysis(self, sequence, target_value, sequence_name):
        """Analyze how the sequence converges to target"""
        if len(sequence) < 10:
            return {}

        # Calculate errors from target
        errors = np.abs(sequence - target_value)

        # Convergence rate analysis
        log_errors = np.log(errors[errors > 0])
        indices = np.arange(len(log_errors))

        if len(log_errors) > 5:
            # Fit exponential decay
            try:
                slope, intercept = np.polyfit(indices, log_errors, 1)
                convergence_rate = -slope
            except:
                convergence_rate = 0
        else:
            convergence_rate = 0

        # Final accuracy
        final_error = errors[-1] if len(errors) > 0 else float('inf')
        final_rel_error = final_error / target_value * 100

        # Oscillation analysis
        if len(sequence) > 10:
            recent_std = np.std(sequence[-10:])
            oscillation_score = recent_std / np.mean(np.abs(sequence[-10:]))
        else:
            oscillation_score = 0

        results = {
            'convergence_rate': convergence_rate,
            'final_error': final_error,
            'final_rel_error': final_rel_error,
            'oscillation_score': oscillation_score
        }

        print(f"  {sequence_name}:")
        print(f"    Convergence rate: {convergence_rate:.4f}")
        print(f"    Final error: {final_error:.2e}")
        print(f"    Final rel error: {final_rel_error:.4f}%")

        return results

def run_pi_evolution_test():
    """Run the complete π evolution test"""

    print("\n🚀 STARTING π EVOLUTION ANALYSIS")
    print("=" * 50)

    analyzer = PiEvolutionAnalyzer()

    # Parameters
    n_terms = 2000
    n_digits = 2000

    # Generate sequences
    print("\n📊 GENERATING SEQUENCES")
    print("-" * 30)

    sequences = {}

    # EVOLUTION SEQUENCES (should show period-4)
    print("\n🔄 Evolution Sequences (should show period-4):")
    sequences['leibniz_pi'] = analyzer.leibniz_pi_sequence(n_terms)
    sequences['machin_pi'] = analyzer.machin_pi_sequence(min(n_terms, 200))  # Slower convergence
    sequences['sqrt_pi_evolution'] = analyzer.sqrt_pi_evolution(n_terms)
    sequences['pi_over_2_digits'] = analyzer.pi_fraction_digits(2, n_digits)
    sequences['pi_over_3_digits'] = analyzer.pi_fraction_digits(3, n_digits)
    sequences['pi_over_4_digits'] = analyzer.pi_fraction_digits(4, n_digits)

    # FINAL STATE SEQUENCES (should NOT show period-4)
    print("\n🎯 Final State Sequences (should NOT show period-4):")
    sequences['pi_final_digits'] = analyzer.pi_final_digits(n_digits)
    sequences['e_final_digits'] = analyzer.e_final_digits(n_digits)

    # ANALYSIS
    print("\n🔍 PERIOD-4 ANALYSIS")
    print("-" * 30)

    results = []

    for name, sequence in sequences.items():
        print(f"\nAnalyzing {name}...")

        # Period-3 and Period-4 analysis
        period3_ratio, period4_ratio, freqs, power = analyzer.period_fft_analysis(sequence, name)

        # Uniformity analysis
        chi2, p_val, z_score = analyzer.statistical_uniformity_test(sequence, name)

        # Convergence analysis (for evolution sequences)
        if 'evolution' in name or 'leibniz' in name or 'machin' in name:
            target = PI if 'sqrt' not in name else np.sqrt(PI)
            conv_results = analyzer.convergence_quality_analysis(sequence, target, name)
        else:
            conv_results = {}

        results.append({
            'sequence': name,
            'type': 'evolution' if any(keyword in name for keyword in ['leibniz', 'machin', 'sqrt', 'over']) else 'final',
            'period3_ratio': period3_ratio,
            'period4_ratio': period4_ratio,
            'uniformity_z': z_score,
            'length': len(sequence),
            **conv_results
        })

    # Create results DataFrame
    df = pd.DataFrame(results)

    # SUMMARY ANALYSIS
    print("\n" + "=" * 50)
    print("📋 RESULTS SUMMARY")
    print("=" * 50)

    evolution_df = df[df['type'] == 'evolution']
    final_df = df[df['type'] == 'final']

    print(f"\nEvolution sequences (should show period patterns):")
    print(f"Average period-3 ratio: {evolution_df['period3_ratio'].mean():.6f}")
    print(f"Average period-4 ratio: {evolution_df['period4_ratio'].mean():.6f}")
    print(f"Sequences with strong period-3 (>0.01): {len(evolution_df[evolution_df['period3_ratio'] > 0.01])}/{len(evolution_df)}")
    print(f"Sequences with strong period-4 (>0.01): {len(evolution_df[evolution_df['period4_ratio'] > 0.01])}/{len(evolution_df)}")
    print(f"Sequences with extreme uniformity (<-3σ): {len(evolution_df[evolution_df['uniformity_z'] < -3])}/{len(evolution_df)}")

    print(f"\nFinal state sequences (should NOT show period patterns):")
    print(f"Average period-3 ratio: {final_df['period3_ratio'].mean():.6f}")
    print(f"Average period-4 ratio: {final_df['period4_ratio'].mean():.6f}")
    print(f"Sequences with strong period-3 (>0.01): {len(final_df[final_df['period3_ratio'] > 0.01])}/{len(final_df)}")
    print(f"Sequences with strong period-4 (>0.01): {len(final_df[final_df['period4_ratio'] > 0.01])}/{len(final_df)}")
    print(f"Sequences with extreme uniformity (<-3σ): {len(final_df[final_df['uniformity_z'] < -3])}/{len(final_df)}")

    # Statistical tests for both periods
    evolution_period3 = evolution_df['period3_ratio'].values
    final_period3 = final_df['period3_ratio'].values
    evolution_period4 = evolution_df['period4_ratio'].values
    final_period4 = final_df['period4_ratio'].values

    if len(evolution_period3) > 0 and len(final_period3) > 0:
        statistic3, p_value3 = stats.mannwhitneyu(evolution_period3, final_period3, alternative='greater')
        print(f"\nStatistical test - Period-3 (Mann-Whitney U):")
        print(f"Evolution vs Final period-3 ratios: p = {p_value3:.6f}")

        if p_value3 < 0.01:
            print("🎯 HIGHLY SIGNIFICANT: Evolution sequences show more period-3!")
        elif p_value3 < 0.05:
            print("✅ SIGNIFICANT: Evolution sequences show more period-3!")
        else:
            print("❌ Period-3 not significant")

    if len(evolution_period4) > 0 and len(final_period4) > 0:
        statistic4, p_value4 = stats.mannwhitneyu(evolution_period4, final_period4, alternative='greater')
        print(f"\nStatistical test - Period-4 (Mann-Whitney U):")
        print(f"Evolution vs Final period-4 ratios: p = {p_value4:.6f}")

        if p_value4 < 0.01:
            print("🎯 HIGHLY SIGNIFICANT: Evolution sequences show more period-4!")
        elif p_value4 < 0.05:
            print("✅ SIGNIFICANT: Evolution sequences show more period-4!")
        else:
            print("❌ Period-4 not significant")

    # Compare period-3 vs period-4 strength overall
    all_period3 = df['period3_ratio'].values
    all_period4 = df['period4_ratio'].values

    if len(all_period3) > 0 and len(all_period4) > 0:
        stat_comparison, p_comparison = stats.wilcoxon(all_period3, all_period4, alternative='greater')
        print(f"\nPeriod-3 vs Period-4 comparison (all sequences):")
        print(f"Period-3 stronger than Period-4: p = {p_comparison:.6f}")

        avg_p3 = np.mean(all_period3)
        avg_p4 = np.mean(all_period4)
        print(f"Average period-3: {avg_p3:.6f}")
        print(f"Average period-4: {avg_p4:.6f}")
        print(f"Ratio (P3/P4): {avg_p3/avg_p4:.2f}" if avg_p4 > 0 else "Ratio: undefined")

    # VISUALIZATION
    plot_pi_evolution_results(df, sequences)

    # FINAL VERDICT
    print("\n" + "=" * 50)
    print("🎯 FINAL VERDICT")
    print("=" * 50)

    strong_evolution_period3 = len(evolution_df[evolution_df['period3_ratio'] > 0.01])
    strong_evolution_period4 = len(evolution_df[evolution_df['period4_ratio'] > 0.01])
    strong_final_period3 = len(final_df[final_df['period3_ratio'] > 0.01])
    strong_final_period4 = len(final_df[final_df['period4_ratio'] > 0.01])

    # Check which period is stronger
    avg_p3_evolution = evolution_df['period3_ratio'].mean()
    avg_p4_evolution = evolution_df['period4_ratio'].mean()
    avg_p3_final = final_df['period3_ratio'].mean()
    avg_p4_final = final_df['period4_ratio'].mean()

    p3_wins = avg_p3_evolution > avg_p4_evolution
    evolution_stronger = (avg_p3_evolution + avg_p4_evolution) > (avg_p3_final + avg_p4_final)

    print(f"Period-3 dominance: {'YES' if p3_wins else 'NO'}")
    print(f"Evolution stronger than final: {'YES' if evolution_stronger else 'NO'}")

    significant_period3 = p_value3 < 0.05 if 'p_value3' in locals() else False
    significant_period4 = p_value4 < 0.05 if 'p_value4' in locals() else False

    if significant_period3 and not significant_period4:
        print("🔥 PERIOD-3 THEORY CONFIRMED: π systems show period-3, not period-4!")
        print("This suggests triangular rather than square harmonic organization!")
    elif significant_period4 and not significant_period3:
        print("🔥 PERIOD-4 THEORY CONFIRMED: Evolution sequences show period-4!")
        print("This supports the original square harmonic hypothesis!")
    elif significant_period3 and significant_period4:
        print("🌀 DUAL PERIOD STRUCTURE: Both period-3 and period-4 are significant!")
        print("This suggests complex harmonic interactions!")
    elif strong_evolution_period3 > strong_final_period3 or strong_evolution_period4 > strong_final_period4:
        print("📈 PARTIAL SUPPORT: Some evolution sequences show periodic structure")
        print(f"Period-3 strength: Evolution={avg_p3_evolution:.4f}, Final={avg_p3_final:.4f}")
        print(f"Period-4 strength: Evolution={avg_p4_evolution:.4f}, Final={avg_p4_final:.4f}")
        if p3_wins:
            print("Period-3 appears to be the dominant structure!")
        else:
            print("Period-4 appears to be the dominant structure!")
    else:
        print("🤔 THEORY NEEDS REVISION: No clear periodic difference")
        print("May need to reconsider the universality claim or test other periods")

    return df, sequences

def plot_pi_evolution_results(df, sequences):
    """Plot comprehensive results for both period-3 and period-4"""

    fig = plt.figure(figsize=(18, 12))

    # Period-3 vs Period-4 comparison
    ax1 = plt.subplot(2, 4, 1)
    evolution_p3 = df[df['type'] == 'evolution']['period3_ratio']
    evolution_p4 = df[df['type'] == 'evolution']['period4_ratio']
    final_p3 = df[df['type'] == 'final']['period3_ratio']
    final_p4 = df[df['type'] == 'final']['period4_ratio']

    x = [1, 2, 3, 4]
    means = [evolution_p3.mean(), evolution_p4.mean(), final_p3.mean(), final_p4.mean()]
    labels = ['Evol P3', 'Evol P4', 'Final P3', 'Final P4']
    colors = ['red', 'orange', 'blue', 'lightblue']

    bars = ax1.bar(x, means, color=colors)
    ax1.set_xticks(x)
    ax1.set_xticklabels(labels, rotation=45)
    ax1.set_ylabel('Average Period Ratio')
    ax1.set_title('Period-3 vs Period-4: Evolution vs Final')
    ax1.grid(True, alpha=0.3)

    # Detailed period-3 comparison
    ax2 = plt.subplot(2, 4, 2)
    df_sorted = df.sort_values('period3_ratio', ascending=True)
    colors = ['red' if t == 'evolution' else 'blue' for t in df_sorted['type']]
    bars = ax2.barh(range(len(df_sorted)), df_sorted['period3_ratio'], color=colors)
    ax2.set_yticks(range(len(df_sorted)))
    ax2.set_yticklabels(df_sorted['sequence'], fontsize=8)
    ax2.set_xlabel('Period-3 Ratio')
    ax2.set_title('Period-3 Ratios by Sequence')
    ax2.grid(True, alpha=0.3)

    # Detailed period-4 comparison
    ax3 = plt.subplot(2, 4, 3)
    df_sorted = df.sort_values('period4_ratio', ascending=True)
    colors = ['red' if t == 'evolution' else 'blue' for t in df_sorted['type']]
    bars = ax3.barh(range(len(df_sorted)), df_sorted['period4_ratio'], color=colors)
    ax3.set_yticks(range(len(df_sorted)))
    ax3.set_yticklabels(df_sorted['sequence'], fontsize=8)
    ax3.set_xlabel('Period-4 Ratio')
    ax3.set_title('Period-4 Ratios by Sequence')
    ax3.grid(True, alpha=0.3)

    # Period-3 vs Period-4 scatter
    ax4 = plt.subplot(2, 4, 4)
    evolution_mask = df['type'] == 'evolution'
    ax4.scatter(df[evolution_mask]['period3_ratio'], df[evolution_mask]['period4_ratio'],
               c='red', s=50, alpha=0.7, label='Evolution')
    ax4.scatter(df[~evolution_mask]['period3_ratio'], df[~evolution_mask]['period4_ratio'],
               c='blue', s=50, alpha=0.7, label='Final State')

    # Add diagonal line
    max_val = max(df['period3_ratio'].max(), df['period4_ratio'].max())
    ax4.plot([0, max_val], [0, max_val], 'k--', alpha=0.5, label='P3=P4')

    ax4.set_xlabel('Period-3 Ratio')
    ax4.set_ylabel('Period-4 Ratio')
    ax4.set_title('Period-3 vs Period-4 Strength')
    ax4.legend()
    ax4.grid(True, alpha=0.3)

    # Sample sequence plots
    ax5 = plt.subplot(2, 4, 5)
    if 'leibniz_pi' in sequences:
        leibniz = sequences['leibniz_pi'][:500]  # First 500 terms
        ax5.plot(leibniz, 'r-', linewidth=1, alpha=0.7)
        ax5.axhline(PI, color='black', linestyle='--', label='π')
        ax5.set_title('Leibniz π/4 Series Convergence')
        ax5.set_ylabel('Value')
        ax5.legend()
        ax5.grid(True, alpha=0.3)

    # FFT spectrum example with both periods marked
    ax6 = plt.subplot(2, 4, 6)
    if 'machin_pi' in sequences:
        machin = sequences['machin_pi'][:200]
        if len(machin) > 16:
            machin_norm = (machin - np.mean(machin)) / np.std(machin)
            fft_vals = np.abs(np.fft.fft(machin_norm))**2
            freqs = np.fft.fftfreq(len(machin_norm))

            # Plot positive frequencies only
            pos_mask = freqs > 0
            ax6.semilogy(freqs[pos_mask], fft_vals[pos_mask], 'b-', linewidth=1)
            ax6.axvline(1/3, color='red', linestyle='--', label='Period-3 (0.333)', alpha=0.7)
            ax6.axvline(0.25, color='orange', linestyle='--', label='Period-4 (0.25)', alpha=0.7)
            ax6.set_xlabel('Frequency')
            ax6.set_ylabel('Power (log scale)')
            ax6.set_title('FFT Spectrum: Machin Formula')
            ax6.legend()
            ax6.grid(True, alpha=0.3)
            ax6.set_xlim(0, 0.5)

    # Uniformity analysis
    ax7 = plt.subplot(2, 4, 7)
    ax7.scatter(df['period3_ratio'], df['uniformity_z'],
               c=['red' if t == 'evolution' else 'blue' for t in df['type']],
               s=50, alpha=0.7, label='Period-3')
    ax7.scatter(df['period4_ratio'], df['uniformity_z'],
               c=['red' if t == 'evolution' else 'blue' for t in df['type']],
               s=50, alpha=0.4, marker='s', label='Period-4')
    ax7.set_xlabel('Period Ratio')
    ax7.set_ylabel('Uniformity Z-score')
    ax7.set_title('Periods vs Uniformity')
    ax7.axhline(-3, color='orange', linestyle='--', alpha=0.5)
    ax7.axhline(-5, color='red', linestyle='--', alpha=0.5)
    ax7.legend()
    ax7.grid(True, alpha=0.3)

    # Period strength comparison by sequence type
    ax8 = plt.subplot(2, 4, 8)
    evolution_df = df[df['type'] == 'evolution']
    final_df = df[df['type'] == 'final']

    categories = ['Evolution\nPeriod-3', 'Evolution\nPeriod-4', 'Final\nPeriod-3', 'Final\nPeriod-4']
    values = [
        evolution_df['period3_ratio'].mean(),
        evolution_df['period4_ratio'].mean(),
        final_df['period3_ratio'].mean(),
        final_df['period4_ratio'].mean()
    ]
    colors = ['darkred', 'red', 'darkblue', 'blue']

    bars = ax8.bar(categories, values, color=colors)
    ax8.set_ylabel('Average Period Ratio')
    ax8.set_title('Period Strength by Type')
    ax8.tick_params(axis='x', rotation=45)
    ax8.grid(True, alpha=0.3)

    # Add values on bars
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax8.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                f'{value:.4f}', ha='center', va='bottom', fontsize=8)

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    results_df, sequences_dict = run_pi_evolution_test()
```
