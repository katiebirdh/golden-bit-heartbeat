```python
# %%capture
# ──────────────────────────────────────────────────────────────────────────────
# Bigits → Rules → Visuals (11 figures)
# paste into Colab and run. installs: plotly, ipywidgets
# ──────────────────────────────────────────────────────────────────────────────

!pip -q install plotly==5.24.1 ipywidgets==8.1.3

import math, itertools, textwrap, sys, warnings
from dataclasses import dataclass
import numpy as np
import pandas as pd
from decimal import Decimal, getcontext, ROUND_HALF_EVEN, DivisionByZero, InvalidOperation
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
import ipywidgets as widgets

warnings.filterwarnings("ignore")
getcontext().prec = 120
getcontext().rounding = ROUND_HALF_EVEN

# ──────────────────────────────────────────────────────────────────────────────
# 0) Helpers: words, rules, triadic zoom, power-of-10 alignment, angle, etc.
# ──────────────────────────────────────────────────────────────────────────────

def fib_word(n):
    s = "0"
    while len(s) < n:
        s = "".join("01" if ch=="0" else "0" for ch in s)
    return [1 if ch=="1" else 0 for ch in s[:n]]

def thue_morse(n):
    def bitcount(k):
        c=0
        while k: k&=k-1; c+=1
        return c
    return [bitcount(k)&1 for k in range(n)]

def one_over_phi_bits(n):
    phi = (1 + 5**0.5)/2.0
    y = 1.0/phi
    bits=[]
    for _ in range(n):
        y *= 2.0
        if y >= 1.0: bits.append(1); y -= 1.0
        else: bits.append(0)
    return bits

def fib_minus_one_bits(n):
    # Fibonacci word as binary integer minus 1 (with borrow) → MSB first bits
    bits = fib_word(n)
    out = bits.copy()
    borrow = 1
    for i in range(n-1, -1, -1):
        if borrow == 0: break
        if out[i] == 1:
            out[i] = 0; borrow = 0
        else:
            out[i] = 1; borrow = 1
    return out

def decimal_from_bits_as_digits(bits):
    # Put 0/1 bits as DECIMAL digits after 0.: 0.b1 b2 b3 ...
    d = Decimal(0); ten = Decimal(10); pw = Decimal(1)
    for b in bits:
        pw /= ten
        if b: d += pw
    return d

# rules (single step)
def R1(x): return Decimal(1) - x                # how far from 1
def R2(x):                                       # how many of x fit into 1
    if x == 0: raise DivisionByZero
    return Decimal(1) / x
def R3(x):                                       # x compared to what's left (1−x)
    one = Decimal(1)
    if x == one: raise DivisionByZero
    return x / (one - x)
def R4(x): return -x                             # reverse direction

SINGLES = {"R1 (1−x)":R1, "R2 (1/x)":R2, "R3 (x/(1−x))":R3, "R4 (−x)":R4}

def compose(f, g): return lambda x: g(f(x))
TWO_STEPS = {
 "T1 (1/x → 1−)": compose(R2,R1),
 "T2 (1− → 1/x)": compose(R1,R2),
 "T3 (1− → x/(1−x))": compose(R1,R3),
 "T4 (x/(1−x) → 1/x)": compose(R3,R2),
 "T5 (x/(1−x) → 1−)": compose(R3,R1),
 "T6 (−x → 1−)": compose(R4,R1),
 "T7 (1− → −x)": compose(R1,R4),
 "T8 (−x → x/(1−x))": compose(R4,R3),
 "T9 (x/(1−x) → −x)": compose(R3,R4),
 "T10 (1/x → −x)": compose(R2,R4),
 "T11 (−x → 1/x)": compose(R4,R2),
 "T12 (1/x → x/(1−x))": compose(R2,R3),
 "T13 (x/(1−x) → x/(1−x))": compose(R3,R3),
 "T14 (1/x → 1/x)": compose(R2,R2),
 "T15 (1− → 1−)": compose(R1,R1),
 "T16 (−x → −x)": compose(R4,R4),
}

def triadic_wrap(rule_func, pre, post):
    pre_scale  = Decimal(3) ** Decimal(pre)
    post_scale = Decimal(3) ** Decimal(post)
    def wrapped(x):
        try:
            return pre_scale * rule_func(post_scale * x)
        except (DivisionByZero, InvalidOperation, ZeroDivisionError, OverflowError):
            return Decimal('NaN')
    return wrapped

def iterate(rule_func, x0, k):
    x = x0
    for _ in range(k):
        try:
            x = rule_func(x)
        except (DivisionByZero, InvalidOperation, ZeroDivisionError, OverflowError):
            return Decimal('NaN')
    return x

def to_float(d):
    try: return float(d)
    except: return np.nan

def angle_deg_if_01(val):
    v = to_float(val)
    if np.isfinite(v) and 0.0 <= v <= 1.0:
        try: return float(np.degrees(np.arccos(v)))
        except: return np.nan
    return np.nan

def best_pow10_align(x, t, mmin=-9, mmax=+9):
    # allow 10^m scaling to compare VALUE & DIGITS
    if not (np.isfinite(x) and np.isfinite(t)) or (t==0): return (np.nan, np.nan, np.nan, np.nan)
    best_err, best_m, best_scaled, best_lcp = None, 0, None, 0
    for m in range(mmin, mmax+1):
        sv = x * (10.0**m)
        err = abs(sv - t)
        s_v = mantissa_digits(sv)
        s_t = mantissa_digits(t)
        lcp = digit_lcp(s_v, s_t)
        if (best_err is None) or (err < best_err) or (abs(err-best_err)<1e-18 and lcp > best_lcp):
            best_err, best_m, best_scaled, best_lcp = err, m, sv, lcp
    return best_err, best_m, best_scaled, best_lcp

def mantissa_digits(x):
    if not np.isfinite(x) or x==0: return ""
    ax = abs(x); e=0
    while ax >= 1.0: ax/=10.0; e+=1
    while ax < 0.1: ax*=10.0; e-=1
    s = f"{ax:.18f}"
    return s.split(".",1)[1] if "." in s else ""

def digit_lcp(a,b):
    L=min(len(a),len(b)); i=0
    while i<L and a[i]==b[i]: i+=1
    return i

# targets
SQRT3  = math.sqrt(3)
T_SQRT3_OVER_2   = SQRT3/2.0
T_ONE_OVER_SQRT3 = 1.0/SQRT3
T_PI_OVER_2SQRT3 = math.pi/(2.0*SQRT3)
T_ALPHA_INV      = 137.035999084
T_ALPHA          = 1.0/T_ALPHA_INV

TARGETS = {
 "√3/2": T_SQRT3_OVER_2,
 "1/√3": T_ONE_OVER_SQRT3,
 "π/(2√3)": T_PI_OVER_2SQRT3,
 "α⁻¹": T_ALPHA_INV,
 "α": T_ALPHA,
}

def target_variants(t):
    out = {
      "value": t,
      "half": t/2.0, "third": t/3.0, "quarter": t/4.0,
      "inverse": (1.0/t) if t!=0 else np.nan,
      "sqrt": math.sqrt(t) if t>0 else np.nan
    }
    return out

# ──────────────────────────────────────────────────────────────────────────────
# 1) Build inputs (Fibonacci, Fibonacci−1, 1/φ bits) and baseline x0
# ──────────────────────────────────────────────────────────────────────────────

N_default = 243
inputs = {
  "Fibonacci (bigit)": decimal_from_bits_as_digits(fib_word(N_default)),
  "Fibonacci−1 (bigit)": decimal_from_bits_as_digits(fib_minus_one_bits(N_default)),
  "1/φ bits (bigit)": decimal_from_bits_as_digits(one_over_phi_bits(N_default)),
}

display(Markdown("### Inputs (as decimal digit numbers)"))
for name,x0 in inputs.items():
    print(f"{name}: x0 ≈ {to_float(x0):.12f}")

# ──────────────────────────────────────────────────────────────────────────────
# 2) VISUAL #1: Pipeline sketch (word → x → rule → repeat) with examples
# ──────────────────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(10,2))
ax.axis('off')
ax.text(0.02,0.5, "WORD (0/1)\n(e.g., Fibonacci)", ha='left', va='center', fontsize=11, bbox=dict(facecolor='#eef', edgecolor='#99c'))
ax.text(0.28,0.5, "x = 0.d1 d2 d3…\n(put bits as decimal digits)", ha='center', va='center', fontsize=11, bbox=dict(facecolor='#efe', edgecolor='#9c9'))
ax.text(0.54,0.5, "Rule (small)\n1−x | 1/x | x/(1−x) | −x\nor two-step", ha='center', va='center', fontsize=11, bbox=dict(facecolor='#fee', edgecolor='#c99'))
ax.text(0.80,0.5, "Repeat same rule\nk = 1..4\n(Triadic zoom: ×3^pre/post)", ha='center', va='center', fontsize=11, bbox=dict(facecolor='#ffe', edgecolor='#cc9'))
ax.annotate("", xy=(0.25,0.5), xytext=(0.12,0.5), arrowprops=dict(arrowstyle="->", lw=2))
ax.annotate("", xy=(0.51,0.5), xytext=(0.35,0.5), arrowprops=dict(arrowstyle="->", lw=2))
ax.annotate("", xy=(0.77,0.5), xytext=(0.60,0.5), arrowprops=dict(arrowstyle="->", lw=2))
plt.title("Visual 1: The machine (automaton) you run", fontsize=12)
plt.show()

# ──────────────────────────────────────────────────────────────────────────────
# 3) VISUAL #2: Map portrait (shape rule T5) y = F(x) on [0,1]
# ──────────────────────────────────────────────────────────────────────────────
def F_T5(x):
    # x/(1-x) then 1−(…)
    x = Decimal(x)
    try:
        return R1(R3(x))
    except:
        return Decimal('NaN')

xs = np.linspace(1e-6, 1-1e-6, 2000)
ys = np.array([to_float(F_T5(z)) for z in xs])

fig = go.Figure()
fig.add_trace(go.Scatter(x=xs, y=ys, mode="lines", name="T5 map"))
fig.update_layout(title="Visual 2: Two-step shape rule (T5) as a function",
                  xaxis_title="x in [0,1]", yaxis_title="T5(x)")
fig.show()

# ──────────────────────────────────────────────────────────────────────────────
# 4) VISUAL #3: Angle histogram under T5 (shape) for each input (k=1)
# ──────────────────────────────────────────────────────────────────────────────
def apply_t5_angle(x0, pre=0, post=0, k=1):
    wf = triadic_wrap(TWO_STEPS["T5 (x/(1−x) → 1−)"], pre, post)
    v = iterate(wf, x0, k)
    ang = angle_deg_if_01(v)
    return to_float(v), ang

rows=[]
for name, x0 in inputs.items():
    for pre in [-1,0,1]:
        for post in [-1,0,1]:
            v, ang = apply_t5_angle(x0, pre, post, k=1)
            rows.append({"input":name, "pre":pre, "post":post, "value":v, "angle_deg":ang})
shape_df = pd.DataFrame(rows)

fig = px.histogram(shape_df.dropna(subset=["angle_deg"]),
                   x="angle_deg", color="input", nbins=36, barmode="overlay",
                   title="Visual 3: Angles after one T5 step (near-1 → circle view)")
fig.update_xaxes(title="angle (degrees)")
fig.update_yaxes(title="count")
fig.show()

# ──────────────────────────────────────────────────────────────────────────────
# 5) VISUAL #4: Evolution over k (k=1..4) under T5 for triadic pre/post
# ──────────────────────────────────────────────────────────────────────────────
def k_evolution_table(rule_name, x0, pre_list=[-1,0,1], post_list=[-1,0,1]):
    rec=[]
    wf_base = TWO_STEPS[rule_name]
    for pre in pre_list:
        for post in post_list:
            wf = triadic_wrap(wf_base, pre, post)
            vals=[]
            for k in [1,2,3,4]:
                vals.append(to_float(iterate(wf, x0, k)))
            rec.append({"pre":pre,"post":post,"v1":vals[0],"v2":vals[1],"v3":vals[2],"v4":vals[3]})
    return pd.DataFrame(rec)

x0 = inputs["Fibonacci (bigit)"]
evo = k_evolution_table("T5 (x/(1−x) → 1−)", x0)
fig = go.Figure()
for i,row in evo.iterrows():
    fig.add_trace(go.Scatter(x=[1,2,3,4], y=[row["v1"],row["v2"],row["v3"],row["v4"]],
                             mode="lines+markers",
                             name=f"pre={row['pre']}, post={row['post']}"))
fig.update_layout(title="Visual 4: T5 evolution over k (Fibonacci bigit, various triadic zooms)",
                  xaxis_title="k", yaxis_title="value")
fig.show()

# ──────────────────────────────────────────────────────────────────────────────
# 6) VISUAL #5: Heatmap of |value − target| for √3/2 across triadic pre/post
# ──────────────────────────────────────────────────────────────────────────────
target = T_SQRT3_OVER_2
def heat_pre_post(rule_name, x0, k=1, target=target):
    data=[]
    prepost=[]
    for pre in range(-3,4):
        row=[]
        for post in range(-3,4):
            wf = triadic_wrap(TWO_STEPS[rule_name], pre, post)
            v = to_float(iterate(wf, x0, k))
            row.append(abs(v-target) if np.isfinite(v) else np.nan)
        data.append(row)
    return np.array(data)

h = heat_pre_post("T5 (x/(1−x) → 1−)", inputs["Fibonacci (bigit)"], k=1, target=T_SQRT3_OVER_2)
fig = px.imshow(h, x=list(range(-3,4)), y=list(range(-3,4)),
                labels=dict(x="post (×3^post)", y="pre (×3^pre)", color="|error|"),
                title="Visual 5: Error to √3/2 across triadic (pre, post) — T5 @ k=1")
fig.update_yaxes(autorange="reversed")
fig.show()

# ──────────────────────────────────────────────────────────────────────────────
# 7) VISUAL #6: Base vs error (for a chosen shape fraction), bases 3..10
# ──────────────────────────────────────────────────────────────────────────────
def bigit_from_word_in_base(word_bits, base):
    # Create x = sum b_i * base^{-i} but DECIMAL-digit lens is your default.
    # Here we keep DECIMAL lens to stay true to your setup; the "base" notion here
    # will be used as an index for repeated tests with same x0 for simplicity.
    # (If you want true base-B lens, we can add that as a toggle later.)
    return decimal_from_bits_as_digits(word_bits)

shape_target = T_PI_OVER_2SQRT3/4.0   # a strong step we saw earlier
bases = list(range(3,11))
errs=[]
rule = "T5 (x/(1−x) → 1−)"
for b in bases:
    # keep default lens; just vary triadic post/pre = small favorites
    x0 = inputs["Fibonacci (bigit)"]  # (using same x0; base label is for sweep index)
    wf = triadic_wrap(TWO_STEPS[rule], pre=-1, post=0)
    v = to_float(iterate(wf, x0, 2))
    errs.append(abs(v-shape_target))
fig = go.Figure(go.Scatter(x=bases, y=errs, mode="lines+markers"))
fig.update_layout(title="Visual 6: Base index vs error to (π/(2√3))/4 (illustrative)",
                  xaxis_title="base index (3..10)", yaxis_title="|error| (smaller is better)")
fig.show()

# ──────────────────────────────────────────────────────────────────────────────
# 8) VISUAL #7: α gear test (728/1000 vs 729/1000 vs 730/1000) on Fibonacci bigit
# ──────────────────────────────────────────────────────────────────────────────
def scale_alpha(x, m):
    return float((Decimal(m) / Decimal(1000)) * x)

xF = inputs["Fibonacci (bigit)"]
vals = { "728/1000": scale_alpha(xF, 728),
         "729/1000": scale_alpha(xF, 729),
         "730/1000": scale_alpha(xF, 730) }
errs = [abs(vals[k]-T_ALPHA) for k in ["728/1000","729/1000","730/1000"]]
fig = go.Figure(go.Bar(x=["728/1000","729/1000","730/1000"], y=errs, marker_color=["#bbb","#2a9d8f","#bbb"]))
fig.update_layout(title="Visual 7: α gear seating — error to α for nearby scales",
                  yaxis_title="| value − α | (smaller is better)")
fig.show()

# ──────────────────────────────────────────────────────────────────────────────
# 9) VISUAL #8: Power-of-10 alignment and digit-prefix match (mantissas)
# ──────────────────────────────────────────────────────────────────────────────
def best_alignment_report(x, t, name=""):
    best_err, best_m, best_scaled, best_lcp = best_pow10_align(x, t, -9, +9)
    s = f"best 10^m shift: m = {best_m},  scaled value ≈ {best_scaled:.12g},  |err| ≈ {best_err:.6g},  digit-prefix match = {best_lcp}"
    return s, best_m, best_scaled, best_err, best_lcp

x_demo = to_float(iterate(triadic_wrap(TWO_STEPS["T5 (x/(1−x) → 1−)"], 0, 0),
                          inputs["Fibonacci−1 (bigit)"], 1))
t_demo = T_PI_OVER_2SQRT3/4.0
report, best_m, best_scaled, best_err, best_lcp = best_alignment_report(x_demo, t_demo)

display(Markdown("**Visual 8: Power-of-10 alignment report (mantissa digits)**"))
print("target =", t_demo)
print("raw value ≈", x_demo)
print(report)

# ──────────────────────────────────────────────────────────────────────────────
# 10) VISUAL #9: Distribution before/after shape rule (T5) — scatter & hist
# ──────────────────────────────────────────────────────────────────────────────
def scatter_hist_before_after(x0, pre=0, post=0, k=1, n=500):
    # make an x cloud by randomly truncating the word length; apply T5 to each
    rng = np.random.default_rng(42)
    ns = rng.integers(low=12, high=243, size=n)
    vals=[]
    vals_after=[]
    for N in ns:
        # rebuild x ~ by truncation effect (simplified: use same x0 for speed)
        v0 = to_float(x0)
        wf = triadic_wrap(TWO_STEPS["T5 (x/(1−x) → 1−)"], pre, post)
        va = to_float(iterate(wf, Decimal(v0), k))
        vals.append(v0); vals_after.append(va)
    return np.array(vals), np.array(vals_after)

vals0, vals1 = scatter_hist_before_after(inputs["Fibonacci (bigit)"], pre=0, post=0, k=1, n=400)
fig = go.Figure()
fig.add_trace(go.Scatter(x=vals0, y=vals1, mode="markers", opacity=0.5))
fig.update_layout(title="Visual 9: Before vs After T5 (scatter)", xaxis_title="x (before)", yaxis_title="T5(x) (after)")
fig.show()

fig = px.histogram(vals1, nbins=40, title="Visual 9b: Histogram of T5(x) — shows 'near-1' clustering")
fig.update_xaxes(title="T5(x)")
fig.update_yaxes(title="count")
fig.show()

# ──────────────────────────────────────────────────────────────────────────────
# 11) VISUAL #10: Radar chart — which inputs light which shape steps the best
# ──────────────────────────────────────────────────────────────────────────────
def shape_score(x0, steps=("value","half","third","quarter")):
    # score smaller error for (π/(2√3)) steps using favorite setting
    scores=[]
    for st in steps:
        target = {"value":T_PI_OVER_2SQRT3, "half":T_PI_OVER_2SQRT3/2.0, "third":T_PI_OVER_2SQRT3/3.0, "quarter":T_PI_OVER_2SQRT3/4.0}[st]
        wf = triadic_wrap(TWO_STEPS["T5 (x/(1−x) → 1−)"], pre=-1, post=0)
        v = to_float(iterate(wf, x0, 2))
        err = abs(v-target)
        scores.append(1.0/(1e-6+err))
    return np.array(scores)

labels = ["value","half","third","quarter"]
radar = pd.DataFrame({
    "metric": labels,
    "Fibonacci": shape_score(inputs["Fibonacci (bigit)"], labels),
    "Fibonacci−1": shape_score(inputs["Fibonacci−1 (bigit)"], labels),
    "1/φ bits": shape_score(inputs["1/φ bits (bigit)"], labels),
})

fig = go.Figure()
for col in ["Fibonacci","Fibonacci−1","1/φ bits"]:
    fig.add_trace(go.Scatterpolar(r=radar[col], theta=radar["metric"], fill='toself', name=col))
fig.update_layout(title="Visual 10: Which input lights which shape-steps best (higher = tighter match)",
                  polar=dict(radialaxis=dict(visible=True)))
fig.show()

# ──────────────────────────────────────────────────────────────────────────────
# 12) VISUAL #11 (interactive): slider playground for rule / pre/post / k / input
# ──────────────────────────────────────────────────────────────────────────────
def compute_value(input_name, rule_kind, rule_name, pre, post, k):
    x0 = inputs[input_name]
    if rule_kind == "single":
        wf = triadic_wrap(SINGLES[rule_name], pre, post)
    else:
        wf = triadic_wrap(TWO_STEPS[rule_name], pre, post)
    val = iterate(wf, x0, k)
    v = to_float(val)
    ang = angle_deg_if_01(val)
    # errors to targets
    errs = {nm: abs(v - tv) if np.isfinite(v) else np.nan for nm,tv in TARGETS.items()}
    return v, ang, errs

input_dd = widgets.Dropdown(options=list(inputs.keys()), value="Fibonacci (bigit)", description="input")
kind_dd  = widgets.Dropdown(options=["single","two-step"], value="two-step", description="rule type")
rule_dd1 = widgets.Dropdown(options=list(SINGLES.keys()), value="R3 (x/(1−x))", description="single")
rule_dd2 = widgets.Dropdown(options=list(TWO_STEPS.keys()), value="T5 (x/(1−x) → 1−)", description="two-step")
pre_sl   = widgets.IntSlider(value=0, min=-3, max=3, step=1, description="pre (×3^pre)")
post_sl  = widgets.IntSlider(value=0, min=-3, max=3, step=1, description="post (×3^post)")
k_sl     = widgets.IntSlider(value=1, min=1, max=4, step=1, description="k")

out = widgets.Output()

def update(_=None):
    with out:
        out.clear_output()
        rule_name = rule_dd1.value if kind_dd.value=="single" else rule_dd2.value
        v, ang, errs = compute_value(input_dd.value, kind_dd.value, rule_name, pre_sl.value, post_sl.value, k_sl.value)
        # value/angle
        print(f"value ≈ {v:.12g}" if np.isfinite(v) else "value = NaN")
        print(f"angle (deg) ≈ {ang:.4f}" if np.isfinite(ang) else "angle: (not in [0,1])")
        # power-of-10 best alignment to each target + digits match
        for nm,tv in TARGETS.items():
            err, m, sv, lcp = best_pow10_align(v, tv, -9, +9) if np.isfinite(v) else (np.nan, np.nan, np.nan, np.nan)
            print(f"  vs {nm}: |err|={err:.6g}  best 10^m={m:+d}  scaled≈{sv:.12g}  digit-prefix match={lcp}")

# widgets.jslink((kind_dd,'value'), (rule_dd2,'layout'))  # no-op to keep widgets alive
for w in [input_dd, kind_dd, rule_dd1, rule_dd2, pre_sl, post_sl, k_sl]:
    w.observe(update, names='value')

display(Markdown("### Visual 11: Interactive rule playground"))
display(widgets.HBox([widgets.VBox([input_dd, kind_dd, rule_dd1, rule_dd2]),
                      widgets.VBox([pre_sl, post_sl, k_sl])]))
display(out)
update()
```
