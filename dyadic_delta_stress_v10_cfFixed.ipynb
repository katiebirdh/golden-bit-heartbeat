```python
n_max=40
powers=[2]
bases_int=[2]
CONST={'pi': mp.pi, 'phi': (1+mp.sqrt(5))/2}
```

```python
def contfrac(x, k=10):
    digits=[]; y=mp.mpf(x)
    for _ in range(k):
        a=int(mp.floor(y)); digits.append(a)
        y -= a
        if y==0: break
        y = 1/y
    digits += [0]*(k-len(digits))
    return digits
```

```python
ref_kappa = {'pi':1, 'phi':1}
X=[]; y=[]
for name in ref_kappa:
    cf10 = contfrac(mp.sqrt(CONST[name]),10)
    X.append(cf10); y.append(math.log(ref_kappa[name]))
print('digits', X)
```

```python
plt.scatter([d[0] for d in X], y)
plt.xlabel('first CF digit of sqrt(C)')
plt.ylabel('log F(C)'); plt.show()
```
