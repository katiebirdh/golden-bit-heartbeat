# üß† Golden-Bit Harmonic Residue Explorer

Explore 4 dimensions of symbolic emergence:
- Decimal ‚Üî Recursive folding
- Base-9 harmonic compression
- Fibonacci vs De Bruijn bit density
- Fine-structure residue dynamics

This notebook is a visually rich, self-contained tool to demonstrate that constants like Œ±‚Åª¬π emerge from symbolic residues ‚Äî visibly and repeatedly.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import sqrt, pi, N
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d
```

## üîÅ Base Setup and Recursive Values

```python
phi = (1 + sqrt(5)) / 2
alpha_inv = 137.035999
alpha_scaled = alpha_inv / 1000
residue_delta = 0.013717421 - 0.00137033999
residue_folded = residue_delta / 9
one_over_81 = 1/81
decimal_seed = 0.123456789
```

## üîÑ Visual 1: Spiral Emergence of Harmonic Residues

```python
theta = np.linspace(0, 6 * np.pi, 800)
radii = np.linspace(0.01, 2, 800)

x_spiral = radii * np.cos(theta)
y_spiral = radii * np.sin(theta)

points = {
    "Digits (1‚Äì9)": [i / 9 for i in range(1, 10)],
    "Œ±‚Åª¬π / 1000": alpha_scaled,
    "Œî residue": residue_delta,
    "Œî / 9": residue_folded,
    "1 / 81": one_over_81
}

angles = {
    "Digits (1‚Äì9)": np.linspace(0, 2*np.pi, 9),
    "Œ±‚Åª¬π / 1000": 1.2 * np.pi,
    "Œî residue": 2.4 * np.pi,
    "Œî / 9": 3.1 * np.pi,
    "1 / 81": 3.7 * np.pi
}

scale = 1000
fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(x_spiral, y_spiral, color='lightgray', lw=0.5)

for key in points:
    if isinstance(points[key], list):
        r_vals = np.array(points[key]) * scale
        a_vals = angles[key]
    else:
        r_vals = np.array([points[key] * scale])
        a_vals = np.array([angles[key]])
    x = r_vals * np.cos(a_vals)
    y = r_vals * np.sin(a_vals)
    ax.scatter(x, y, label=key)

ax.set_title("Spiral Map of Symbolic Residues")
ax.axis('equal')
ax.legend()
plt.show()
```

## üåê Visual 2: 3D Spiral Embedding of Constants

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

theta3 = np.linspace(0, 8 * np.pi, 600)
z = np.linspace(0, 1, 600)
r3 = np.linspace(0.05, 1.5, 600)
x3 = r3 * np.cos(theta3)
y3 = r3 * np.sin(theta3)

ax.plot3D(x3, y3, z, color='gray', alpha=0.4)

points_3d = {
    "Œ±‚Åª¬π / 1000": (alpha_scaled * 400, 1.2 * np.pi, 0.2),
    "Œî": (residue_delta * 400, 2.4 * np.pi, 0.5),
    "Œî / 9": (residue_folded * 2000, 3.1 * np.pi, 0.8),
    "1 / 81": (one_over_81 * 1000, 3.7 * np.pi, 0.65)
}

for label, (r, angle, z_val) in points_3d.items():
    x = r * np.cos(angle)
    y = r * np.sin(angle)
    ax.scatter(x, y, z_val, label=label)

ax.set_title("3D Spiral with Residual Constants")
ax.legend()
plt.show()
```

## üßÆ Visual 3: Bit Growth of Fibonacci vs De Bruijn

```python
def fib_density(n):
    return n / float(N(1 / phi**2))

fibs = [3**i for i in range(1, 7)]
bits_needed = [fib_density(p) for p in fibs]
debruijn_lengths = [2**i for i in range(1, 7)]

plt.figure(figsize=(8,5))
plt.plot(fibs, bits_needed, 'o-', label='Fibonacci bits for 3^n ones')
plt.plot(fibs, debruijn_lengths, 's--', label='De Bruijn (2^n length)')
plt.xlabel('n')
plt.ylabel('Bits')
plt.title('Fibonacci vs De Bruijn Bit Growth')
plt.grid(True)
plt.legend()
plt.show()
```

## üß≤ Visual 4: Polar Encoding of Recursive Compression

```python
residues = [alpha_scaled, residue_delta, residue_folded, one_over_81]
labels = ["Œ±‚Åª¬π / 1000", "Œî", "Œî / 9", "1 / 81"]
radii = np.array(residues) * 1000
angles = np.linspace(0, 2*np.pi, len(radii))

fig, ax = plt.subplots(figsize=(6,6), subplot_kw={'projection':'polar'})
bars = ax.bar(angles, radii, width=0.5, bottom=0.0)

for bar, label in zip(bars, labels):
    bar.set_alpha(0.6)
    bar.set_label(label)

ax.set_title("Radial Compression of Emergent Constants")
ax.legend(loc='lower center')
plt.show()
```
