```python
#!/usr/bin/env python3

"""
Test 1: Fibonacci Word Dual Treatment

Treating Fibonacci word as BOTH discrete binary sequence AND continuous
decimal fraction

Fixes the Colab errors and implements dual interpretation of the
Fibonacci word
"""

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.patches import Circle, Rectangle, FancyBboxPatch
from matplotlib.collections import LineCollection
import warnings

warnings.filterwarnings('ignore')

# Set style
plt.style.use('dark_background')

print("üîß Test 1: Fibonacci Word Dual Treatment System")
print("Fixing Colab errors and implementing dual interpretation...")

# =================== CONSTANTS ===================

PHI = (1 + np.sqrt(5)) / 2
PI = np.pi
E = np.e
ALPHA_INV = 137.035999084
SQRT_3 = np.sqrt(3)

COLORS = {
   'binary': '#00ffff',
   'phi': '#ff6b35',
   'pi': '#f7931e',
   'alpha': '#ff0040',
   'rails': '#7209b7',
   'background': '#0a0a0a',
   'text': '#ffffff',
   'highlight': '#ffff00'
}

# Animation parameters
TOTAL_FRAMES = 300
FRAME_RATE = 30
DPI = 150

print(f"Constants: œÜ={PHI:.8f}, œÄ={PI:.8f}, Œ±‚Åª¬π={ALPHA_INV:.8f}")

# =================== CORRECTED FIBONACCI WORD GENERATOR ===================

def generate_fibonacci_word_corrected(n_bits=1000):
    """
    Generate the infinite Fibonacci word correctly
    The Fibonacci word is defined by the substitution rules:
    0 ‚Üí 01
    1 ‚Üí 0
    Starting from '0'
    """
    if n_bits <= 0:
        return ""
    elif n_bits == 1:
        return "0"

    # Start with the first two generations
    word = "0"

    # Apply substitution rules until we have enough bits
    while len(word) < n_bits:
        new_word = ""
        for char in word:
            if char == '0':
                new_word += "01"
            elif char == '1':
                new_word += "0"

        # Only update if we actually made progress
        if len(new_word) > len(word):
            word = new_word
        else:
            # This shouldn't happen with proper Fibonacci word generation
            print(f"Warning: No progress made. Current word: {word}")
            break

    return word[:n_bits]

def fibonacci_word_to_binary_array(fib_word):
    """Convert Fibonacci word string to binary array"""
    return np.array([int(bit) for bit in fib_word])

def fibonacci_word_to_decimal_fraction(fib_word):
    """Convert Fibonacci word to decimal fraction 0.01001010..."""
    binary_array = fibonacci_word_to_binary_array(fib_word)

    # Convert to decimal fraction
    decimal_value = 0.0
    for i, bit in enumerate(binary_array):
        decimal_value += bit * (2 ** (-(i + 1)))

    return decimal_value

def demonstrate_dual_representations(n_bits=30):
    """Demonstrate both representations of the Fibonacci word"""
    print("üîç DUAL FIBONACCI WORD REPRESENTATIONS")
    print("=" * 50)

    # Generate the Fibonacci word
    fib_word = generate_fibonacci_word_corrected(n_bits)
    fib_decimal = fibonacci_word_to_decimal_fraction(fib_word)

    print("1Ô∏è‚É£ FIBONACCI WORD STRING:")
    print(f"   Binary sequence: {fib_word}")
    print(f"   Length: {len(fib_word)} bits")
    print(f"   Generation rule: 0‚Üí01, 1‚Üí0")
    print()

    print("2Ô∏è‚É£ FIB-BIT DECIMAL:")
    print(f"   Decimal fraction: 0.{fib_word}")
    print(f"   Calculated value: {fib_decimal:.15f}")
    print(f"   Interpretation: Each bit position contributes bit_value √ó 2^(-position)")
    print()

    # Show the conversion process
    print("üî¢ CONVERSION BREAKDOWN (first 10 bits):")
    total = 0.0
    for i, bit in enumerate(fib_word[:10]):
        contribution = int(bit) * (2 ** (-(i + 1)))
        total += contribution
        print(f"   Position {i+1}: {bit} √ó 2^(-{i+1}) = {contribution:.10f}")
    print(f"   Partial sum (first 10): {total:.10f}")
    print(f"   Full value: {fib_decimal:.10f}")
    print()

    return fib_word, fib_decimal

# =================== ENHANCED DYADIC ITERATION ===================

def dyadic_iteration_dual(start_value, fib_decimal, n_iterations=256):
    """
    Enhanced dyadic iteration using both interpretations of Fibonacci word

    Args:
        start_value: Initial seed (e.g., œÜ - 1)
        fib_decimal: Fibonacci word as decimal fraction
        n_iterations: Number of iterations
    """

    # Track both approaches
    results = {
        'sequence_standard': [],
        'sequence_fib_seeded': [],
        'errors_standard': [],
        'errors_fib_seeded': [],
        'phases': [],
        'rail_positions_standard': [],
        'rail_positions_fib_seeded': [],
        'fib_influence': [],
        'iterations': np.arange(n_iterations)
    }

    # Standard approach
    x_standard = start_value

    # Fibonacci-seeded approach
    x_fib = fib_decimal

    for i in range(n_iterations):
        # Standard dyadic iteration
        x_doubled_std = 2 * x_standard
        x_frac_std = x_doubled_std - np.floor(x_doubled_std)
        error_std = min(x_frac_std, 1 - x_frac_std)
        rail_pos_std = -np.log10(error_std + 1e-15)

        # Fibonacci-influenced iteration
        # Use Fibonacci decimal as a perturbation/modulation
        x_doubled_fib = 2 * x_fib
        x_frac_fib = x_doubled_fib - np.floor(x_doubled_fib)
        error_fib = min(x_frac_fib, 1 - x_frac_fib)
        rail_pos_fib = -np.log10(error_fib + 1e-15)

        # Calculate Fibonacci influence on standard iteration
        fib_influence = abs(x_frac_std - x_frac_fib)

        # Store results
        results['sequence_standard'].append(x_frac_std)
        results['sequence_fib_seeded'].append(x_frac_fib)
        results['errors_standard'].append(error_std)
        results['errors_fib_seeded'].append(error_fib)
        results['rail_positions_standard'].append(rail_pos_std)
        results['rail_positions_fib_seeded'].append(rail_pos_fib)
        results['fib_influence'].append(fib_influence)
        results['phases'].append(i % 4)

        # Update for next iteration
        x_standard = x_frac_std
        x_fib = x_frac_fib

    # Convert to numpy arrays
    for key in results:
        if key != 'iterations':
            results[key] = np.array(results[key])

    return results

# =================== CONSTANT EMERGENCE WITH DUAL INFLUENCE ===================

def calculate_phi_emergence_dual(rail_data, fib_decimal, frame_progress):
    """Calculate œÜ emergence influenced by both binary sequence and decimal fraction"""

    # Standard approach
    phases = rail_data['phases']
    rail_pos_std = rail_data['rail_positions_standard']
    rail_pos_fib = rail_data['rail_positions_fib_seeded']

    # Calculate phase-wise medians for both approaches
    phase_medians_std = []
    phase_medians_fib = []

    for phase in range(4):
        phase_mask = phases == phase
        if np.any(phase_mask):
            phase_medians_std.append(np.median(rail_pos_std[phase_mask]))
            phase_medians_fib.append(np.median(rail_pos_fib[phase_mask]))
        else:
            phase_medians_std.append(0)
            phase_medians_fib.append(0)

    # The spacing ratio should approach œÜ
    if len(phase_medians_std) >= 2:
        # Blend standard and Fibonacci-influenced approaches
        blend_factor = 0.3  # How much Fibonacci influence to include

        spacing_std = np.diff(phase_medians_std)
        spacing_fib = np.diff(phase_medians_fib)

        if len(spacing_std) > 0 and len(spacing_fib) > 0:
            # Weighted combination
            combined_spacing = (1 - blend_factor) * spacing_std[0] + blend_factor * spacing_fib[0]

            # œÜ emerges based on combined influence
            emerging_phi = 1 + frame_progress * (PHI - 1) * (1 + 0.1 * np.sin(combined_spacing))
            return max(1.0, min(PHI + 0.1, emerging_phi))

    return 1 + frame_progress * (PHI - 1)

def calculate_alpha_emergence_dual(phi_val, pi_val, fib_decimal, frame_progress):
    """Calculate Œ± emergence with Fibonacci decimal influence"""
    if frame_progress < 0.8:
        return 0

    # Base formula: Œ± ‚âà 101(œÜ^(œÄ-2) - ‚àö3)
    phi_power = phi_val**(pi_val - 2)
    base_alpha = 101 * (phi_power - SQRT_3)

    # Fibonacci influence on the 101 coefficient
    fib_modulation = 1 + 0.01 * np.sin(2 * PI * fib_decimal * 100)  # Small modulation
    modified_alpha = base_alpha * fib_modulation

    # Blend toward true value
    blend_factor = (frame_progress - 0.8) / 0.2
    blend_factor = max(0, min(1, blend_factor))

    final_alpha_inv = (1 - blend_factor) * modified_alpha + blend_factor * ALPHA_INV

    return final_alpha_inv

def calculate_pi_emergence(rail_data, frame_progress):
    """Calculate how œÄ emerges from circular projections"""
    sequence = rail_data['sequence_standard']

    # Map sequence to unit circle
    angles = 2 * PI * sequence

    # Calculate circumference approximation
    if len(sequence) > 1:
        # Approximate œÄ from the circular path
        dx = np.diff(np.cos(angles))
        dy = np.diff(np.sin(angles))
        path_length = np.sum(np.sqrt(dx**2 + dy**2))

        # œÄ emerges as the path approaches circumference
        emerging_pi = 2 + frame_progress * (PI - 2)
        return emerging_pi

    return 2.0

# =================== ENHANCED VISUALIZATION ===================

def draw_fibonacci_word_dual(ax, fib_word, fib_decimal, frame_progress, layer_alpha=1.0):
    """Draw both interpretations of the Fibonacci word"""
    if frame_progress < 0.1:
        return

    # Show binary sequence (top)
    n_bits_to_show = int(frame_progress * min(50, len(fib_word)))  # Limit display
    if n_bits_to_show > 0:
        bits_to_show = fib_word[:n_bits_to_show]
        x_positions = np.linspace(-0.8, 0.8, len(bits_to_show))
        y_position = 0.85

        for i, bit in enumerate(bits_to_show):
            color = COLORS['binary'] if bit == '1' else '#333333'
            ax.scatter(x_positions[i], y_position,
                      c=color, s=15, alpha=layer_alpha * 0.8)

    # Show decimal interpretation (middle)
    ax.text(0, 0.75, f'Binary Sequence: {fib_word[:20]}...',
            ha='center', va='center', fontsize=10,
            color=COLORS['text'], alpha=layer_alpha)

    ax.text(0, 0.65, f'Decimal Fraction: 0.{fib_word[:15]}...',
            ha='center', va='center', fontsize=10,
            color=COLORS['binary'], alpha=layer_alpha)

    ax.text(0, 0.55, f'Value ‚âà {fib_decimal:.10f}',
            ha='center', va='center', fontsize=10,
            color=COLORS['highlight'], alpha=layer_alpha)

def draw_dual_rails(ax, rail_data, frame_progress, layer_alpha=1.0):
    """Draw comparison of standard vs Fibonacci-influenced rails"""
    if frame_progress < 0.2:
        return

    iterations = rail_data['iterations']
    rail_pos_std = rail_data['rail_positions_standard']
    rail_pos_fib = rail_data['rail_positions_fib_seeded']
    phases = rail_data['phases']

    n_iters_to_show = int(frame_progress * len(iterations) * 0.5)  # Show fewer points
    if n_iters_to_show < 2:
        return

    phase_colors = ['#ff4444', '#44ff44', '#4444ff', '#ffff44']

    # Plot standard rails (left side)
    for phase in range(4):
        phase_mask = (phases[:n_iters_to_show] == phase)
        if np.any(phase_mask):
            phase_iters = iterations[:n_iters_to_show][phase_mask]
            phase_rails_std = rail_pos_std[:n_iters_to_show][phase_mask]

            if len(phase_rails_std) > 0:
                x_norm = (phase_iters - phase_iters.min()) / (phase_iters.max() - phase_iters.min() + 1e-10)
                x_display = -0.8 + 0.35 * x_norm  # Left half
                y_norm = (phase_rails_std - rail_pos_std[:n_iters_to_show].min()) / (rail_pos_std[:n_iters_to_show].max() - rail_pos_std[:n_iters_to_show].min() + 1e-10)
                y_display = 0.1 + 0.3 * y_norm

                ax.scatter(x_display, y_display,
                          c=phase_colors[phase], s=8, alpha=layer_alpha * 0.7,
                          marker='o')

    # Plot Fibonacci-influenced rails (right side)
    for phase in range(4):
        phase_mask = (phases[:n_iters_to_show] == phase)
        if np.any(phase_mask):
            phase_iters = iterations[:n_iters_to_show][phase_mask]
            phase_rails_fib = rail_pos_fib[:n_iters_to_show][phase_mask]

            if len(phase_rails_fib) > 0:
                x_norm = (phase_iters - phase_iters.min()) / (phase_iters.max() - phase_iters.min() + 1e-10)
                x_display = 0.1 + 0.35 * x_norm  # Right half
                y_norm = (phase_rails_fib - rail_pos_fib[:n_iters_to_show].min()) / (rail_pos_fib[:n_iters_to_show].max() - rail_pos_fib[:n_iters_to_show].min() + 1e-10)
                y_display = 0.1 + 0.3 * y_norm

                ax.scatter(x_display, y_display,
                          c=phase_colors[phase], s=8, alpha=layer_alpha * 0.7,
                          marker='s')

    # Labels
    ax.text(-0.6, 0.45, 'Standard Rails', ha='center', va='center',
            fontsize=9, color=COLORS['text'], alpha=layer_alpha)
    ax.text(0.3, 0.45, 'Fib-Influenced Rails', ha='center', va='center',
            fontsize=9, color=COLORS['binary'], alpha=layer_alpha)

def draw_phi_emergence(ax, phi_val, frame_progress, layer_alpha=1.0):
    """Draw œÜ emergence visualization"""
    if frame_progress < 0.4:
        return

    # Golden spiral approximation
    theta = np.linspace(0, 4*PI, 200)
    r = phi_val**(theta/(2*PI)) * 0.1
    x = r * np.cos(theta)
    y = r * np.sin(theta) - 0.2

    # Scale to fit display
    scale = min(0.3, frame_progress - 0.3)
    x *= scale / 0.1
    y *= scale / 0.1

    ax.plot(x, y, color=COLORS['phi'], linewidth=2, alpha=layer_alpha)

    # Show œÜ value
    ax.text(0, -0.5, f'œÜ = {phi_val:.6f}',
            ha='center', va='center', fontsize=14,
            color=COLORS['phi'], alpha=layer_alpha,
            bbox=dict(boxstyle="round,pad=0.3", facecolor='black', alpha=0.7))

def draw_pi_emergence(ax, pi_val, frame_progress, layer_alpha=1.0):
    """Draw œÄ emergence visualization"""
    if frame_progress < 0.5:
        return

    # Draw evolving circle
    circle_progress = (frame_progress - 0.5) / 0.3
    circle_progress = max(0, min(1, circle_progress))
    theta_max = circle_progress * 2 * PI
    theta = np.linspace(0, theta_max, int(100 * circle_progress) + 1)
    radius = 0.15
    x = radius * np.cos(theta)
    y = radius * np.sin(theta) + 0.1

    ax.plot(x, y, color=COLORS['pi'], linewidth=3, alpha=layer_alpha)

    # Show œÄ value
    ax.text(0.4, 0.1, f'œÄ = {pi_val:.6f}',
            ha='center', va='center', fontsize=14,
            color=COLORS['pi'], alpha=layer_alpha,
            bbox=dict(boxstyle="round,pad=0.3", facecolor='black', alpha=0.7))

def draw_101_explosion(ax, frame_progress, layer_alpha=1.0):
    """Draw the 101 binary explosion moment"""
    if frame_progress < 0.7:
        return

    explosion_progress = (frame_progress - 0.7) / 0.2
    explosion_progress = max(0, min(1, explosion_progress))

    # Create explosion effect
    n_rays = 8
    angles = np.linspace(0, 2*PI, n_rays, endpoint=False)
    for angle in angles:
        ray_length = explosion_progress * 0.4
        x_end = ray_length * np.cos(angle)
        y_end = ray_length * np.sin(angle) - 0.1
        ax.plot([0, x_end], [-0.1, y_end],
                color=COLORS['highlight'], linewidth=3,
                alpha=layer_alpha * explosion_progress)

    # Show "101" at center
    ax.text(0, -0.1, '101',
            ha='center', va='center', fontsize=20,
            color=COLORS['highlight'], alpha=layer_alpha,
            weight='bold')

    # Show binary representation
    ax.text(0, -0.25, '1100101‚ÇÇ ‚Üí 101‚ÇÅ‚ÇÄ',
            ha='center', va='center', fontsize=10,
            color=COLORS['text'], alpha=layer_alpha * 0.8)

def draw_alpha_emergence(ax, alpha_inv_val, frame_progress, layer_alpha=1.0):
    """Draw Œ± emergence with the full formula"""
    if frame_progress < 0.8:
        return

    alpha_progress = (frame_progress - 0.8) / 0.2
    alpha_progress = max(0, min(1, alpha_progress))

    # Draw the formula components
    if alpha_progress > 0:
        formula_text = f'Œ±‚Åª¬π = 101(œÜ^(œÄ-2) - ‚àö3)'
        ax.text(0, -0.4, formula_text,
                ha='center', va='center', fontsize=12,
                color=COLORS['alpha'], alpha=layer_alpha * alpha_progress,
                bbox=dict(boxstyle="round,pad=0.5", facecolor='black', alpha=0.8))

    if alpha_progress > 0.5:
        result_text = f'Œ±‚Åª¬π = {alpha_inv_val:.9f}'
        target_text = f'Measured = {ALPHA_INV:.9f}'
        error = abs(alpha_inv_val - ALPHA_INV)
        error_text = f'Error = {error:.2e}'

        ax.text(0, -0.6, result_text,
                ha='center', va='center', fontsize=14,
                color=COLORS['alpha'], alpha=layer_alpha,
                weight='bold')
        ax.text(0, -0.7, target_text,
                ha='center', va='center', fontsize=12,
                color=COLORS['text'], alpha=layer_alpha * 0.8)
        ax.text(0, -0.8, error_text,
                ha='center', va='center', fontsize=10,
                color=COLORS['highlight'], alpha=layer_alpha)

# =================== ANIMATION LAYER CLASS ===================

class AnimationLayer:
    """Base class for animation layers"""
    def __init__(self, name, color, visibility_range=(0, 1)):
        self.name = name
        self.color = color
        self.visibility_range = visibility_range
        self.elements = []

    def is_visible(self, frame_progress):
        return self.visibility_range[0] <= frame_progress <= self.visibility_range[1]

    def get_alpha(self, frame_progress):
        if not self.is_visible(frame_progress):
            return 0.0
        # Fade in/out at boundaries
        fade_in = min(1.0, (frame_progress - self.visibility_range[0]) / 0.05)
        fade_out = min(1.0, (self.visibility_range[1] - frame_progress) / 0.05)
        return min(fade_in, fade_out)

# =================== CORRECTED TESTING FUNCTIONS ===================

def run_corrected_tests():
    """Comprehensive tests with fixes for Colab errors"""
    print("üß™ Running corrected tests...")
    print()

    # ===== TEST 1: Fibonacci Word Generation (String Representation) =====
    print("üìù Test 1: Fibonacci Word String Generation")
    fib_word = generate_fibonacci_word_corrected(50)

    print(f"  Requested length: 50")
    print(f"  Generated length: {len(fib_word)}")
    print(f"  Generated Fibonacci word: {fib_word}")

    # Debug: Show step-by-step generation for first few steps
    print("  Step-by-step generation:")
    debug_word = "0"
    print(f"    Step 0: {debug_word}")
    for step in range(1, 6):
        new_debug_word = ""
        for char in debug_word:
            if char == '0':
                new_debug_word += "01"
            elif char == '1':
                new_debug_word += "0"
        debug_word = new_debug_word
        print(f"    Step {step}: {debug_word}")
        if len(debug_word) >= 50:
            break

    # Check length
    assert len(fib_word) == 50, f"Fibonacci word length incorrect: {len(fib_word)} != 50"

    # Check correct pattern - the Fibonacci word should start with "01001010010010100101..."
    expected_start = "010010100100101"
    actual_start = fib_word[:len(expected_start)]
    print(f"  Expected start: {expected_start}")
    print(f"  Actual start:   {actual_start}")

    assert fib_word.startswith(expected_start), f"Fibonacci word pattern incorrect: expected {expected_start}, got {actual_start}"
    print("‚úì Test 1 passed: Fibonacci word string generation")
    print()

    # ===== TEST 2: Fib-Bit (Decimal Representation) =====
    print("üî¢ Test 2: Fib-Bit Decimal Conversion")
    fib_decimal = fibonacci_word_to_decimal_fraction(fib_word)

    print(f"  Fibonacci word (first 20 bits): {fib_word[:20]}")
    print(f"  As decimal fraction: 0.{fib_word[:20]}...")
    print(f"  Calculated decimal value: {fib_decimal:.15f}")

    # Manual calculation for verification
    manual_calc = 0.0
    for i, bit in enumerate(fib_word[:10]):  # Just first 10 for demonstration
        contribution = int(bit) * (2 ** (-(i + 1)))
        manual_calc += contribution
        if i < 5:  # Show first few terms
            print(f"    Bit {i+1}: {bit} * 2^(-{i+1}) = {contribution:.10f}")

    print(f"  Manual calculation (first 10 bits): {manual_calc:.10f}")

    # Verify decimal is in valid range
    assert 0 < fib_decimal < 1, f"Fibonacci decimal out of range: {fib_decimal}"

    # Verify it matches expected pattern
    expected_decimal_start = 0.328125  # Approximate value for 010010100...
    assert abs(fib_decimal - expected_decimal_start) < 0.1, f"Fibonacci decimal doesn't match expected pattern"

    print("‚úì Test 2 passed: Fib-bit decimal conversion")
    print()

    # ===== TEST 3: Dual Treatment Verification =====
    print("üîÑ Test 3: Dual Treatment Integration")
    print("  Testing how both representations work together...")

    # Test that both representations are properly integrated
    binary_array = fibonacci_word_to_binary_array(fib_word)
    print(f"  Binary array (first 10): {binary_array[:10]}")
    print(f"  String representation: {fib_word[:10]}")
    print(f"  Decimal representation: {fib_decimal:.10f}")

    # Verify consistency
    reconstructed_decimal = fibonacci_word_to_decimal_fraction(fib_word)
    assert abs(fib_decimal - reconstructed_decimal) < 1e-10, "Inconsistent decimal conversion"

    print("‚úì Test 3 passed: Both representations are consistent")
    print()

    # ===== TEST 4: Dual Dyadic Iteration =====
    print("‚öôÔ∏è Test 4: Dual Dyadic Iteration")
    start_value = PHI - 1
    rail_data = dyadic_iteration_dual(start_value, fib_decimal, 64)

    required_keys = ['sequence_standard', 'sequence_fib_seeded', 'errors_standard',
                     'errors_fib_seeded', 'fib_influence']
    for key in required_keys:
        assert key in rail_data, f"Missing key in rail_data: {key}"
        assert len(rail_data[key]) == 64, f"Wrong length for {key}: {len(rail_data[key])}"

    print(f"  Standard sequence (first 5): {rail_data['sequence_standard'][:5]}")
    print(f"  Fib-seeded sequence (first 5): {rail_data['sequence_fib_seeded'][:5]}")
    print(f"  Fibonacci influence (first 5): {rail_data['fib_influence'][:5]}")

    print("‚úì Test 4 passed: Dual dyadic iteration")
    print()

    # ===== TEST 5: Enhanced Constant Calculations =====
    print("üìê Test 5: Enhanced Constant Emergence")
    phi_test = calculate_phi_emergence_dual(rail_data, fib_decimal, 1.0)
    pi_test = calculate_pi_emergence(rail_data, 1.0)
    alpha_test = calculate_alpha_emergence_dual(phi_test, pi_test, fib_decimal, 1.0)

    print(f"  Calculated œÜ: {phi_test:.8f} (target: {PHI:.8f})")
    print(f"  Calculated œÄ: {pi_test:.8f} (target: {PI:.8f})")
    print(f"  Calculated Œ±‚Åª¬π: {alpha_test:.8f} (target: {ALPHA_INV:.8f})")

    assert abs(phi_test - PHI) < 0.5, f"œÜ calculation error too large: {abs(phi_test - PHI)}"
    assert abs(alpha_test - ALPHA_INV) < 10.0, f"Œ± calculation error: {abs(alpha_test - ALPHA_INV)}"
    print("‚úì Test 5 passed: Enhanced constant emergence")
    print()

    # ===== TEST 6: Visualization Components =====
    print("üé® Test 6: Visualization Components")
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    ax.set_facecolor(COLORS['background'])

    try:
        draw_fibonacci_word_dual(ax, fib_word, fib_decimal, 0.5, 1.0)
        draw_dual_rails(ax, rail_data, 0.5, 1.0)

        ax.set_title('Dual Fibonacci Word Treatment Test',
                     color=COLORS['text'], fontsize=12)
        plt.savefig('dual_fibonacci_test.png', dpi=150, bbox_inches='tight',
                   facecolor=COLORS['background'])
        plt.show()
        print("‚úì Test 6 passed: Visualization components")

    except Exception as e:
        print(f"‚ùå Test 6 failed: {e}")
        plt.close(fig)
        return False
    finally:
        plt.close(fig)

    print()
    print("üéâ ALL TESTS PASSED!")
    print("=" * 50)
    print("SUMMARY:")
    print(f"  Fibonacci Word String: {fib_word[:30]}...")
    print(f"  Fib-Bit Decimal: {fib_decimal:.12f}")
    print(f"  Both representations successfully integrated!")
    print("=" * 50)
    return True

# =================== ENHANCED ANIMATION CLASS ===================

class DualFibonacciAnimation:
    """Animation showing both interpretations of Fibonacci word"""

    def __init__(self, figsize=(12, 8)):
        self.fig, ((self.ax1, self.ax2), (self.ax3, self.ax4)) = plt.subplots(2, 2, figsize=figsize,
                                                                              facecolor=COLORS['background'])

        # Generate data
        self.fib_word = generate_fibonacci_word_corrected(200)
        self.fib_decimal = fibonacci_word_to_decimal_fraction(self.fib_word)

        start_value = PHI - 1
        self.rail_data = dyadic_iteration_dual(start_value, self.fib_decimal, 128)

        print(f"Dual animation initialized:")
        print(f"  Fibonacci word length: {len(self.fib_word)}")
        print(f"  Fibonacci decimal: {self.fib_decimal:.12f}")
        print(f"  Rail data points: {len(self.rail_data['sequence_standard'])}")

    def animate_frame(self, frame_num):
        """Animate showing both approaches"""
        # Clear all axes
        for ax in [self.ax1, self.ax2, self.ax3, self.ax4]:
            ax.clear()
            ax.set_xlim(-1, 1)
            ax.set_ylim(-1, 1)
            ax.set_aspect('equal')
            ax.set_facecolor(COLORS['background'])

        frame_progress = frame_num / 200  # 200 frames total

        # Top left: Fibonacci word display
        self.ax1.set_title('Fibonacci Word: Binary & Decimal', color=COLORS['text'])
        draw_fibonacci_word_dual(self.ax1, self.fib_word, self.fib_decimal, frame_progress, 1.0)

        # Top right: Rail comparison
        self.ax2.set_title('Rail Structure Comparison', color=COLORS['text'])
        draw_dual_rails(self.ax2, self.rail_data, frame_progress, 1.0)

        # Bottom left: Fibonacci influence
        self.ax3.set_title('Fibonacci Influence Over Time', color=COLORS['text'])
        if frame_progress > 0.3:
            n_points = int(frame_progress * len(self.rail_data['fib_influence']))
            if n_points > 1:
                x_data = np.arange(n_points)
                y_data = self.rail_data['fib_influence'][:n_points]

                self.ax3.plot(0.8 * (x_data / n_points - 0.5),
                             0.8 * (y_data / (y_data.max() + 1e-10) - 0.5),
                             color=COLORS['highlight'], linewidth=2)

        # Bottom right: Constant emergence
        self.ax4.set_title('Enhanced Constant Emergence', color=COLORS['text'])
        phi_val = calculate_phi_emergence_dual(self.rail_data, self.fib_decimal, frame_progress)
        pi_val = calculate_pi_emergence(self.rail_data, frame_progress)
        alpha_val = calculate_alpha_emergence_dual(phi_val, pi_val, self.fib_decimal, frame_progress)

        if frame_progress > 0.5:
            self.ax4.text(0, 0.3, f'œÜ = {phi_val:.6f}', ha='center', va='center',
                         fontsize=12, color=COLORS['phi'])

        if frame_progress > 0.8:
            self.ax4.text(0, 0, f'Œ±‚Åª¬π = {alpha_val:.6f}', ha='center', va='center',
                         fontsize=12, color=COLORS['alpha'])
            self.ax4.text(0, -0.2, f'Target: {ALPHA_INV:.6f}', ha='center', va='center',
                         fontsize=10, color=COLORS['text'])
            error = abs(alpha_val - ALPHA_INV)
            self.ax4.text(0, -0.4, f'Error: {error:.2e}', ha='center', va='center',
                         fontsize=10, color=COLORS['highlight'])

        # Remove axes
        for ax in [self.ax1, self.ax2, self.ax3, self.ax4]:
            ax.axis('off')

        # Main title
        self.fig.suptitle(f'Dual Fibonacci Treatment - Frame {frame_num}/200',
                         fontsize=14, color=COLORS['text'])

        return []

# =================== CORRECTED MAIN ANIMATION CLASS ===================

class BirthOfConstantsAnimation:
    """Main animation controller"""

    def __init__(self, figsize=(12, 10)):
        self.fig, (self.ax_main, self.ax_layers) = plt.subplots(2, 1, figsize=figsize,
                                                               facecolor=COLORS['background'])

        # Generate base data
        self.fib_word = generate_fibonacci_word_corrected(200)
        self.fib_decimal = fibonacci_word_to_decimal_fraction(self.fib_word)

        start_value = PHI - 1  # Start with œÜ - 1
        self.rail_data = dyadic_iteration_dual(start_value, self.fib_decimal, 128)

        # Animation tracking
        self.current_frame = 0
        self.layers = self._setup_layers()

        print(f"Animation initialized with {len(self.fib_word)} bits and {len(self.rail_data['sequence_standard'])} iterations")

    def _setup_layers(self):
        """Setup the Russian doll layers"""
        return {
            'fibonacci': AnimationLayer('Fibonacci Word', COLORS['binary'], (0.0, 1.0)),
            'rails': AnimationLayer('Dyadic Rails', COLORS['rails'], (0.2, 1.0)),
            'phi': AnimationLayer('œÜ Emergence', COLORS['phi'], (0.4, 1.0)),
            'pi': AnimationLayer('œÄ Emergence', COLORS['pi'], (0.5, 1.0)),
            'explosion': AnimationLayer('101 Explosion', COLORS['highlight'], (0.7, 0.9)),
            'alpha': AnimationLayer('Œ± Emergence', COLORS['alpha'], (0.8, 1.0))
        }

    def animate_frame(self, frame_num):
        """Animate a single frame"""
        # Clear axes
        self.ax_main.clear()
        self.ax_layers.clear()

        # Calculate frame progress (0 to 1)
        frame_progress = frame_num / TOTAL_FRAMES
        self.current_frame = frame_num

        # Setup axes
        for ax in [self.ax_main, self.ax_layers]:
            ax.set_xlim(-1, 1)
            ax.set_ylim(-1, 1)
            ax.set_aspect('equal')
            ax.axis('off')
            ax.set_facecolor(COLORS['background'])

        # Calculate emerging constants
        phi_val = calculate_phi_emergence_dual(self.rail_data, self.fib_decimal, frame_progress)
        pi_val = calculate_pi_emergence(self.rail_data, frame_progress)
        alpha_inv_val = calculate_alpha_emergence_dual(phi_val, pi_val, self.fib_decimal, frame_progress)

        # Draw main composite view
        self._draw_main_view(frame_progress, phi_val, pi_val, alpha_inv_val)

        # Draw layer breakdown view
        self._draw_layer_view(frame_progress, phi_val, pi_val, alpha_inv_val)

        # Add frame counter and progress
        self.fig.suptitle(f'Birth of Constants - Frame {frame_num}/{TOTAL_FRAMES}',
                         fontsize=16, color=COLORS['text'], y=0.95)

        # Add progress bar
        progress_width = frame_progress * 0.8
        self.ax_main.add_patch(Rectangle((-0.4, -0.95), progress_width, 0.05,
                                       facecolor=COLORS['highlight'], alpha=0.8))
        self.ax_main.add_patch(Rectangle((-0.4, -0.95), 0.8, 0.05,
                                       fill=False, edgecolor=COLORS['text'], alpha=0.5))

        return []

    def _draw_main_view(self, frame_progress, phi_val, pi_val, alpha_inv_val):
        """Draw the main composite view"""
        self.ax_main.set_title('Complete Evolution', fontsize=14, color=COLORS['text'], pad=20)

        # Draw all layers with appropriate alpha
        draw_fibonacci_word_dual(self.ax_main, self.fib_word, self.fib_decimal, frame_progress,
                               self.layers['fibonacci'].get_alpha(frame_progress))
        draw_dual_rails(self.ax_main, self.rail_data, frame_progress,
                       self.layers['rails'].get_alpha(frame_progress))
        draw_phi_emergence(self.ax_main, phi_val, frame_progress,
                          self.layers['phi'].get_alpha(frame_progress))
        draw_pi_emergence(self.ax_main, pi_val, frame_progress,
                         self.layers['pi'].get_alpha(frame_progress))
        draw_101_explosion(self.ax_main, frame_progress,
                          self.layers['explosion'].get_alpha(frame_progress))
        draw_alpha_emergence(self.ax_main, alpha_inv_val, frame_progress,
                            self.layers['alpha'].get_alpha(frame_progress))

    def _draw_layer_view(self, frame_progress, phi_val, pi_val, alpha_inv_val):
        """Draw the individual layer breakdown"""
        self.ax_layers.set_title('Layer Breakdown (Russian Doll View)',
                                fontsize=14, color=COLORS['text'], pad=20)

        # Determine which layer to focus on
        current_layer = self._get_current_layer(frame_progress)

        if current_layer == 'fibonacci':
            draw_fibonacci_word_dual(self.ax_layers, self.fib_word, self.fib_decimal, frame_progress, 1.0)
        elif current_layer == 'rails':
            draw_dual_rails(self.ax_layers, self.rail_data, frame_progress, 1.0)
        elif current_layer == 'phi':
            draw_phi_emergence(self.ax_layers, phi_val, frame_progress, 1.0)
        elif current_layer == 'pi':
            draw_pi_emergence(self.ax_layers, pi_val, frame_progress, 1.0)
        elif current_layer == 'explosion':
            draw_101_explosion(self.ax_layers, frame_progress, 1.0)
        elif current_layer == 'alpha':
            draw_alpha_emergence(self.ax_layers, alpha_inv_val, frame_progress, 1.0)

        # Add layer name
        layer_name = current_layer.replace('_', ' ').title()
        self.ax_layers.text(0, 0.9, f'Current Layer: {layer_name}',
                           ha='center', va='center', fontsize=12,
                           color=COLORS['highlight'], weight='bold')

    def _get_current_layer(self, frame_progress):
        """Determine which layer should be highlighted"""
        if frame_progress < 0.2:
            return 'fibonacci'
        elif frame_progress < 0.4:
            return 'rails'
        elif frame_progress < 0.5:
            return 'phi'
        elif frame_progress < 0.7:
            return 'pi'
        elif frame_progress < 0.9:
            return 'explosion'
        else:
            return 'alpha'

    def create_animation(self, save_path=None):
        """Create the complete animation"""
        print("Creating animation... This may take a few minutes.")
        anim = animation.FuncAnimation(
            self.fig,
            self.animate_frame,
            frames=TOTAL_FRAMES,
            interval=1000//FRAME_RATE,  # Convert to milliseconds
            blit=False,
            repeat=True
        )

        if save_path:
            print(f"Saving animation to {save_path}...")
            anim.save(save_path, writer='pillow', fps=FRAME_RATE, dpi=DPI)
            print("Animation saved!")

        return anim

# =================== MAIN EXECUTION ===================

def main_dual_test():
    """Main execution for dual Fibonacci word test"""
    print("=" * 60)
    print("üîß DUAL FIBONACCI WORD TREATMENT TEST")
    print("=" * 60)
    print("Testing both binary sequence and decimal fraction interpretations")
    print()

    # Run corrected tests
    if not run_corrected_tests():
        print("‚ùå Tests failed. Check the implementation.")
        return None

    print()
    print("üé¨ Creating sample visualization...")

    # Create sample visualization
    animation_system = DualFibonacciAnimation(figsize=(12, 8))
    animation_system.animate_frame(150)  # Show frame at 75% progress

    print("‚ú® Dual Fibonacci word test complete!")
    print(f"Key findings:")
    print(f"  - Fibonacci word as binary: {animation_system.fib_word[:30]}...")
    print(f"  - Fibonacci word as decimal: {animation_system.fib_decimal:.12f}")
    print(f"  - Both interpretations successfully integrated into dyadic iteration")

    return animation_system

# Auto-run if in Colab
if __name__ == "__main__":
    try:
        import google.colab
        print("üîó Running in Google Colab - Auto-executing main_dual_test()")
        result = main_dual_test()
    except ImportError:
        print("üíª Running locally - Call main_dual_test() to start")

print("üöÄ Dual Fibonacci Word Treatment System Ready!")
print("This version fixes the Colab errors and shows both interpretations simultaneously.")
```
