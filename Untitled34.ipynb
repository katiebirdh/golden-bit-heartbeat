```python
# %% [markdown]
# # Universal δ–φ–γ Suite (bias-controlled, Colab-ready)
# Tests the candidate balance:
#     δ/(δ−1) + √φ   ≈   exp(γ · s)
# with s ∈ {1, φ} and all (a,b) ∈ {0,1}² evaluated (pre-registered grid).
# Includes:
#   • Feigenbaum δ_z estimation via robust superstable solver (no lookups; fallback for z=2 if needed)
#   • Monte-Carlo null controlling for selection (min over grid)
#   • Universality residuals over z
#   • Golden “extremality” check vs metallic means
# Saves CSVs/PNGs with prefix OUT_PREFIX.

# ========== Imports & precision ==========
import math, random, time, json, csv
import mpmath as mp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

mp.mp.dps = 120  # high precision
RNG = random.Random(1337)

# ========== Canonical constants (computed) ==========
phi = (1 + mp.sqrt(5)) / 2
sqrt_phi = mp.sqrt(phi)
gamma = mp.euler   # Euler–Mascheroni

# ========== Unimodal map family ==========
def iterate(x0, mu, z, k):
    """Iterate f_z(x) = 1 - mu * |x|^z, k steps."""
    x = mp.mpf(x0)
    z = mp.mpf(z)
    mu = mp.mpf(mu)
    for _ in range(k):
        x = 1 - mu * (abs(x) ** z)
    return x

def F_superstable(mu, z, n):
    """Superstable condition value: f_z^{(2^n)}(0)."""
    return iterate(0, mu, z, 1 << n)

# ========== Robust superstable solver (touching-zero aware) ==========
def find_superstable_mu(z: float, n: int,
                        mu_min: float = 0.0, mu_max: float = 4.0,
                        coarse: int = 20000, tol: float = 1e-40) -> mp.mpf:
    """
    Solve f_z^{(2^n)}(0) = 0 robustly:
      1) coarse scan; collect local minima of |F| (roots can 'kiss' zero)
      2) refine each candidate with mp.findroot (two-seed & one-seed)
      3) fallback: sign-change brackets → hybrid secant/bisection
    Accepts root iff |F(mu*)| < tol.
    """
    z = mp.mpf(z)
    mu_lo, mu_hi = mp.mpf(mu_min), mp.mpf(mu_max)
    K = 1 << n
    def F(mu): return iterate(0, mu, z, K)

    # (1) coarse scan & local minima of |F|
    grid = [mu_lo + (mu_hi - mu_lo) * mp.mpf(i) / (coarse - 1) for i in range(coarse)]
    vals = [F(mu) for mu in grid]
    absvals = [abs(v) for v in vals]

    minima_idx = []
    for i in range(1, coarse - 1):
        if absvals[i] <= absvals[i-1] and absvals[i] <= absvals[i+1]:
            minima_idx.append(i)
    # try up to 12 best minima
    seeds = sorted(minima_idx, key=lambda i: absvals[i])[:12]

    # (2) mp.findroot refinements
    for i in seeds:
        a = grid[max(0, i-1)]
        b = grid[min(coarse-1, i+1)]
        try:
            mu_star = mp.findroot(F, a, b, tol=mp.mpf('1e-40'), maxsteps=100)
            if abs(F(mu_star)) < mp.mpf(tol):
                return mu_star
        except:
            c = grid[i]
            try:
                mu_star = mp.findroot(F, c, tol=mp.mpf('1e-40'), maxsteps=100)
                if abs(F(mu_star)) < mp.mpf(tol):
                    return mu_star
            except:
                pass

    # (3) sign-change fallback
    brackets = []
    for i in range(coarse - 1):
        if vals[i] == 0:
            return grid[i]
        if vals[i] * vals[i+1] < 0:
            brackets.append((grid[i], grid[i+1]))

    def _hybrid_zero(a, b):
        fa, fb = F(a), F(b)
        # a few secant steps with bracketing guards
        for _ in range(12):
            if abs(fb - fa) < mp.mpf('1e-60'):
                break
            s = b - fb * (b - a) / (fb - fa)
            if s <= min(a, b) or s >= max(a, b):
                s = (a + b) / 2
            fs = F(s)
            if fa * fs <= 0:
                b, fb = s, fs
            else:
                a, fa = s, fs
            if abs(b - a) < mp.mpf('1e-30') or abs(fs) < mp.mpf(tol):
                if abs(F(s)) < mp.mpf(tol):
                    return s
                break
        # polish by bisection
        aa, bb = a, b
        for _ in range(200):
            m = (aa + bb) / 2
            fm = F(m)
            if fm == 0 or abs(bb - aa) < mp.mpf('1e-30'):
                return m
            if F(aa) * fm < 0:
                bb = m
            else:
                aa = m
        return (aa + bb) / 2

    for a, b in brackets:
        root = _hybrid_zero(a, b)
        if abs(F(root)) < mp.mpf(tol):
            return root

    raise RuntimeError(f"superstable root not found (z={z}, n={n})")

# ========== Feigenbaum δ_z estimator ==========
def estimate_delta_z(z: float,
                     n_start: int = 2, n_end: int = 9,
                     coarse: int = 20000) -> tuple[mp.mpf, list[mp.mpf], list[mp.mpf]]:
    """
    δ_z ≈ (μ'_{n-1}-μ'_{n-2}) / (μ'_n-μ'_{n-1}), averaged over last few stable levels.
    Skips numerically unstable steps; uses robust superstable solver above.
    """
    if z == 1.5:
        # You can remove this guard if you're okay with longer runtimes.
        raise RuntimeError("z=1.5 skipped (slow & fragile)")

    mus = []
    for n in range(n_start, n_end + 1):
        mu_n = find_superstable_mu(z, n, mu_min=0.0, mu_max=4.0, coarse=coarse)
        mus.append(mu_n)

    deltas = []
    for i in range(2, len(mus)):
        num = mus[i-1] - mus[i-2]
        den = mus[i]   - mus[i-1]
        if abs(den) < mp.mpf('1e-28'):
            continue
        deltas.append(num / den)

    if not deltas:
        raise RuntimeError(f"δ_z could not be estimated for z={z}")

    K = min(3, len(deltas))
    return sum(deltas[-K:]) / K, mus, deltas

# ========== Pre-registered expression grid ==========
def lhs_value(delta, a, b): return (mp.mpf(a)*(delta/(delta-1))) + (mp.mpf(b)*sqrt_phi)
def rhs_value(s): return mp.e ** (gamma * s)
EXPR_GRID = [(a,b,s) for a in (0,1) for b in (0,1) for s in (mp.mpf(1), phi)]

def evaluate_expression_grid(delta):
    rows=[]
    for (a,b,s) in EXPR_GRID:
        L=lhs_value(delta,a,b); R=rhs_value(s)
        abs_err=abs(L-R); rel_err=abs_err/abs(R)
        rows.append({"a":int(a),"b":int(b),"s_label":"1" if s==1 else "phi",
                     "LHS":str(L),"RHS":str(R),
                     "abs_err":float(abs_err),"rel_err":float(rel_err)})
    return pd.DataFrame(rows).sort_values("rel_err").reset_index(drop=True)

# ========== Monte-Carlo null (selection-controlled) ==========
def monte_carlo_null(num_trials=2000):
    mins=[]
    for _ in range(num_trials):
        delta_r=mp.mpf(4.5 + RNG.random()*0.5)      # around Feigenbaum scale
        phi_r  =mp.mpf(1.55 + RNG.random()*0.25)    # around φ
        gamma_r=mp.mpf(0.56 + RNG.random()*0.03)    # around γ
        sqrt_phi_r=mp.sqrt(phi_r)
        best=mp.inf
        for (a,b,s_choice) in EXPR_GRID:
            L=(mp.mpf(a)*(delta_r/(delta_r-1))) + (mp.mpf(b)*sqrt_phi_r)
            s_val = 1 if s_choice==1 else phi_r
            R=mp.e**(gamma_r*s_val)
            err=abs(L-R)/abs(R)
            if err<best: best=err
        mins.append(float(best))
    return mins

# ========== Metallic means & golden extremality ==========
def metallic_mean(m): return (m + mp.sqrt(m**2 + 4))/2

def golden_extremality_table(delta, M_max=10):
    rows=[]
    L_phi=(delta/(delta-1))+sqrt_phi
    s_hat=mp.log(L_phi)/gamma
    rows.append({"mode":"baseline","m":1,
                 "best_s_hat":float(s_hat),
                 "abs_best_s_minus_phi":float(abs(s_hat-phi))})
    for m in range(1,M_max+1):
        M_m=metallic_mean(m)
        L_m=(delta/(delta-1))+mp.sqrt(M_m)
        s_hat_m=mp.log(L_m)/gamma
        rows.append({"mode":"swap_sqrt_phi","m":m,
                     "best_s_hat":float(s_hat_m),
                     "abs_best_s_minus_phi":float(abs(s_hat_m-phi))})
    return pd.DataFrame(rows)

# ========== Main run (everything tied together) ==========
OUT_PREFIX="delta_phi_gamma_suite"
Z_LIST=[1.5, 2.0, 2.5, 3.0]      # feel free to extend later
N_START, N_END = 2, 9
COARSE_SCAN   = 20000
NULL_TRIALS   = 2000

print("== Mechanistic scaffold ==")
print("""
We test a scaling balance between (i) dyadic renormalization gain and (ii) a golden
mid-inflation step against (iii) a discrete–continuous correction budget:
    LHS = δ/(δ−1) + √φ   versus   RHS = exp(γ · s)
The pre-registered class fixes LHS to {0, δ/(δ−1)} + {0, √φ} and s ∈ {1, φ}. All
combinations are evaluated; selection bias is handled via a Monte-Carlo null that
replaces (δ, φ, γ) with random irrationals of similar magnitude.
""")

# ---- δ sweep (with fallback for z=2 if solver refuses) ----
delta_estimates=[]
for i, z in enumerate(Z_LIST):
    t0=time.time()
    try:
        dz, mus, dlist = estimate_delta_z(z, n_start=N_START, n_end=N_END, coarse=COARSE_SCAN)
        dt=time.time()-t0
        delta_estimates.append({"z":z,"delta_est":float(dz),
                                "mus":[str(m) for m in mus],
                                "delta_by_n":[str(dv) for dv in dlist],
                                "seconds":dt})
        print(f"[{i}] z={z}: δ≈{dz}  ({dt:.1f}s)")
    except Exception as e:
        if abs(z-2.0) < 1e-12:
            # Canonical high-precision δ fallback so downstream analyses run
            dz = mp.mpf("4.669201609102990671853203820466")
            delta_estimates.append({"z":z,"delta_est":float(dz),
                                    "mus":[], "delta_by_n":[], "seconds":None})
            print(f"[{i}] z={z}: WARNING using canonical δ (solver failed: {e})")
        else:
            print(f"[{i}] z={z}: skipped ({e})")

df_delta = pd.DataFrame(delta_estimates)
display(df_delta)
df_delta.to_csv(f"{OUT_PREFIX}_delta_sweep.csv", index=False)

# ---- choose baseline δ (z=2) for expression grid ----
if not df_delta.empty and (df_delta['z']==2.0).any():
    delta_quadratic = mp.mpf(df_delta[df_delta['z']==2.0]['delta_est'].iloc[0])
else:
    delta_quadratic = mp.mpf("4.669201609102990671853203820466")  # canonical

# ---- pre-registered expression grid ----
df_grid = evaluate_expression_grid(delta_quadratic)
display(df_grid)
df_grid.to_csv(f"{OUT_PREFIX}_expression_grid.csv", index=False)
best_rel_err = float(df_grid['rel_err'].min())
print("Best combo:")
print(df_grid.iloc[0].to_string(index=False))

# ---- Monte-Carlo null (selection-controlled) ----
print(f"\nRunning Monte-Carlo null with {NULL_TRIALS} trials …")
null_mins = monte_carlo_null(NULL_TRIALS)
p_hat = sum(1 for x in null_mins if x <= best_rel_err) / len(null_mins)
print(f"Observed best rel_err = {best_rel_err:.8e}")
print(f"Monte-Carlo p̂ (min-over-grid) = {p_hat:.6f}")
pd.DataFrame({"min_rel_err": null_mins}).to_csv(f"{OUT_PREFIX}_null_montecarlo.csv", index=False)

plt.figure(figsize=(6,4))
plt.hist(null_mins, bins=40)
plt.axvline(best_rel_err, linestyle='--', linewidth=2)
plt.xlabel("Min relative error over expression grid (null trials)")
plt.ylabel("Count")
plt.title("Monte-Carlo Null — Selection Controlled")
plt.tight_layout()
plt.savefig(f"{OUT_PREFIX}_null_hist.png", dpi=160)
plt.show()

# ---- Universality residuals (apply δ_z to candidate law) ----
target_R = mp.e ** (gamma * phi)
rows_uni = []
for rec in delta_estimates:
    if "delta_est" not in rec:
        continue
    dz = mp.mpf(rec["delta_est"])
    L = (dz / (dz - 1)) + sqrt_phi
    abs_err = abs(L - target_R)
    rel_err = abs_err / abs(target_R)
    rows_uni.append({"z":rec["z"], "delta_est":rec["delta_est"],
                     "abs_err":float(abs_err), "rel_err":float(rel_err)})
df_uni = pd.DataFrame(rows_uni).sort_values("rel_err")
display(df_uni)
df_uni.to_csv(f"{OUT_PREFIX}_universality_residuals.csv", index=False)

# ---- Golden extremality (constant-space) ----
df_gold = golden_extremality_table(delta_quadratic, M_max=10)
display(df_gold.sort_values(["mode","abs_best_s_minus_phi"]))
df_gold.to_csv(f"{OUT_PREFIX}_golden_extremality.csv", index=False)

plt.figure(figsize=(6,4))
swap = df_gold[df_gold.mode=="swap_sqrt_phi"].copy()
plt.plot(swap["m"], swap["abs_best_s_minus_phi"], marker='o')
plt.xlabel("m (metallic mean index)")
plt.ylabel("| ŝ  −  φ |")
plt.title("Golden Extremality (constant-space)")
plt.tight_layout()
plt.savefig(f"{OUT_PREFIX}_golden_extremality_proximity.png", dpi=160)
plt.show()

print("\n== Done. Outputs saved with prefix:", OUT_PREFIX, "==")
```
