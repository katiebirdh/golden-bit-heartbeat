```python
import numpy as np
from numpy.random import default_rng
rng = default_rng(137)

def emd_hist_01(x, bins=160):
    h,e = np.histogram(np.asarray(x)%1.0, bins=bins, range=(0,1), density=True)
    c = (e[:-1]+e[1:])/2
    C = np.cumsum(h); C /= C[-1] if C[-1] else 1
    return float(np.trapezoid(np.abs(C - c), c))

def kuiper_p(x, m=1000):
    x = np.sort(np.asarray(x)%1.0); n=len(x);
    grid = np.arange(1,n+1)/n
    Dp = np.max(grid - x); Dm = np.max(x - (np.arange(0,n)/n)); V = Dp+Dm
    u = np.sort(rng.random((m,n)), axis=1)
    Dp0 = np.max(grid - u, axis=1);
    u2 = np.sort(rng.random((m,n)), axis=1)
    Dm0 = np.max(u2 - (np.arange(0,n)/n), axis=1)
    V0 = Dp0+Dm0
    return (np.sum(V0>=V)+1)/(m+1)
```
