```python
#!/usr/bin/env python3
"""
Comprehensive Binary Substrate Reality Tests

Testing the hypothesis that reality emerges from binary computational substrate
through systematic mathematical analysis of:
1. Binary substrate ‚Üí constant emergence
2. Phase transitions at critical thresholds
3. Scale-dependent determinism vs apparent randomness
4. Geometric relationships across mathematical domains
5. Statistical patterns in prime sequences
6. Mirror/inverse relationships in fundamental constants

Self-contained, triple-tested, user-free execution
Designed for ~15 minute runtime in Colab environment
"""

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import Circle, Rectangle, Polygon
import warnings
from scipy import stats
from scipy.fft import fft, fftfreq, rfft, rfftfreq
from scipy.optimize import minimize_scalar
from scipy.special import factorial, gamma
import time
from datetime import datetime
import itertools
from collections import defaultdict, Counter
import random

warnings.filterwarnings('ignore')
plt.style.use('dark_background')

print("üåå COMPREHENSIVE BINARY SUBSTRATE REALITY TESTS")
print("=" * 70)
print(f"Initiated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("Testing the computational substrate hypothesis across multiple domains")
print("Estimated runtime: ~15 minutes")
print()

# =================== FUNDAMENTAL CONSTANTS & PRECISION ===================

# Mathematical constants (extended precision)
PHI = (1 + np.sqrt(5)) / 2
PI = np.pi
E = np.e
ALPHA_INV = 137.035999084
SQRT_2 = np.sqrt(2)
SQRT_3 = np.sqrt(3)
LN_2 = np.log(2)
LN_PHI = np.log(PHI)

# Critical thresholds from theory
E_THRESHOLD = E
PHI_SQUARED = PHI**2
QUARTER_DYADIC = 2**(-1/4)

# Physical constants for validation
PROTON_ELECTRON_MASS_RATIO = 1836.15267343
FINE_STRUCTURE = 1/ALPHA_INV

# Precision parameters
DEFAULT_PRECISION = 1e-12
HIGH_PRECISION = 1e-15
ULTRA_PRECISION = 1e-18

# Visual configuration
COLORS = {
    'substrate': '#1a1a2e',
    'binary': '#16213e',
    'emergence': '#e94560',
    'phi': '#f39c12',
    'pi': '#8e44ad',
    'alpha': '#e74c3c',
    'mirror': '#00ffff',
    'phase_pre': '#9b59b6',
    'phase_post': '#f39c12',
    'background': '#0f0f23',
    'text': '#ffffff',
    'success': '#2ecc71',
    'warning': '#f1c40f',
    'error': '#e74c3c'
}

print(f"Constants initialized to {DEFAULT_PRECISION:.0e} precision:")
print(f"  œÜ = {PHI:.12f}")
print(f"  œÄ = {PI:.12f}")
print(f"  e = {E:.12f} (critical threshold)")
print(f"  Œ±‚Åª¬π = {ALPHA_INV:.12f}")
print(f"  ‚àö2 = {SQRT_2:.12f}")
print(f"  ‚àö3 = {SQRT_3:.12f}")
print()

# =================== BINARY SUBSTRATE GENERATORS ===================

def generate_fibonacci_word_ultra(n_bits=1000):
    """
    Ultra-robust Fibonacci word generation with validation
    Rules: 0 ‚Üí 01, 1 ‚Üí 0, starting from '0'
    """
    if n_bits <= 0:
        return ""

    word = "0"
    generation = 0
    max_generations = 50  # Safety limit

    while len(word) < n_bits and generation < max_generations:
        new_word = ""
        for char in word:
            if char == '0':
                new_word += "01"
            elif char == '1':
                new_word += "0"

        if len(new_word) <= len(word):
            break

        word = new_word
        generation += 1

    # Validate against known Fibonacci word properties
    result = word[:n_bits]

    # Check golden ratio property in length ratios
    if len(result) > 20:
        # Count 0s and 1s - should approach golden ratio
        count_0 = result.count('0')
        count_1 = result.count('1')
        if count_1 > 0:
            ratio = count_0 / count_1
            expected_ratio = PHI
            if abs(ratio - expected_ratio) > 0.1:
                print(f"‚ö†Ô∏è  Fibonacci word ratio validation: {ratio:.6f} vs expected {expected_ratio:.6f}")

    return result

def generate_de_bruijn_sequence(k=2, n=3):
    """
    Generate de Bruijn sequence B(k,n) - each n-length string appears exactly once
    Using Martin's algorithm for efficiency
    """
    alphabet = list(range(k))
    a = [0] * k * n
    sequence = []

    def db(t, p):
        if t > n:
            if n % p == 0:
                sequence.extend(a[1:p + 1])
        else:
            a[t] = a[t - p]
            db(t + 1, p)
            for j in range(a[t - p] + 1, k):
                a[t] = j
                db(t + 1, t)

    db(1, 1)
    return ''.join(map(str, sequence))

def binary_to_decimal_substrate(binary_str):
    """Convert binary string to decimal fraction with substrate properties"""
    if not binary_str:
        return 0.0

    decimal_value = 0.0
    for i, bit in enumerate(binary_str):
        if bit == '1':
            decimal_value += 2 ** (-(i + 1))

    return decimal_value

def substrate_modulation(x, fib_decimal, modulation_strength=0.1):
    """Apply Fibonacci substrate modulation to iteration"""
    # Phase modulation using Fibonacci decimal as frequency
    phase = 2 * PI * fib_decimal * x
    modulation = modulation_strength * np.sin(phase)
    return x * (1 + modulation)

# =================== ADVANCED DYADIC ITERATION ENGINE ===================

def advanced_dyadic_iteration(start_value, fib_decimal, n_iterations=512,
                            modulation=True, track_phases=True):
    """
    Advanced dyadic iteration with substrate modulation and phase tracking
    """
    results = {
        'sequence': [],
        'errors': [],
        'phases': [],
        'rail_positions': [],
        'substrate_influence': [],
        'crystallization_measure': [],
        'mirror_candidates': [],
        'geometric_ratios': [],
        'iterations': np.arange(n_iterations)
    }

    x = start_value

    for i in range(n_iterations):
        # Apply substrate modulation if enabled
        if modulation:
            x_mod = substrate_modulation(x, fib_decimal, 0.05)
        else:
            x_mod = x

        # Dyadic doubling with fractional part
        x_doubled = 2 * x_mod
        x_frac = x_doubled - np.floor(x_doubled)

        # Error measurement (distance to binary boundaries)
        error = min(x_frac, 1 - x_frac)

        # Phase tracking (mod 4 for period-4 rails)
        phase = i % 4 if track_phases else 0

        # Rail position (logarithmic error scale)
        rail_pos = -np.log10(error + ULTRA_PRECISION)

        # Substrate influence measure
        substrate_inf = abs(x_frac - (x if not modulation else x)) if modulation else 0

        # Crystallization measure (stability of iteration)
        if i > 10:
            recent_errors = results['errors'][-10:]
            crystallization = 1.0 / (1.0 + np.std(recent_errors))
        else:
            crystallization = 0.0

        # Mirror candidate extraction
        x_str = f"{x_frac:.15f}"
        decimal_part = x_str.split('.')[1] if '.' in x_str else ""
        if len(decimal_part) >= 3:
            first_three = decimal_part[:3]
            mirror_cand = int(first_three[::-1]) if first_three.isdigit() else 0
        else:
            mirror_cand = 0

        # Geometric ratio extraction
        geometric_ratio = x_frac / (1 - x_frac) if x_frac != 1.0 else float('inf')

        # Store results
        results['sequence'].append(x_frac)
        results['errors'].append(error)
        results['phases'].append(phase)
        results['rail_positions'].append(rail_pos)
        results['substrate_influence'].append(substrate_inf)
        results['crystallization_measure'].append(crystallization)
        results['mirror_candidates'].append(mirror_cand)
        results['geometric_ratios'].append(geometric_ratio)

        # Update for next iteration
        x = x_frac

    # Convert lists to numpy arrays
    for key in results:
        if key != 'iterations':
            results[key] = np.array(results[key])

    return results

# =================== CONSTANT EMERGENCE ANALYSIS ===================

def analyze_constant_emergence(dyadic_results, target_constants=None):
    """
    Analyze emergence of mathematical constants from dyadic iteration
    """
    if target_constants is None:
        target_constants = {
            'phi': PHI,
            'pi': PI,
            'e': E,
            'alpha_inv': ALPHA_INV,
            'sqrt2': SQRT_2,
            'sqrt3': SQRT_3
        }

    emergence_analysis = {
        'constant_matches': {},
        'golden_ladder_positions': [],
        'phase_4_alignment': {},
        'mirror_relationships': {},
        'statistical_significance': {}
    }

    sequence = dyadic_results['sequence']
    phases = dyadic_results['phases']
    rail_positions = dyadic_results['rail_positions']
    mirror_candidates = dyadic_results['mirror_candidates']

    # Analyze constant emergence through rail positions
    for const_name, const_value in target_constants.items():
        # Look for sequences that converge to constant values
        matches = []
        for i, seq_val in enumerate(sequence):
            # Multiple approaches to find constant emergence

            # Direct value matching
            direct_error = abs(seq_val - const_value)
            if direct_error < DEFAULT_PRECISION:
                matches.append(('direct', i, direct_error))

            # Golden ladder scaling (œÜ^n relationships)
            for n in range(-3, 4):
                scaled_val = seq_val * (PHI ** n)
                scaled_error = abs(scaled_val - const_value)
                if scaled_error < DEFAULT_PRECISION:
                    matches.append(('golden_scaled', i, scaled_error, n))

            # Inverse relationships
            if seq_val != 0:
                inv_error = abs(1/seq_val - const_value)
                if inv_error < DEFAULT_PRECISION:
                    matches.append(('inverse', i, inv_error))

        emergence_analysis['constant_matches'][const_name] = matches

    # Phase-4 alignment analysis
    for phase in range(4):
        phase_mask = phases == phase
        if np.any(phase_mask):
            phase_positions = rail_positions[phase_mask]
            phase_median = np.median(phase_positions)
            emergence_analysis['phase_4_alignment'][phase] = {
                'median_position': phase_median,
                'std_deviation': np.std(phase_positions),
                'count': np.sum(phase_mask)
            }

    # Mirror relationship analysis
    mirror_counts = Counter(mirror_candidates)
    for mirror_val, count in mirror_counts.most_common(10):
        if mirror_val > 0:
            # Check against known constants
            for const_name, const_value in target_constants.items():
                const_int = int(const_value) if const_value < 1000 else int(const_value * 1000)
                if abs(mirror_val - const_int) <= 1:
                    emergence_analysis['mirror_relationships'][const_name] = {
                        'mirror_value': mirror_val,
                        'target_value': const_int,
                        'count': count,
                        'accuracy': abs(mirror_val - const_int)
                    }

    return emergence_analysis

# =================== PRIME RESIDUE ANALYSIS ===================

def analyze_prime_residues(modulus=137, n_primes=50000):
    """
    Analyze prime residue patterns for non-random structure
    """
    print(f"Generating first {n_primes} primes for residue analysis...")

    # Generate primes using sieve of Eratosthenes
    def sieve_of_eratosthenes(limit):
        sieve = [True] * (limit + 1)
        sieve[0] = sieve[1] = False

        for i in range(2, int(limit**0.5) + 1):
            if sieve[i]:
                for j in range(i*i, limit + 1, i):
                    sieve[j] = False

        return [i for i in range(2, limit + 1) if sieve[i]]

    # Estimate upper bound for nth prime (approximation)
    if n_primes < 6:
        upper_bound = 15
    else:
        upper_bound = int(n_primes * (np.log(n_primes) + np.log(np.log(n_primes))))

    primes = sieve_of_eratosthenes(upper_bound)

    # Extend if necessary
    while len(primes) < n_primes:
        upper_bound *= 2
        primes = sieve_of_eratosthenes(upper_bound)

    primes = primes[:n_primes]

    # Calculate residues
    residues = [p % modulus for p in primes]

    # Statistical analysis
    residue_counts = Counter(residues)

    # Expected uniform distribution
    expected_count = n_primes / modulus

    # Chi-square test for uniformity
    observed = np.array([residue_counts.get(i, 0) for i in range(modulus)])
    expected = np.full(modulus, expected_count)

    # Remove zero class as it's deterministically absent (except for prime 2 when modulus is odd)
    if modulus % 2 == 1:  # Odd modulus
        non_zero_observed = observed[1:]
        non_zero_expected = expected[1:]
        dof = modulus - 2
    else:
        non_zero_observed = observed[observed > 0]
        non_zero_expected = expected[expected > 0]
        dof = len(non_zero_observed) - 1

    chi2_stat = np.sum((non_zero_observed - non_zero_expected)**2 / non_zero_expected)
    p_value = 1 - stats.chi2.cdf(chi2_stat, dof)

    # FFT analysis for period-4 pattern
    residue_sequence = np.array(residues)
    fft_result = np.abs(rfft(residue_sequence))
    freqs = rfftfreq(len(residue_sequence))

    # Look for quarter-cycle peak
    quarter_freq_idx = np.argmin(np.abs(freqs - 0.25))
    quarter_power = fft_result[quarter_freq_idx]
    total_power = np.sum(fft_result)
    quarter_fraction = quarter_power / total_power

    analysis_results = {
        'n_primes': len(primes),
        'modulus': modulus,
        'residue_counts': dict(residue_counts),
        'chi2_statistic': chi2_stat,
        'chi2_p_value': p_value,
        'degrees_of_freedom': dof,
        'uniformity_deviation': chi2_stat / dof,  # Normalized deviation
        'quarter_cycle_power': quarter_power,
        'quarter_cycle_fraction': quarter_fraction,
        'fft_frequencies': freqs[:50],  # First 50 frequencies
        'fft_powers': fft_result[:50],
        'residues': residues[:1000]  # First 1000 for visualization
    }

    return analysis_results

# =================== GEOMETRIC RELATIONSHIP TESTS ===================

def test_reuleaux_relationships():
    """
    Test geometric relationships involving Reuleaux triangles and fundamental constants
    """
    print("Testing Reuleaux triangle - fundamental constant relationships...")

    # Reuleaux triangle area formula: A_R = (œÄ - ‚àö3)/2
    reuleaux_area = (PI - SQRT_3) / 2

    # Test relationship to fine structure constant
    # From paper: Œ± ‚âà 101(œÜ^(œÄ-2) - ‚àö3)
    phi_power = PHI**(PI - 2)
    alpha_formula_result = 101 * (phi_power - SQRT_3)
    alpha_error = abs(alpha_formula_result - ALPHA_INV)

    # Test if the (œÄ - ‚àö3) term connects Reuleaux area to Œ±
    pi_minus_sqrt3 = PI - SQRT_3
    reuleaux_connection = 2 * reuleaux_area  # Should equal œÄ - ‚àö3
    reuleaux_error = abs(reuleaux_connection - pi_minus_sqrt3)

    # Test hexagonal lattice connection
    # Hex unit cell area: ‚àö3/2
    hex_area = SQRT_3 / 2
    half_circle_area = PI / 2
    hex_circle_gap = half_circle_area - hex_area

    # Golden ratio geometric relationships
    golden_rectangle_area = PHI * 1  # œÜ √ó 1 rectangle
    golden_spiral_pitch = 2 * PI / LN_PHI  # Spiral pitch constant

    # Test binary relationship: ‚àö2 as diagonal scaler
    unit_square_diagonal = SQRT_2
    area_doubling_factor = SQRT_2  # Linear scale factor for 2x area

    geometric_tests = {
        'reuleaux_area': reuleaux_area,
        'pi_minus_sqrt3': pi_minus_sqrt3,
        'reuleaux_connection_error': reuleaux_error,
        'alpha_formula_result': alpha_formula_result,
        'alpha_formula_error': alpha_error,
        'alpha_formula_accuracy': 1 - (alpha_error / ALPHA_INV),
        'hex_area': hex_area,
        'hex_circle_gap': hex_circle_gap,
        'golden_rectangle_area': golden_rectangle_area,
        'golden_spiral_pitch': golden_spiral_pitch,
        'sqrt2_relationships': {
            'unit_diagonal': unit_square_diagonal,
            'area_doubler': area_doubling_factor,
            'geometric_mean_sqrt2_phi': np.sqrt(SQRT_2 * PHI)
        }
    }

    return geometric_tests

# =================== PHASE TRANSITION ANALYSIS ===================

def analyze_phase_transitions(scaling_range=(0.5, 5.0), n_points=20):
    """
    Analyze phase transitions around critical thresholds (e, œÜ¬≤, etc.)
    """
    print(f"Analyzing phase transitions across {n_points} scaling factors...")

    scaling_factors = np.linspace(scaling_range[0], scaling_range[1], n_points)

    transition_data = {
        'scaling_factors': scaling_factors,
        'phi_stability': [],
        'alpha_stability': [],
        'crystallization_metrics': [],
        'substrate_coherence': [],
        'phase_classifications': []
    }

    # Generate base substrate
    fib_word = generate_fibonacci_word_ultra(300)
    fib_decimal = binary_to_decimal_substrate(fib_word)

    for sf in scaling_factors:
        # Run scaled iteration
        start_val = (PHI - 1) * sf
        results = advanced_dyadic_iteration(start_val, fib_decimal,
                                          n_iterations=128, modulation=True)

        # Measure stability (how much values fluctuate)
        sequence = results['sequence']

        # œÜ stability: how close sequence converges to œÜ-related values
        phi_targets = [PHI, PHI-1, 1/PHI, PHI**2, PHI**(-1)]
        phi_min_distances = []
        for seq_val in sequence[-20:]:  # Last 20 iterations
            min_dist = min(abs(seq_val - target) for target in phi_targets)
            phi_min_distances.append(min_dist)
        phi_stability = 1.0 / (1.0 + np.mean(phi_min_distances))

        # Œ± stability: how close to Œ±-related values
        alpha_targets = [1/ALPHA_INV, ALPHA_INV/1000, ALPHA_INV/100]
        alpha_min_distances = []
        for seq_val in sequence[-20:]:
            min_dist = min(abs(seq_val - target) for target in alpha_targets)
            alpha_min_distances.append(min_dist)
        alpha_stability = 1.0 / (1.0 + np.mean(alpha_min_distances))

        # Crystallization: inverse of variation
        crystallization = np.mean(results['crystallization_measure'][-20:])

        # Substrate coherence: how much substrate modulation affects iteration
        substrate_coherence = np.mean(results['substrate_influence'])

        # Phase classification
        if sf < E_THRESHOLD:
            phase = 'pre_e'
        elif sf < PHI_SQUARED:
            phase = 'transition'
        else:
            phase = 'post_e'

        # Store results
        transition_data['phi_stability'].append(phi_stability)
        transition_data['alpha_stability'].append(alpha_stability)
        transition_data['crystallization_metrics'].append(crystallization)
        transition_data['substrate_coherence'].append(substrate_coherence)
        transition_data['phase_classifications'].append(phase)

    # Convert to numpy arrays
    for key in ['phi_stability', 'alpha_stability', 'crystallization_metrics', 'substrate_coherence']:
        transition_data[key] = np.array(transition_data[key])

    return transition_data

# =================== SCALE-DEPENDENT RANDOMNESS TESTS ===================

def test_scale_dependent_randomness(binary_sequence, scales=[1, 2, 4, 8, 16, 32]):
    """
    Test whether apparent randomness is scale-dependent observation artifact
    """
    print("Testing scale-dependent randomness hypothesis...")

    randomness_analysis = {
        'scales': scales,
        'entropy_by_scale': [],
        'autocorrelation_by_scale': [],
        'predictability_by_scale': [],
        'pattern_strength_by_scale': []
    }

    # Convert binary string to numeric array
    if isinstance(binary_sequence, str):
        bit_array = np.array([int(b) for b in binary_sequence])
    else:
        bit_array = np.array(binary_sequence)

    for scale in scales:
        # Downsample sequence at this scale
        if scale == 1:
            scaled_sequence = bit_array
        else:
            # Group into blocks of 'scale' size and take majority vote
            n_blocks = len(bit_array) // scale
            scaled_sequence = []
            for i in range(n_blocks):
                block = bit_array[i*scale:(i+1)*scale]
                scaled_sequence.append(1 if np.mean(block) > 0.5 else 0)
            scaled_sequence = np.array(scaled_sequence)

        if len(scaled_sequence) < 10:
            continue

        # Calculate entropy (measure of randomness)
        p0 = np.mean(scaled_sequence == 0)
        p1 = np.mean(scaled_sequence == 1)
        if p0 > 0 and p1 > 0:
            entropy = -p0 * np.log2(p0) - p1 * np.log2(p1)
        else:
            entropy = 0

        # Calculate autocorrelation (measure of self-similarity)
        if len(scaled_sequence) > 20:
            autocorr = np.corrcoef(scaled_sequence[:-1], scaled_sequence[1:])[0,1]
            if np.isnan(autocorr):
                autocorr = 0
        else:
            autocorr = 0

        # Calculate predictability (inverse of complexity)
        # Use run-length encoding efficiency as proxy
        runs = []
        current_run = 1
        for i in range(1, len(scaled_sequence)):
            if scaled_sequence[i] == scaled_sequence[i-1]:
                current_run += 1
            else:
                runs.append(current_run)
                current_run = 1
        runs.append(current_run)

        # Compression ratio: fewer, longer runs = more predictable
        compression_ratio = len(runs) / len(scaled_sequence)
        predictability = 1 - compression_ratio

        # Pattern strength: FFT peak analysis
        if len(scaled_sequence) > 8:
            fft_result = np.abs(fft(scaled_sequence))
            # Exclude DC component
            pattern_strength = np.max(fft_result[1:len(fft_result)//2]) / np.sum(fft_result[1:])
        else:
            pattern_strength = 0

        randomness_analysis['entropy_by_scale'].append(entropy)
        randomness_analysis['autocorrelation_by_scale'].append(autocorr)
        randomness_analysis['predictability_by_scale'].append(predictability)
        randomness_analysis['pattern_strength_by_scale'].append(pattern_strength)

    return randomness_analysis

# =================== COMPREHENSIVE VISUALIZATION ===================

def create_comprehensive_visualization(all_results):
    """
    Create comprehensive visualization of all test results
    """
    fig = plt.figure(figsize=(20, 16), facecolor=COLORS['background'])

    # Create complex subplot layout
    gs = fig.add_gridspec(4, 3, hspace=0.3, wspace=0.3)

    # 1. Dyadic iteration results
    ax1 = fig.add_subplot(gs[0, 0])
    ax1.set_facecolor(COLORS['background'])

    dyadic_data = all_results['dyadic_analysis']
    iterations = dyadic_data['iterations']
    sequence = dyadic_data['sequence']
    rail_positions = dyadic_data['rail_positions']

    # Plot sequence evolution
    ax1.plot(iterations[:200], sequence[:200], color=COLORS['emergence'],
             linewidth=1, alpha=0.8, label='Dyadic Sequence')
    ax1.set_xlabel('Iteration', color=COLORS['text'])
    ax1.set_ylabel('Sequence Value', color=COLORS['text'])
    ax1.set_title('Binary Substrate Iteration', color=COLORS['text'])
    ax1.grid(True, alpha=0.3)
    ax1.legend()

    # 2. Rail structure visualization
    ax2 = fig.add_subplot(gs[0, 1])
    ax2.set_facecolor(COLORS['background'])

    phases = dyadic_data['phases']
    phase_colors = [COLORS['phi'], COLORS['pi'], COLORS['alpha'], COLORS['mirror']]

    for phase in range(4):
        phase_mask = phases[:200] == phase
        if np.any(phase_mask):
            ax2.scatter(iterations[:200][phase_mask], rail_positions[:200][phase_mask],
                       c=phase_colors[phase], s=10, alpha=0.7, label=f'Phase {phase}')

    ax2.set_xlabel('Iteration', color=COLORS['text'])
    ax2.set_ylabel('Log Rail Position', color=COLORS['text'])
    ax2.set_title('Period-4 Rail Structure', color=COLORS['text'])
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    # 3. Constant emergence
    ax3 = fig.add_subplot(gs[0, 2])
    ax3.set_facecolor(COLORS['background'])

    emergence_data = all_results['emergence_analysis']
    const_names = list(emergence_data['constant_matches'].keys())
    match_counts = [len(emergence_data['constant_matches'][name]) for name in const_names]

    bars = ax3.bar(range(len(const_names)), match_counts,
                   color=[COLORS['phi'], COLORS['pi'], COLORS['alpha'],
                         COLORS['emergence'], COLORS['mirror'], COLORS['binary']])
    ax3.set_xticks(range(len(const_names)))
    ax3.set_xticklabels(const_names, rotation=45, color=COLORS['text'])
    ax3.set_ylabel('Emergence Events', color=COLORS['text'])
    ax3.set_title('Constant Emergence Detection', color=COLORS['text'])
    ax3.grid(True, alpha=0.3)

    # 4. Prime residue analysis
    ax4 = fig.add_subplot(gs[1, 0])
    ax4.set_facecolor(COLORS['background'])

    prime_data = all_results['prime_analysis']
    residues = prime_data['residues']

    ax4.hist(residues, bins=50, color=COLORS['alpha'], alpha=0.7, edgecolor='none')
    ax4.axhline(y=len(residues)/137, color=COLORS['text'], linestyle='--', alpha=0.7,
                label='Expected Uniform')
    ax4.set_xlabel('Residue (mod 137)', color=COLORS['text'])
    ax4.set_ylabel('Frequency', color=COLORS['text'])
    ax4.set_title(f'Prime Residues (œá¬≤ = {prime_data["chi2_statistic"]:.2f})', color=COLORS['text'])
    ax4.legend()
    ax4.grid(True, alpha=0.3)

    # 5. FFT analysis
    ax5 = fig.add_subplot(gs[1, 1])
    ax5.set_facecolor(COLORS['background'])

    freqs = prime_data['fft_frequencies']
    powers = prime_data['fft_powers']

    ax5.plot(freqs, powers, color=COLORS['emergence'], linewidth=2)
    ax5.axvline(x=0.25, color=COLORS['warning'], linestyle='--', alpha=0.8,
                label='Quarter Cycle')
    ax5.set_xlabel('Frequency (cycles/bit)', color=COLORS['text'])
    ax5.set_ylabel('FFT Power', color=COLORS['text'])
    ax5.set_title('Prime Residue FFT Analysis', color=COLORS['text'])
    ax5.legend()
    ax5.grid(True, alpha=0.3)

    # 6. Phase transition analysis
    ax6 = fig.add_subplot(gs[1, 2])
    ax6.set_facecolor(COLORS['background'])

    if 'phase_transitions' in all_results:
        transition_data = all_results['phase_transitions']
        scaling_factors = transition_data['scaling_factors']
        phi_stability = transition_data['phi_stability']
        alpha_stability = transition_data['alpha_stability']

        ax6.plot(scaling_factors, phi_stability, 'o-', color=COLORS['phi'],
                label='œÜ Stability', linewidth=2)
        ax6.plot(scaling_factors, alpha_stability, 's-', color=COLORS['alpha'],
                label='Œ± Stability', linewidth=2)
        ax6.axvline(x=E, color=COLORS['warning'], linestyle='-', alpha=0.8,
                   label=f'e-threshold = {E:.3f}')
        ax6.set_xlabel('Scaling Factor', color=COLORS['text'])
        ax6.set_ylabel('Stability Measure', color=COLORS['text'])
        ax6.set_title('Phase Transition Analysis', color=COLORS['text'])
        ax6.legend()
        ax6.grid(True, alpha=0.3)

    # 7. Geometric relationships
    ax7 = fig.add_subplot(gs[2, 0])
    ax7.set_facecolor(COLORS['background'])

    if 'geometric_tests' in all_results:
        geo_data = all_results['geometric_tests']

        # Create geometric relationship network
        constants_geo = ['œÄ', '‚àö3', 'œÜ', 'Œ±‚Åª¬π', '‚àö2']
        values_geo = [PI, SQRT_3, PHI, ALPHA_INV, SQRT_2]

        # Reuleaux area connection
        reuleaux_area = geo_data['reuleaux_area']
        alpha_accuracy = geo_data['alpha_formula_accuracy']

        ax7.scatter(range(len(constants_geo)), values_geo,
                   s=[200*alpha_accuracy if c=='Œ±‚Åª¬π' else 100 for c in constants_geo],
                   c=[COLORS['alpha'] if c=='Œ±‚Åª¬π' else COLORS['phi'] for c in constants_geo],
                   alpha=0.8)

        ax7.set_xticks(range(len(constants_geo)))
        ax7.set_xticklabels(constants_geo, color=COLORS['text'])
        ax7.set_ylabel('Constant Value', color=COLORS['text'])
        ax7.set_title(f'Geometric Relationships (Œ± accuracy: {alpha_accuracy:.1%})', color=COLORS['text'])
        ax7.grid(True, alpha=0.3)

    # 8. Scale-dependent randomness
    ax8 = fig.add_subplot(gs[2, 1])
    ax8.set_facecolor(COLORS['background'])

    if 'randomness_analysis' in all_results:
        rand_data = all_results['randomness_analysis']
        scales = rand_data['scales']
        entropy = rand_data['entropy_by_scale']
        predictability = rand_data['predictability_by_scale']

        ax8.plot(scales, entropy, 'o-', color=COLORS['emergence'],
                label='Entropy', linewidth=2)
        ax8.plot(scales, predictability, 's-', color=COLORS['success'],
                label='Predictability', linewidth=2)
        ax8.set_xscale('log', base=2)
        ax8.set_xlabel('Observation Scale', color=COLORS['text'])
        ax8.set_ylabel('Measure', color=COLORS['text'])
        ax8.set_title('Scale-Dependent Randomness', color=COLORS['text'])
        ax8.legend()
        ax8.grid(True, alpha=0.3)

    # 9. Mirror relationships
    ax9 = fig.add_subplot(gs[2, 2])
    ax9.set_facecolor(COLORS['background'])

    mirror_data = emergence_data.get('mirror_relationships', {})
    if mirror_data:
        mirror_names = list(mirror_data.keys())
        mirror_accuracies = [1/(1+mirror_data[name]['accuracy']) for name in mirror_names]

        bars = ax9.bar(range(len(mirror_names)), mirror_accuracies,
                      color=COLORS['mirror'], alpha=0.8)
        ax9.set_xticks(range(len(mirror_names)))
        ax9.set_xticklabels(mirror_names, rotation=45, color=COLORS['text'])
        ax9.set_ylabel('Mirror Accuracy', color=COLORS['text'])
        ax9.set_title('Mirror Relationships (731‚Üî137 type)', color=COLORS['text'])
        ax9.grid(True, alpha=0.3)

    # 10-12. Summary statistics
    ax10 = fig.add_subplot(gs[3, :])
    ax10.set_facecolor(COLORS['background'])
    ax10.axis('off')

    # Summary text
    summary_text = f"""
COMPREHENSIVE BINARY SUBSTRATE ANALYSIS SUMMARY
‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê

Dyadic Iteration: {len(dyadic_data['sequence'])} iterations generated
Prime Analysis: {prime_data['n_primes']} primes analyzed (œá¬≤ = {prime_data['chi2_statistic']:.2f}, p = {prime_data['chi2_p_value']:.2e})
Constant Emergence: {sum(len(matches) for matches in emergence_data['constant_matches'].values())} total matches detected
Geometric Accuracy: Œ± formula achieves {geo_data['alpha_formula_accuracy']:.1%} precision
Phase Transitions: Analyzed across {len(transition_data['scaling_factors']) if 'phase_transitions' in all_results else 0} scaling factors

Binary Substrate Hypothesis Status: {'SUPPORTED' if prime_data['chi2_p_value'] < 0.01 else 'INCONCLUSIVE'}
    """

    ax10.text(0.05, 0.5, summary_text, transform=ax10.transAxes,
             fontfamily='monospace', fontsize=10, color=COLORS['text'],
             verticalalignment='center')

    plt.suptitle('Binary Substrate Reality Tests - Comprehensive Analysis',
                fontsize=16, color=COLORS['text'], y=0.98)

    return fig

# =================== MAIN COMPREHENSIVE TEST EXECUTION ===================

def run_comprehensive_tests():
    """
    Execute all tests in logical sequence
    """
    print("üî¨ PHASE 1: Binary Substrate Generation")
    start_time = time.time()

    # Generate multiple substrate sequences
    fib_word = generate_fibonacci_word_ultra(500)
    fib_decimal = binary_to_decimal_substrate(fib_word)
    debruijn_seq = generate_de_bruijn_sequence(2, 4)  # B(2,4)

    print(f"  Fibonacci word: {len(fib_word)} bits")
    print(f"  Fibonacci decimal: {fib_decimal:.15f}")
    print(f"  First 50 bits: {fib_word[:50]}")
    print(f"  De Bruijn B(2,4): {debruijn_seq}")
    print()

    print("üî¨ PHASE 2: Advanced Dyadic Iteration Analysis")

    # Run advanced dyadic analysis
    dyadic_results = advanced_dyadic_iteration(
        start_value=PHI - 1,
        fib_decimal=fib_decimal,
        n_iterations=1024,
        modulation=True,
        track_phases=True
    )

    print(f"  Generated {len(dyadic_results['sequence'])} iterations")
    print(f"  Mean crystallization: {np.mean(dyadic_results['crystallization_measure']):.6f}")
    print(f"  Substrate influence: {np.mean(dyadic_results['substrate_influence']):.6f}")
    print()

    print("üî¨ PHASE 3: Constant Emergence Analysis")

    emergence_analysis = analyze_constant_emergence(dyadic_results)

    total_matches = sum(len(matches) for matches in emergence_analysis['constant_matches'].values())
    print(f"  Total constant matches detected: {total_matches}")

    for const_name, matches in emergence_analysis['constant_matches'].items():
        if matches:
            best_match = min(matches, key=lambda x: x[2])  # Best by error
            print(f"    {const_name}: {len(matches)} matches, best error: {best_match[2]:.2e}")
    print()

    print("üî¨ PHASE 4: Prime Residue Structure Analysis")

    prime_analysis = analyze_prime_residues(modulus=137, n_primes=25000)  # Reduced for speed

    print(f"  Analyzed {prime_analysis['n_primes']} primes mod {prime_analysis['modulus']}")
    print(f"  œá¬≤ statistic: {prime_analysis['chi2_statistic']:.2f} (p = {prime_analysis['chi2_p_value']:.2e})")
    print(f"  Uniformity deviation: {prime_analysis['uniformity_deviation']:.2f}")
    print(f"  Quarter-cycle power fraction: {prime_analysis['quarter_cycle_fraction']:.4f}")
    print()

    print("üî¨ PHASE 5: Geometric Relationship Testing")

    geometric_tests = test_reuleaux_relationships()

    print(f"  Reuleaux area: {geometric_tests['reuleaux_area']:.8f}")
    print(f"  Œ± formula result: {geometric_tests['alpha_formula_result']:.8f}")
    print(f"  Œ± formula accuracy: {geometric_tests['alpha_formula_accuracy']:.1%}")
    print(f"  Reuleaux connection error: {geometric_tests['reuleaux_connection_error']:.2e}")
    print()

    print("üî¨ PHASE 6: Phase Transition Analysis")

    phase_transitions = analyze_phase_transitions(scaling_range=(1.0, 4.0), n_points=15)  # Focused range

    e_idx = np.argmin(np.abs(phase_transitions['scaling_factors'] - E))
    phi_stability_at_e = phase_transitions['phi_stability'][e_idx]
    alpha_stability_at_e = phase_transitions['alpha_stability'][e_idx]

    print(f"  œÜ stability at e-threshold: {phi_stability_at_e:.4f}")
    print(f"  Œ± stability at e-threshold: {alpha_stability_at_e:.4f}")
    print(f"  Phase transition detected: {'YES' if phi_stability_at_e > 0.5 else 'NO'}")
    print()

    print("üî¨ PHASE 7: Scale-Dependent Randomness Analysis")

    randomness_analysis = test_scale_dependent_randomness(fib_word)

    scales = randomness_analysis['scales']
    entropies = randomness_analysis['entropy_by_scale']

    print(f"  Entropy at scale 1: {entropies[0]:.4f}")
    print(f"  Entropy at scale {scales[-1]}: {entropies[-1]:.4f}")
    print(f"  Scale-dependent pattern: {'DETECTED' if entropies[0] > entropies[-1] else 'NOT DETECTED'}")
    print()

    elapsed_time = time.time() - start_time
    print(f"‚è±Ô∏è  Analysis completed in {elapsed_time:.2f} seconds")
    print()

    # Compile all results
    all_results = {
        'dyadic_analysis': dyadic_results,
        'emergence_analysis': emergence_analysis,
        'prime_analysis': prime_analysis,
        'geometric_tests': geometric_tests,
        'phase_transitions': phase_transitions,
        'randomness_analysis': randomness_analysis,
        'substrate_sequences': {
            'fibonacci_word': fib_word,
            'fibonacci_decimal': fib_decimal,
            'debruijn_sequence': debruijn_seq
        },
        'execution_time': elapsed_time
    }

    return all_results

# =================== STATISTICAL VALIDATION ===================

def generate_final_assessment(results):
    """
    Generate final statistical assessment of binary substrate hypothesis
    """
    print("üìä FINAL STATISTICAL ASSESSMENT")
    print("=" * 50)

    evidence_scores = []

    # 1. Prime residue non-randomness
    prime_p_value = results['prime_analysis']['chi2_p_value']
    if prime_p_value < 0.01:
        prime_evidence = 1.0
        print(f"‚úÖ Prime residue structure: SIGNIFICANT (p = {prime_p_value:.2e})")
    elif prime_p_value < 0.05:
        prime_evidence = 0.5
        print(f"‚ö†Ô∏è  Prime residue structure: MARGINAL (p = {prime_p_value:.2e})")
    else:
        prime_evidence = 0.0
        print(f"‚ùå Prime residue structure: NOT SIGNIFICANT (p = {prime_p_value:.2e})")

    evidence_scores.append(prime_evidence)

    # 2. Constant emergence detection
    total_matches = sum(len(matches) for matches in results['emergence_analysis']['constant_matches'].values())
    if total_matches > 10:
        emergence_evidence = 1.0
        print(f"‚úÖ Constant emergence: DETECTED ({total_matches} matches)")
    elif total_matches > 3:
        emergence_evidence = 0.5
        print(f"‚ö†Ô∏è  Constant emergence: PARTIAL ({total_matches} matches)")
    else:
        emergence_evidence = 0.0
        print(f"‚ùå Constant emergence: MINIMAL ({total_matches} matches)")

    evidence_scores.append(emergence_evidence)

    # 3. Geometric relationship accuracy
    alpha_accuracy = results['geometric_tests']['alpha_formula_accuracy']
    if alpha_accuracy > 0.99:
        geometric_evidence = 1.0
        print(f"‚úÖ Geometric relationships: HIGH ACCURACY ({alpha_accuracy:.1%})")
    elif alpha_accuracy > 0.95:
        geometric_evidence = 0.5
        print(f"‚ö†Ô∏è  Geometric relationships: MODERATE ACCURACY ({alpha_accuracy:.1%})")
    else:
        geometric_evidence = 0.0
        print(f"‚ùå Geometric relationships: LOW ACCURACY ({alpha_accuracy:.1%})")

    evidence_scores.append(geometric_evidence)

    # 4. Phase transitions
    if 'phase_transitions' in results:
        e_idx = np.argmin(np.abs(results['phase_transitions']['scaling_factors'] - E))
        stability_at_e = results['phase_transitions']['phi_stability'][e_idx]

        if stability_at_e > 0.7:
            phase_evidence = 1.0
            print(f"‚úÖ Phase transitions: DETECTED (stability = {stability_at_e:.3f})")
        elif stability_at_e > 0.5:
            phase_evidence = 0.5
            print(f"‚ö†Ô∏è  Phase transitions: PARTIAL (stability = {stability_at_e:.3f})")
        else:
            phase_evidence = 0.0
            print(f"‚ùå Phase transitions: NOT DETECTED (stability = {stability_at_e:.3f})")

        evidence_scores.append(phase_evidence)

    # 5. Scale-dependent randomness
    entropy_scale_1 = results['randomness_analysis']['entropy_by_scale'][0]
    entropy_scale_max = results['randomness_analysis']['entropy_by_scale'][-1]

    if entropy_scale_1 > entropy_scale_max + 0.1:
        randomness_evidence = 1.0
        print(f"‚úÖ Scale-dependent randomness: CONFIRMED (Œîentropy = {entropy_scale_1 - entropy_scale_max:.3f})")
    elif entropy_scale_1 > entropy_scale_max:
        randomness_evidence = 0.5
        print(f"‚ö†Ô∏è  Scale-dependent randomness: WEAK (Œîentropy = {entropy_scale_1 - entropy_scale_max:.3f})")
    else:
        randomness_evidence = 0.0
        print(f"‚ùå Scale-dependent randomness: NOT DETECTED (Œîentropy = {entropy_scale_1 - entropy_scale_max:.3f})")

    evidence_scores.append(randomness_evidence)

    # Overall assessment
    overall_score = np.mean(evidence_scores)

    print()
    print(f"üìà OVERALL EVIDENCE SCORE: {overall_score:.2f}/1.00")

    if overall_score >= 0.8:
        assessment = "STRONG SUPPORT"
        color = "‚úÖ"
    elif overall_score >= 0.6:
        assessment = "MODERATE SUPPORT"
        color = "‚ö†Ô∏è "
    elif overall_score >= 0.4:
        assessment = "WEAK SUPPORT"
        color = "‚ö†Ô∏è "
    else:
        assessment = "INSUFFICIENT EVIDENCE"
        color = "‚ùå"

    print(f"{color} BINARY SUBSTRATE HYPOTHESIS: {assessment}")
    print()
    print("Note: This assessment is based on mathematical pattern detection.")
    print("Results require peer review and independent replication.")

    return {
        'evidence_scores': evidence_scores,
        'overall_score': overall_score,
        'assessment': assessment
    }

# =================== MAIN EXECUTION ===================

if __name__ == "__main__":

    try:
        # Run comprehensive test suite
        print("üöÄ INITIATING COMPREHENSIVE BINARY SUBSTRATE ANALYSIS")
        print()

        all_results = run_comprehensive_tests()

        print("üé® GENERATING COMPREHENSIVE VISUALIZATION")

        fig = create_comprehensive_visualization(all_results)
        plt.show()

        print()

        # Generate final statistical assessment
        final_assessment = generate_final_assessment(all_results)

        print()
        print("=" * 70)
        print("‚úÖ COMPREHENSIVE ANALYSIS COMPLETED SUCCESSFULLY")
        print(f"‚è±Ô∏è  Total execution time: {all_results['execution_time']:.2f} seconds")
        print(f"üìä Evidence score: {final_assessment['overall_score']:.2f}")
        print(f"üéØ Assessment: {final_assessment['assessment']}")
        print("=" * 70)
        print()
        print("All results stored in 'all_results' variable for further analysis")
        print("Access individual components:")
        print("  - all_results['dyadic_analysis']")
        print("  - all_results['prime_analysis']")
        print("  - all_results['geometric_tests']")
        print("  - all_results['phase_transitions']")
        print("  - all_results['randomness_analysis']")

    except Exception as e:
        print(f"‚ùå ANALYSIS FAILED: {e}")
        import traceback
        traceback.print_exc()

print()
print("üåå Binary Substrate Reality Tests Complete!")
print("Testing whether mathematical reality emerges from computational substrate")
print("through systematic analysis of patterns across multiple domains.")
```
