```python
# ================================
# α encoder — quote the numbers
# ================================
import os, math, numpy as np, pandas as pd
import matplotlib.pyplot as plt

# Helper to load a CSV or say it's missing
def need(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"Missing file: {path}")
    return pd.read_csv(path)

# 1) Rank & p-value
try:
    outlier = need("alpha_outlier_score.csv")
    best_med = float(outlier.loc[0,"best_median_ppm"])
    rank     = int(outlier.loc[0,"rank_among_lanes"])
    nlanes   = int(outlier.loc[0,"num_lanes"])
    p_emp    = float(outlier.loc[0,"empirical_p_value"])
    gap2     = float(outlier.loc[0,"gap_to_2nd_best_ppm"])
    print("== Outlier score ==")
    print(f"  Best median ppm: {best_med:.6f}")
    print(f"  Rank among lanes: {rank} / {nlanes}  -> empirical p ≈ {p_emp:.3e}")
    print(f"  Gap to #2 (ppm): {gap2:.3f}")
except Exception as e:
    print("Outlier summary not found:", e)

# 2) Offset invariance (min/median/max per N)
try:
    off = need("alpha_offset_invariance.csv")
    byN = off.groupby("N")["ppm"].agg(["min","median","max"]).reset_index()
    print("\n== Offset invariance (ppm by N) ==")
    for _,r in byN.iterrows():
        print(f"  N={int(r['N'])}: min {r['min']:.6f} | median {r['median']:.6f} | max {r['max']:.6f}")
except Exception as e:
    print("Offset invariance not found:", e)

# 3) λ identifiability (local quadratic fit)
try:
    ident = need("alpha_lambda_identifiability.csv")
    a = float(ident.loc[0,"ppm_at_min_approx"])
    b = float(ident.loc[0,"slope_dppm_dppmLam"])   # ppm(α) per ppm(λ)
    c = float(ident.loc[0,"curvature"])
    print("\n== λ identifiability (local quadratic near Δλ=0 ppm) ==")
    print(f"  ppm at min (≈a): {a:.6f}")
    print(f"  slope d(ppm)/d(Δλ_ppm): {b:.3f}  (ppm-in-α per ppm-in-λ)")
    print(f"  curvature: {c:.3f}")
except Exception as e:
    print("λ identifiability not found:", e)

# 4) λ-robustness minima per N
try:
    rob = need("alpha_lambda_robustness.csv")
    mins=[]
    for N, sub in rob.groupby("N"):
        j = int(sub["ppm"].values.argmin())
        mins.append((int(N), float(sub.iloc[j]["d_ppm_on_lambda"]), float(sub.iloc[j]["ppm"])))
    print("\n== λ-robustness minima (per N) ==")
    for N, dppm, yppm in mins:
        print(f"  N={N}: min at Δλ≈{dppm:.1f} ppm -> ppm≈{yppm:.6f}")
except Exception as e:
    print("λ-robustness not found:", e)

# 5) Winner lane + complexity
winner = None
try:
    summ = need("alpha_summary.csv")
    winner = {
        "base": int(summ.loc[0,"best_base"]),
        "word": str(summ.loc[0,"best_word"]),
        "lambda": str(summ.loc[0,"best_lambda"]),
        "op": str(summ.loc[0,"best_op"]),
        "median_ppm": float(summ.loc[0,"best_median_ppm"]),
        "plateau": bool(summ.loc[0,"best_best_plateau_adjacent"]) if "best_best_plateau_adjacent" in summ.columns else bool(summ.loc[0,"best_plateau_adjacent"])
    }
except Exception:
    pass

try:
    lanes = need("alpha_best_lanes.csv")
    if winner is None:
        # fall back: first row is already the best (file sorted)
        row0 = lanes.iloc[0]
        winner = {
            "base": int(row0["base"]), "word": str(row0["word"]),
            "lambda": str(row0["lambda"]), "op": f"{row0['op_kind']}/{row0['op_code']}",
            "median_ppm": float(row0["median_ppm"]), "plateau": bool(row0["plateau_adjacent"])
        }
    # find winner complexity
    mask = (lanes["base"]==winner["base"]) & (lanes["word"]==winner["word"]) & (lanes["lambda"]==winner["lambda"]) \
           & (lanes["op_kind"]==winner["op"].split('/')[0]) & (lanes["op_code"]==winner["op"].split('/')[1])
    wcomp = float(lanes.loc[mask, "complexity"].iloc[0]) if mask.any() else float('nan')
    print("\n== Winner lane & complexity ==")
    print(f"  base={winner['base']}, word={winner['word']}, λ={winner['lambda']}, op={winner['op']}")
    print(f"  median ppm: {winner['median_ppm']:.6f}, plateau: {winner['plateau']}, complexity: {wcomp:.1f}")
except Exception as e:
    print("Best lanes file not found or format mismatch:", e)

# --------------------------
# Small, clean plots
# --------------------------

# A) λ-robustness curves
try:
    if 'rob' not in globals():
        rob = need("alpha_lambda_robustness.csv")
    plt.figure()
    for N, sub in rob.groupby("N"):
        plt.plot(sub["d_ppm_on_lambda"], sub["ppm"], label=f"N={int(N)}")
    plt.axvline(0, linestyle='--')
    plt.xlabel("Δλ (ppm)"); plt.ylabel("ppm to α⁻¹")
    plt.title("λ-robustness around best λ")
    plt.legend(); plt.tight_layout()
    plt.savefig("fig_lambda_robustness.png", dpi=150)
    print("\nSaved: fig_lambda_robustness.png")
except Exception as e:
    print("Robustness plot skipped:", e)

# B) Offset invariance scatter
try:
    if 'off' not in globals():
        off = need("alpha_offset_invariance.csv")
    plt.figure()
    # jitter x by tiny amount to see overlapping points per N
    xs = off["N"] + (np.random.RandomState(0).rand(len(off)) - 0.5) * 2.0
    plt.scatter(xs, off["ppm"], s=12)
    plt.xlabel("N (with tiny jitter)"); plt.ylabel("ppm to α⁻¹")
    plt.title("Offset/window invariance")
    plt.tight_layout()
    plt.savefig("fig_offset_invariance.png", dpi=150)
    print("Saved: fig_offset_invariance.png")
except Exception as e:
    print("Offset invariance plot skipped:", e)

# C) Perturbation error bars
try:
    pert = need("alpha_perturbation.csv")
    plt.figure()
    for N, sub in pert.groupby("N"):
        plt.errorbar(sub["flip_rate"], sub["ppm_mean"], yerr=sub["ppm_std"], marker='o', label=f"N={int(N)}")
    plt.xlabel("bit flip rate"); plt.ylabel("ppm to α⁻¹ (mean ± sd)")
    plt.title("Perturbation sensitivity")
    plt.legend(); plt.tight_layout()
    plt.savefig("fig_perturbation.png", dpi=150)
    print("Saved: fig_perturbation.png")
except Exception as e:
    print("Perturbation plot skipped:", e)

# D) Complexity–error scatter (sampled) with winner annotated
try:
    if 'lanes' not in globals():
        lanes = need("alpha_best_lanes.csv")
    # reservoir sample ~10k points to keep plotting light
    target = 10000
    rng = np.random.RandomState(42)
    if len(lanes) > target:
        idx = rng.choice(len(lanes), size=target, replace=False)
        samp = lanes.iloc[idx]
    else:
        samp = lanes
    plt.figure()
    plt.scatter(samp["complexity"].astype(float), samp["median_ppm"].astype(float), s=8)
    if winner is not None:
        wx = wcomp
        wy = winner["median_ppm"]
        plt.scatter([wx],[wy], s=50, marker='x')
        plt.annotate("winner", (wx, wy), xytext=(5,5), textcoords="offset points")
    plt.xlabel("description complexity (proxy)")
    plt.ylabel("median ppm to α⁻¹")
    plt.title("Complexity vs error (sampled lanes)")
    plt.tight_layout()
    plt.savefig("fig_complexity_vs_error.png", dpi=150)
    print("Saved: fig_complexity_vs_error.png")
except Exception as e:
    print("Complexity–error plot skipped:", e)
```
