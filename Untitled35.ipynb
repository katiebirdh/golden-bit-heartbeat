```python
# === ONE-CELL SETUP & RUN FOR TEST A ===
# It creates utils/, runs the Fibbit lane (λ = 729/1000 -> invert), writes CSV, shows a plot, and zips everything.

# 0) Minimal deps
!pip -q install mpmath

# 1) Imports & folders
import os, json, textwrap, csv, zipfile
from pathlib import Path
from math import isnan
from google.colab import files
from mpmath import mp
import sys # Import sys to modify the path

BASE = Path("math_emergence_tests")
UTILS = BASE / "utils"
NB_DIR = BASE / "notebooks"
ART = BASE / "artifacts" / "test_A"
for p in [UTILS, NB_DIR, ART]:
    p.mkdir(parents=True, exist_ok=True)

# Add the base directory to the system path
sys.path.insert(0, str(BASE))


# 2) Write utils/ modules
(UTILS / "precision.py").write_text(textwrap.dedent(r'''
"""Precision utilities for high-precision, reproducible math."""
from mpmath import mp
def set_precision(dps: int = 80):
    mp.dps = int(dps)
    return mp.dps
'''))

(UTILS / "binary_words.py").write_text(textwrap.dedent(r'''
"""
Binary/morphic words with simple, deterministic generators.
All functions return strings of '0'/'1' of at least length n.
"""
def fib_word(n: int) -> str:
    """Return Fibonacci word prefix of length >= n.
    F0='0', F1='01', Fk = F_{k-1} + F_{k-2}.
    """
    a, b = "0", "01"
    if n <= 1:
        return a[:n]
    s = b
    while len(s) < n:
        a, b = b, b + a
        s = b
    return s[:n]
'''))

(UTILS / "bigits.py").write_text(textwrap.dedent(r'''
"""Bigit encoders: interpret binary words as base-10 digit strings on purpose."""
from mpmath import mp

def to_bigit_decimal(word: str) -> mp.mpf:
    """Interpret '01001...' as decimal digits: 0.01001..."""
    s = "0." + "".join('0' if c == '0' else '1' for c in word)
    return mp.mpf(s)

def scale_lambda(x, num=729, den=1000):
    """Apply fixed rational scale λ = 729/1000 by default."""
    return x * (mp.mpf(num) / mp.mpf(den))

def invert(x):
    """Return 1/x (caller responsible for x != 0)."""
    return 1 / x

def leading_digits(y, k=14):
    """Return the first k decimal digits of y as 'DDD.ddd…' (padded)."""
    mp.dps += 10
    s = mp.nstr(y, n=k+10)
    if '.' not in s:
        s = s + '.0'
    whole, frac = s.split('.', 1)
    frac = (frac + "0"*k)[:k]
    return whole + "." + frac
'''))

(UTILS / "metrics.py").write_text(textwrap.dedent(r'''
"""Metrics and helpers for stability and comparison."""
def stable_prefix_len(values, k=16):
    if not values:
        return 0
    shortest = min(len(v) for v in values)
    d = 0
    for i in range(min(shortest, k)):
        ch = values[0][i]
        if all(v[i] == ch for v in values):
            d += 1
        else:
            break
    return d

def first_diff_index(values):
    if not values:
        return -1
    shortest = min(len(v) for v in values)
    for i in range(shortest):
        ch = values[0][i]
        if any(v[i] != ch for v in values[1:]):
            return i
    return -1
'''))


# 3) Run Test A (Fibbit lane)
from utils.precision import set_precision
from utils.binary_words import fib_word
from utils.bigits import to_bigit_decimal, scale_lambda, invert, leading_digits
from utils.metrics import stable_prefix_len, first_diff_index

set_precision(80)  # bump to 120 if you ever see flicker
Ns = [27, 81, 243, 729, 2187]
rows = []
ldigs = []
for N in Ns:
    w = fib_word(N)
    x = to_bigit_decimal(w)
    y = invert(scale_lambda(x))
    s = leading_digits(y, k=14)
    rows.append([N, str(x), str(scale_lambda(x)), str(y), s])
    ldigs.append(s)

csv_path = ART / "test_A_fibbit_alpha_inverse.csv"
with open(csv_path, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(["N", "x_bigit", "x_scaled", "y_inv", "leading_digits"])
    for r in rows: w.writerow(r)

print("CSV written:", csv_path)
print("Leading digits per N:", ldigs)
print("Stable prefix length (chars, up to 16):", stable_prefix_len(ldigs, k=min(16, len(ldigs[0]))))
print("First differing index among strings:", first_diff_index(ldigs))

# 4) Optional: quick plot (matplotlib, no custom colors/styles)
import matplotlib.pyplot as plt
ys = []
for N in Ns:
    w = fib_word(N)
    x = to_bigit_decimal(w)
    ys.append(float(invert(scale_lambda(x))))

plt.figure()
plt.plot(Ns, ys, marker='o')
plt.title('Test A: y = 1/(λ·Fibbit_N)')
plt.xlabel('N (Fibonacci word digits)')
plt.ylabel('y')
plt.grid(True)
plt.show()

# 5) Zip everything and prompt a download
zip_path = "math_emergence_testA_drop.zip"
with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as z:
    for root, _, files_ in os.walk(BASE):
        for name in files_:
            full = os.path.join(root, name)
            z.write(full, arcname=os.path.relpath(full, BASE))
print("Zipped:", zip_path)
files.download(zip_path)
```
