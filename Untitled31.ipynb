```python
# ================================================================
# PREREG ENGINE FOR  (fine-structure)  plateau, robustness, base-invariance,
# perturbation sensitivity, and complexity vs error  NO RIGGING.
#
# What this does:
#    Builds 0/1 words (Fib, TM, PD, RS, Rand, FibShuf)
#    Interprets them as base-B fractions (B  {2,3,5,10}) in (0,1)
#    Applies scale  from a small, symmetric, predeclared family near 1
#    Applies rails f  {S1..S4, T1..T16} (with inverses)
#    Targets ^{-1} = 137.035999084 (ppm)
#    Finds lanes with lowest median ppm across N  {12,27,81,243,729,2187}
#    Tests -robustness around the best lane (20 ppm)
#    Base-invariance: repeats search over bases
#    Perturbation: ppm vs flip-rate curve (random seeds fixed)
#    Complexity vs error: plots & CSVs
#
# Notes:
#   - Deterministic seeding (no randomness except controlled perturbations)
#   - Decimal high precision (sized by max N)
#   - Matplotlib only, one plot per figure, default colors
# ================================================================

from decimal import Decimal as D, getcontext, ROUND_HALF_EVEN
import math, numpy as np, pandas as pd
import itertools
import matplotlib.pyplot as plt

# ---------- Precision ----------
NS = [12, 27, 81, 243, 729, 2187]
MAX_N = max(NS)
getcontext().prec = max(2600, int(MAX_N*1.2) + 200)  # long enough for base-2 weights & ops
getcontext().rounding = ROUND_HALF_EVEN

# ---------- Target ----------
ALPHA_INV = D("137.035999084")

def ppm(v: D, ref: D) -> D:
    return (abs(v - ref) / ref) * D(1_000_000)

# ---------- Words ----------
def fib_word(n):
    a, b = "0", "01"
    while len(b) < n:
        a, b = b, b + a
    return b[:n]

def thue_morse(n):
    return ''.join('1' if bin(i).count('1')%2 else '0' for i in range(n))

def period_doubling(n):
    s="0"
    while len(s)<n:
        s = s.replace("0","X").replace("1","Y")
        s = s.replace("X","01").replace("Y","00")
    return s[:n]

def rudin_shapiro(n):
    out=[]
    for i in range(n):
        b = bin(i)[2:]
        cnt = sum(1 for j in range(len(b)-1) if b[j]=='1' and b[j+1]=='1')
        out.append('1' if cnt%2 else '0')
    return ''.join(out)

def random_word(n, seed=123456):
    rng = np.random.RandomState(seed)
    return ''.join(str(int(x)) for x in rng.randint(0,2,size=n))

def fib_shuffled(n, seed=424242):
    digs = list(fib_word(n))
    rng = np.random.RandomState(seed)
    rng.shuffle(digs)
    return ''.join(digs)

WORDS = {
    "Fib": fib_word,
    "TM": thue_morse,
    "PD": period_doubling,
    "RS": rudin_shapiro,
    "Rand": random_word,
    "FibShuf": fib_shuffled,
}

# ---------- Interpret as base-B fraction in (0,1) ----------
def word_to_fraction_base(digits_str: str, B: int) -> D:
    w = D(0)
    weight = D(1) / D(B)
    for ch in digits_str:
        if ch == '1':
            w += weight
        weight = weight / D(B)
    return w

BASES = [2,3,5,10]

# ---------- Rails (ops) and inverses ----------
def S1(x): return D(1) - x
def S2(x): return None if x == 0 else D(1) / x
def S3(x): return None if x == 1 else x / (D(1) - x)
def S4(x): return -x

OPS = {"S1":S1, "S2":S2, "S3":S3, "S4":S4}
INV = {"S1":S1, "S2":S2, "S3":(lambda y: None if y == -1 else y/(D(1)+y)), "S4":S4}

T = {
    "T1":("S2","S1"), "T2":("S1","S2"), "T3":("S1","S3"), "T4":("S3","S2"),
    "T5":("S3","S1"), "T6":("S4","S1"), "T7":("S1","S4"), "T8":("S4","S3"),
    "T9":("S3","S4"), "T10":("S2","S4"), "T11":("S4","S2"), "T12":("S2","S3"),
    "T13":("S3","S3"), "T14":("S2","S2"), "T15":("S1","S1"), "T16":("S4","S4"),
}

def apply_seq(x: D, seq) -> D:
    y = x
    for s in seq:
        y = OPS[s](y)
        if y is None:
            return None
    return y

def inv_seq(y: D, seq) -> D:
    z = y
    for s in reversed(seq):
        z = INV[s](z)
        if z is None:
            return None
    return z

ALL_OPS = {("S",k):(k,) for k in OPS.keys()}
ALL_OPS.update({("T",k):T[k] for k in T.keys()})

# ---------- Scale grid (symmetric, predeclared; includes 729/1000 but not privileged) ----------
def generate_scales():
    #  = 2^u * 3^v * 5^w / 10^k, with small integer exponents (symmetric)
    scopes = {
        "u": range(-2,3),   # powers of 2
        "v": range(-6,7),   # powers of 3
        "w": range(-1,2),   # powers of 5
        "k": range(-3,4),   # powers of 10 in denominator exponent
    }
    vals = set()
    for u in scopes["u"]:
        for v in scopes["v"]:
            for w in scopes["w"]:
                for k in scopes["k"]:
                    lam = (D(2) ** u) * (D(3) ** v) * (D(5) ** w) / (D(10) ** k)
                    # Keep within a reasonable band around 1
                    if D("0.03") <= lam <= D("30"):
                        vals.add(lam)
    # Ensure exact 729/1000 is present (3^6/10^3)
    vals.add((D(3) ** 6) / (D(10) ** 3))
    # Return as sorted list of Decimals
    return sorted(vals)
LAMBDAS = generate_scales()

# ---------- Helper: complexity score (small = simpler) ----------
def complexity_score(word_name, base, op_kind, op_code, lam):
    # Decompose lam in our grid coordinates approximately from its decimal (we generated lam via small exponents).
    # Use a simple proxy: number of digits of numerator/denominator when rationalized, plus op depth + word/base indicators.
    # To stay deterministic and light, approximate by (length of lam as string) + depth.
    lam_str = format(lam, 'f').rstrip('0').rstrip('.') if '.' in format(lam, 'f') else str(lam)
    depth = 1 if op_kind=="S" else 2
    word_term = {"Fib":0,"TM":1,"PD":1,"RS":2,"Rand":3,"FibShuf":2}[word_name]
    base_term = {2:1,3:1,5:1,10:0}[base]
    return len(lam_str) + depth + word_term + base_term

# ---------- Core search (per base, all words/ops/scales) ----------
def evaluate_all(bases=BASES, ns=NS, words=WORDS, ops=ALL_OPS, lambdas=LAMBDAS):
    records = []
    for base in bases:
        for word_name, builder in words.items():
            for N in ns:
                digits = builder(N) if word_name not in ("Rand","FibShuf") else (builder(N, seed=777) if word_name=="Rand" else builder(N, seed=888))
                x_raw = word_to_fraction_base(digits, base)
                for lam in lambdas:
                    x0 = x_raw * lam
                    for (okind, oc), seq in ops.items():
                        y = apply_seq(x0, seq)
                        if y is None or (y.is_nan() if hasattr(y,'is_nan') else False):
                            fppm = float('inf')
                        else:
                            fppm = float(ppm(y, ALPHA_INV))
                        records.append({
                            "base": base, "word": word_name, "N": N, "lambda": str(lam),
                            "op_kind": okind, "op_code": oc,
                            "val": str(y) if y is not None else "",
                            "ppm": fppm,
                            "complexity": complexity_score(word_name, base, okind, oc, lam),
                        })
    df = pd.DataFrame(records)
    return df

print("Running preregistered search grid... (this can take a bit for N=2187; hang tight)")
df = evaluate_all()
df.to_csv("alpha_grid_results.csv", index=False)
print("Saved alpha_grid_results.csv")

# ---------- Identify best lanes (by median ppm across N), one per (base,word,op) ----------
def summarize_lanes(df):
    # group by lane (base, word, lambda, op), compute median ppm across N
    grp_cols = ["base","word","lambda","op_kind","op_code"]
    med = df.groupby(grp_cols, as_index=False)["ppm"].median().rename(columns={"ppm":"median_ppm"})
    # Also check plateau: at least one adjacent pair  50 ppm
    plates = []
    for key, sub in df.groupby(grp_cols):
        sub = sub.sort_values("N")
        arr = sub["ppm"].values
        plateau = any((arr[i] <= 50.0 and arr[i+1] <= 50.0) for i in range(len(arr)-1))
        plates.append((*key, plateau))
    plates = pd.DataFrame(plates, columns=grp_cols+["plateau_adjacent"])
    out = med.merge(plates, on=grp_cols)
    out["complexity"] = df.groupby(grp_cols, as_index=False)["complexity"].median()["complexity"]
    out = out.sort_values(["median_ppm","complexity"])
    return out

lanes = summarize_lanes(df)
lanes.to_csv("alpha_best_lanes.csv", index=False)
print("Saved alpha_best_lanes.csv")

print("\nTop 12 lanes by median ppm:")
print(lanes.head(12).to_string(index=False))

# ---------- Pick the single best lane for ROBUSTNESS (lowest median ppm) ----------
best = lanes.iloc[0]
best_base = int(best["base"])
best_word = best["word"]
best_lambda = D(best["lambda"])
best_okind = best["op_kind"]; best_oc = best["op_code"]
best_seq = ALL_OPS[(best_okind, best_oc)]
print("\nBest lane:", dict(base=best_base, word=best_word, lam=str(best_lambda), op=(best_okind,best_oc),
                         median_ppm=float(best['median_ppm']), plateau=bool(best['plateau_adjacent'])))

# ---------- Robustness: sweep 20 ppm around best_lambda ----------
def robustness_sweep(best_lambda, seq, base, word_name, ns=NS, ppm_window=20):
    deltas_ppm = np.linspace(-ppm_window, ppm_window, 41)
    rows=[]
    builder = WORDS[word_name]
    for N in ns:
        digits = builder(N) if word_name not in ("Rand","FibShuf") else (builder(N, seed=777) if word_name=="Rand" else builder(N, seed=888))
        x_raw = word_to_fraction_base(digits, base)
        for d in deltas_ppm:
            # convert ppm shift on  to multiplicative factor (1 + d*1e-6)
            lam = best_lambda * (D(1) + D(d) / D(1_000_000))
            y = apply_seq(x_raw * lam, seq)
            fppm = float(ppm(y, ALPHA_INV)) if y is not None else float('inf')
            rows.append({"N":N, "d_ppm_on_lambda":float(d), "ppm":fppm})
    return pd.DataFrame(rows)

rob = robustness_sweep(best_lambda, best_seq, best_base, best_word, NS, ppm_window=20)
rob.to_csv("alpha_lambda_robustness.csv", index=False)
print("Saved alpha_lambda_robustness.csv")

# Plot robustness (one figure)
plt.figure()
for N in NS:
    sub = rob[rob["N"]==N]
    plt.plot(sub["d_ppm_on_lambda"], sub["ppm"], label=f"N={N}")
plt.axvline(0, linestyle='--')
plt.xlabel("Relative tweak to  (ppm)")
plt.ylabel("ppm to ")
plt.title("-robustness around best lane")
plt.legend()
plt.tight_layout()
plt.savefig("alpha_lambda_robustness.png", dpi=150)
print("Saved alpha_lambda_robustness.png")

# ---------- Base invariance: for each base, best median ppm lane ----------
best_per_base = []
for B in BASES:
    lanes_B = lanes[lanes["base"]==B]
    if len(lanes_B)==0: continue
    best_per_base.append(lanes_B.iloc[0])
best_per_base = pd.DataFrame(best_per_base)
best_per_base.to_csv("alpha_best_per_base.csv", index=False)

print("\nBest lane per base (median ppm):")
print(best_per_base[["base","word","lambda","op_kind","op_code","median_ppm","plateau_adjacent"]].to_string(index=False))

# Plot best-per-base bar (single figure)
plt.figure()
plt.bar(best_per_base["base"].astype(str), best_per_base["median_ppm"].astype(float))
plt.xlabel("Base for fraction interpretation")
plt.ylabel("Best median ppm to ")
plt.title("Best attainable -lane per base (fixed grammar)")
plt.tight_layout()
plt.savefig("alpha_best_per_base.png", dpi=150)
print("Saved alpha_best_per_base.png")

# ---------- Perturbation sensitivity ----------
def flip_bits(s, rate, seed):
    if rate<=0: return s
    rng = np.random.RandomState(seed)
    n = len(s); k = int(round(rate*n))
    idx = rng.choice(n, size=k, replace=False)
    arr = np.array(list(s))
    for i in idx:
        arr[i] = '1' if arr[i]=='0' else '0'
    return ''.join(arr)

def perturbation_curve(seq, base, word_name, lam, rates=(0,0.001,0.01,0.05,0.1), seeds=(101,202,303), ns=(81,243,729)):
    rows=[]
    builder = WORDS[word_name]
    for N in ns:
        base_digits = builder(N) if word_name not in ("Rand","FibShuf") else (builder(N, seed=777) if word_name=="Rand" else builder(N, seed=888))
        for r in rates:
            ppms=[]
            for sd in seeds:
                pert = flip_bits(base_digits, r, sd)
                x_raw = word_to_fraction_base(pert, base)
                y = apply_seq(x_raw * lam, seq)
                fppm = float(ppm(y, ALPHA_INV)) if y is not None else float('inf')
                ppms.append(fppm)
            rows.append({"N":N, "flip_rate":r, "ppm_mean":float(np.mean(ppms)), "ppm_std":float(np.std(ppms))})
    return pd.DataFrame(rows)

pert = perturbation_curve(best_seq, best_base, best_word, best_lambda)
pert.to_csv("alpha_perturbation.csv", index=False)
print("Saved alpha_perturbation.csv")

# Plot perturbation (one figure)
plt.figure()
for N in sorted(pert["N"].unique()):
    sub = pert[pert["N"]==N]
    plt.errorbar(sub["flip_rate"], sub["ppm_mean"], yerr=sub["ppm_std"], marker='o', label=f"N={N}")
plt.xlabel("Bit flip rate")
plt.ylabel("ppm to  (mean  sd)")
plt.title("Perturbation sensitivity (best lane)")
plt.legend()
plt.tight_layout()
plt.savefig("alpha_perturbation.png", dpi=150)
print("Saved alpha_perturbation.png")

# ---------- Complexity vs error (Pareto-ish view) ----------
# Take for each lane its median ppm (across N) and median complexity; plot
plt.figure()
plt.scatter(lanes["complexity"].astype(float), lanes["median_ppm"].astype(float), s=12)
plt.xlabel("Description complexity (proxy)")
plt.ylabel("Median ppm to ")
plt.title("Complexity vs Error (lanes across grammar)")
plt.tight_layout()
plt.savefig("alpha_complexity_vs_error.png", dpi=150)
print("Saved alpha_complexity_vs_error.png")

# ---------- Optional: Energy-scale alignment (user-editable reference list) ----------
# This is a *context* check only (not a derivation). You can refine these references later.
REF_ALPHA_INV = {
    "low_energy": 137.035999084,   # CODATA 2018 static (approx)
    "mZ_pole_eff": 128.95,         # ballpark effective ^{-1}(m_Z^2)
}
# Compare our plateau value (median across N at best lane) to these references.
plateau_ppm_across_N = df[(df["base"]==best_base) &
                          (df["word"]==best_word) &
                          (df["lambda"]==str(best_lambda)) &
                          (df["op_kind"]==best_okind) &
                          (df["op_code"]==best_oc)].sort_values("N")[["N","ppm"]]
print("\nBest lane ppm across N:")
print(plateau_ppm_across_N.to_string(index=False))

# Save a compact summary
summary = {
    "best_base": best_base,
    "best_word": best_word,
    "best_lambda": str(best_lambda),
    "best_op": f"{best_okind}/{best_oc}",
    "best_median_ppm": float(best["median_ppm"]),
    "best_plateau_adjacent": bool(best["plateau_adjacent"]),
}
pd.DataFrame([summary]).to_csv("alpha_summary.csv", index=False)
print("\nSaved alpha_summary.csv")
print("\nAll done. Files saved:\n",
      "- alpha_grid_results.csv\n - alpha_best_lanes.csv\n - alpha_lambda_robustness.csv (and .png)\n",
      "- alpha_best_per_base.csv (and .png)\n - alpha_perturbation.csv (and .png)\n - alpha_complexity_vs_error.png\n - alpha_summary.csv")
```
