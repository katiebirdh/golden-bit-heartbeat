```python
# ===================== α GEOMETRY — OVERHAUL v2 (SINGLE CELL) =====================
# Robust tests with fixed families, mirrored nulls, mass summaries, cross-base maps,
# exact Fibonacci morphism, residual rungs, and complex-lift + tiny Möbius scan.
# Safe defaults; runs quickly. You can widen ranges or TRIALS later.

FAST_MODE  = True
TRIALS     = 500 if FAST_MODE else 5000
D_RANGE    = (40, 260)              # denominators
K_RANGE    = (40, 260)              # multipliers
U_RANGE    = range(-3, 4)           # u ∈ [-3..3]
V_RANGE    = range(-4, 5)           # v ∈ [-4..4]
FIB_NS     = [20, 40, 80, 120, 160, 200, 240, 320]
THETAS_N   = 181 if FAST_MODE else 721
THETAS_MOB = 121 if FAST_MODE else 361

import numpy as np, math, itertools
import mpmath as mp
import matplotlib.pyplot as plt
mp.mp.dps = 60

alpha = mp.mpf("0.007297352566417119")
phi   = (1 + mp.sqrt(5)) / 2
sqrt3 = mp.sqrt(3); pi = mp.pi; lnphi = mp.log(phi)

# --- Two compact approximants ---
A_101  = 101*(phi**(pi-2) - sqrt3)
A_110  = (phi**pi - 2 - sqrt3)/110
def report(label, val):
    ae = mp.fabs(val - alpha); re = ae/alpha
    print(f"{label:30s} = {val}"); print(f"  abs err = {ae},   rel err = {re}")
report("A_101  = 101(φ^{π-2} - √3)", A_101)
report("A_110  = (φ^π - 2 - √3)/110", A_110)

xs = [float(A_110), float(alpha), float(A_101)]
labels = ["(φ^π-2-√3)/110","α (target)","101(φ^{π-2}-√3)"]
plt.figure(figsize=(8,1.6)); plt.scatter(xs,[0,0,0],s=50)
for x,lab in zip(xs, labels): plt.text(x,0.03,lab,rotation=45,ha='left',va='bottom')
plt.hlines(0, min(xs)-1e-5, max(xs)+1e-5); plt.yticks([])
plt.title("Two small-integer φ–π–√3 approximants that bracket α"); plt.xlabel("value")
plt.tight_layout(); plt.show()

# --- Pre-registered wide grids ---
phi_pi = phi**pi; phi_pi_f, sqrt3_f, alpha_f = float(phi_pi), float(sqrt3), float(alpha)
D = np.arange(D_RANGE[0], D_RANGE[1]+1, dtype=np.float64)
K = np.arange(K_RANGE[0], K_RANGE[1]+1, dtype=np.float64)

def search_A():
    best = (np.inf, None, None, None); all_errs = []
    for u in U_RANGE:
        for v in V_RANGE:
            vals = (phi_pi_f + u*sqrt3_f + v) / D
            errs = np.abs(vals - alpha_f); all_errs.append(errs)
            j = np.argmin(errs)
            if errs[j] < best[0]: best = (float(errs[j]), int(D[j]), int(u), int(v))
    return best, np.concatenate(all_errs)
def search_B():
    t = float(phi**(pi-2) - sqrt3); vals = K * t; errs = np.abs(vals - alpha_f)
    j = np.argmin(errs); return (float(errs[j]), int(K[j])), errs

bestA, massA = search_A(); bestB, massB = search_B()
print("\nBest in A-family (grid): err=%.3e at d=%d, u=%+d, v=%+d" % bestA)
print("Best in B-family (grid):  err=%.3e at k=%d" % bestB)

def summary(arr):
    return dict(min=float(arr.min()), p10=float(np.percentile(arr,10)),
                median=float(np.median(arr)), p90=float(np.percentile(arr,90)),
                max=float(arr.max()))
print("\nA-mass summary:", summary(massA)); print("B-mass summary:", summary(massB))
plt.figure(); plt.hist(massA, bins=80); plt.xlabel("|A(d,u,v)-α|"); plt.ylabel("count")
plt.title("A-family error mass over grid"); plt.tight_layout(); plt.show()
plt.figure(); plt.hist(massB, bins=80); plt.xlabel("|B(k)-α|"); plt.ylabel("count")
plt.title("B-family error mass over grid"); plt.tight_layout(); plt.show()

# --- Null models mirroring each search ---
rng = np.random.default_rng(12345)
eps_pi   = rng.uniform(-0.005, 0.005, TRIALS)
eps_lnph = rng.uniform(-0.003, 0.003, TRIALS)
eps_s3   = rng.uniform(-0.02,  0.02,  TRIALS)
phi_pi_j = np.exp((float(pi)+eps_pi)*(float(lnphi)+eps_lnph))
sqrt3_j  = sqrt3_f + eps_s3

best_null_A = np.full(TRIALS, np.inf)
for u in U_RANGE:
    for v in V_RANGE:
        numer = phi_pi_j[:,None] + u*sqrt3_j[:,None] + v
        errs  = np.abs(numer / D[None,:] - alpha_f)
        best_null_A = np.minimum(best_null_A, errs.min(axis=1))
t_prime = np.exp((float(pi)+eps_pi-2.0)*(float(lnphi)+eps_lnph)) - sqrt3_j
best_null_B = np.abs(t_prime[:,None]*K[None,:] - alpha_f).min(axis=1)

best_err_A, best_err_B = bestA[0], bestB[0]
cap = 1.0/(TRIALS+1)
pA = float((best_null_A <= best_err_A).mean()); pB = float((best_null_B <= best_err_B).mean())
print("\nEmpirical p-values (null ≤ observed):")
print("  A-family search p =", f"< {cap:.4f}" if pA==0 else f"{pA:.4f}")
print("  B-family search p =", f"< {cap:.4f}" if pB==0 else f"{pB:.4f}")
plt.figure(); plt.hist(best_null_A, bins=80); plt.axvline(best_err_A, color='k', ls='--', lw=2)
plt.xlabel("best abs error under null (A-family)"); plt.ylabel("count"); plt.title("Null A"); plt.tight_layout(); plt.show()
plt.figure(); plt.hist(best_null_B, bins=80); plt.axvline(best_err_B, color='k', ls='--', lw=2)
plt.xlabel("best abs error under null (B-family)"); plt.ylabel("count"); plt.title("Null B"); plt.tight_layout(); plt.show()

# --- Fibonacci word via exact morphism 0->01, 1->0 ---
def fib_bits_morph(N: int):
    s = "0"; out = []
    while len(out) < N:
        s = ''.join(("01" if c=='0' else "0") for c in s)
        need = min(N - len(out), len(s))
        out.extend(1 if c=='1' else 0 for c in s[:need])
        s = s[need:]
    return np.array(out, dtype=np.int8)

def thue_morse_bits(N: int):
    x = np.array([0], dtype=np.int8)
    while len(x) < N: x = np.concatenate([x, 1-x])
    return x[:N]
def periodic_bits(N: int): return np.resize(np.array([0,1], dtype=np.int8), N)
def random_bits(N: int, seed=123):
    r = np.random.default_rng(seed); return r.integers(0,2,N, dtype=np.int8)

def D_fraction_base(bits: np.ndarray, base: int) -> float:
    n = np.arange(1, len(bits)+1, dtype=np.float64)
    return float((bits * (base**(-n))).sum())
def triadic_scale(base: int) -> float: return (3**6)/(base**3)
def map_err(bits, base): return abs(triadic_scale(base)*D_fraction_base(bits, base) - float(alpha))

print("\nFibonacci (morphism) prefix:", ''.join(map(str, fib_bits_morph(20))))

bases = [10,8,12]
curves = {("Fibonacci",b):[] for b in bases}
curves.update({("Thue–Morse",10):[], ("periodic 01",10):[], ("random",10):[]})

for N in FIB_NS:
    F = fib_bits_morph(N)
    for b in bases: curves[("Fibonacci",b)].append(map_err(F,b))
    curves[("Thue–Morse",10)].append(map_err(thue_morse_bits(N),10))
    curves[("periodic 01",10)].append(map_err(periodic_bits(N),10))
    curves[("random",10)].append(map_err(random_bits(N,seed=100+N),10))

plt.figure()
for b in bases:
    plt.plot(FIB_NS, curves[("Fibonacci",b)], marker='o', label=f"Fibonacci, base-{b}")
plt.yscale('log'); plt.legend(); plt.xlabel("N"); plt.ylabel("error to α (log)")
plt.title("Fibonacci→α across bases (morphism)"); plt.tight_layout(); plt.show()

plt.figure()
plt.plot(FIB_NS, curves[("Fibonacci",10)], marker='o', label="Fibonacci")
plt.plot(FIB_NS, curves[("Thue–Morse",10)], marker='s', label="Thue–Morse")
plt.plot(FIB_NS, curves[("periodic 01",10)], marker='^', label="periodic 0101…")
plt.plot(FIB_NS, curves[("random",10)], marker='d', label="random")
plt.yscale('log'); plt.legend(); plt.xlabel("N"); plt.ylabel("error to α (log)")
plt.title("Sequence controls (base-10)"); plt.tight_layout(); plt.show()

# --- Residual rungs over the whole mass ---
bases_r = [2,3,5,7,10]
def best_rung(ratio, K=12):
    best = (1e9,None,None)
    for b in bases_r:
        for k in range(1,K+1):
            rel = abs(ratio/(b**(-k)) - 1.0)
            if rel < best[0]: best = (rel,b,k)
    return best
ratios = []
for u in U_RANGE:
    for v in V_RANGE:
        vals = (phi_pi_f + u*sqrt3_f + v) / D
        ratios.extend((np.abs(vals - alpha_f)/alpha_f).tolist())
t = float(phi**(pi-2) - sqrt3)
ratios.extend((np.abs(K*t - alpha_f)/alpha_f).tolist())

from collections import Counter
bases_best = [best_rung(r,12)[1] for r in ratios]
cnt = Counter(bases_best); xs = sorted(cnt.keys()); ys = [cnt[x] for x in xs]
plt.figure(); plt.bar([str(x) for x in xs], ys)
plt.xlabel("base b of best rung"); plt.ylabel("count"); plt.title("Residual rung preference (mass)")
plt.tight_layout(); plt.show()
print("Rung counts:", dict(cnt))

# --- Complex lift + tiny Möbius ---
def D10(bits):
    n = np.arange(1, len(bits)+1, dtype=np.float64)
    return float((bits * (10.0**(-n))).sum())
N = 320; F = fib_bits_morph(N); R = 1 - F
Z = complex(D10(F), D10(R)); S = 729/1000
thetas = np.linspace(0, 2*np.pi, THETAS_N)
base_curve = []
for th in thetas:
    zt = Z * complex(math.cos(th), math.sin(th))
    base_curve.append(abs(S*zt.real - float(alpha)))
base_min = min(base_curve)

COEF = [-1,0,1]
best_mob = (np.inf, None, None, None, None)
for a,b,c,d in itertools.product(COF, COEF, COEF, COEF):
    if a*d - b*c == 0: continue
    best_local = np.inf; theta_star = None
    for th in np.linspace(0, 2*np.pi, THETAS_MOB):
        zt = Z * complex(math.cos(th), math.sin(th))
        denom = c*zt + d
        if abs(denom) < 1e-14: continue
        w = (a*zt + b) / denom
        e = abs(S*w.real - float(alpha))
        if e < best_local: best_local, theta_star = e, th
    if best_local < best_mob[0]: best_mob = (best_local,a,b,c,d,theta_star)

print("\nComplex-lift baseline min error:", base_min)
print("Best Möbius (a,b,c,d,θ*):", best_mob[1:6])
print("Best Möbius min error   :", best_mob[0])
if best_mob[0] < base_min: print("Improvement factor      :", base_min/best_mob[0])
print("\nDone.")
```
