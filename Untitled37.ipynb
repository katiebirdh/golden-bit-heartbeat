```python
# === Lean Prime φ Scan: fast EMD sweeps + scaling ===
import os, csv, math, json, numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
rng = default_rng(137)

BASE_DIR = "phi_lean_scan"
os.makedirs(BASE_DIR, exist_ok=True)

# ---- knobs: keep these small to stay fast ----
WINDOWS = [(10**6, 5*10**6), (10**7, 5*10**7)]   # 2 windows
CAP_PER_WINDOW = 60_000                           # max primes per window
GRID = np.round(np.arange(1.55, 1.70+1e-9, 0.005), 3)  # coarse grid
phi = (1 + 5**0.5) / 2
NEIGHBORS = [round(phi-0.01,5), round(phi+0.01,5)]     # for scaling check
SUBBINS = 8                                            # sub-windows per window

def primes_in_window(lo, hi, cap):
    from sympy import primerange
    arr = np.fromiter(primerange(int(lo), int(hi)), dtype=np.int64)
    if arr.size > cap:
        idx = rng.choice(arr.size, size=cap, replace=False)
        arr = arr[idx]
    return np.sort(arr)

def emd_hist_to_uniform01(phases, bins=120):
    # 1D EMD vs uniform via histogram CDFs (fast)
    x = np.asarray(phases, dtype=float) % 1.0
    h, edges = np.histogram(x, bins=bins, range=(0,1), density=True)
    centers = (edges[:-1] + edges[1:]) / 2
    cdf = np.cumsum(h)
    if cdf[-1] == 0: return 0.0
    cdf = cdf / cdf[-1]
    cdf_u = centers  # CDF of U[0,1]
    return float(np.trapezoid(np.abs(cdf - cdf_u), centers))

def phase_log_base(primes, b):
    return (np.log(primes)/np.log(b)) % 1.0

summary_rows = []
scaling_rows = []

# ---------- sweep: EMD vs base (per window) ----------
for (lo, hi) in WINDOWS:
    primes = primes_in_window(lo, hi, CAP_PER_WINDOW)
    emds, bases = [], []
    for b in np.unique(np.concatenate([GRID, [phi]])):
        phases = phase_log_base(primes, b)
        emd = emd_hist_to_uniform01(phases, bins=120)
        bases.append(float(b)); emds.append(emd)
        summary_rows.append([lo, hi, float(b), primes.size, emd])
    # quick plot
    plt.figure()
    order = np.argsort(bases)
    bb = np.array(bases)[order]; ee = np.array(emds)[order]
    plt.plot(bb, ee, marker='o', linewidth=1)
    plt.axvline(phi, linestyle="--")
    plt.title(f"EMD vs base — [{lo},{hi})  (n={primes.size})")
    plt.xlabel("base b"); plt.ylabel("EMD to uniform")
    plt.tight_layout()
    plt.savefig(f"{BASE_DIR}/emd_curve_{lo}_{hi}.png")

# ---------- scaling: φ vs neighbors across sub-windows ----------
for (lo, hi) in WINDOWS:
    primes_full = primes_in_window(lo, hi, CAP_PER_WINDOW*2)  # a bit more for slices
    cuts = np.linspace(lo, hi, SUBBINS+1, dtype=np.int64)
    for j in range(SUBBINS):
        slo, shi = int(cuts[j]), int(cuts[j+1])
        mask = (primes_full >= slo) & (primes_full < shi)
        p = primes_full[mask]
        if p.size < 2000:  # skip tiny slices
            continue
        for b in [phi] + NEIGHBORS:
            emd = emd_hist_to_uniform01(phase_log_base(p, b), bins=100)
            scaling_rows.append([lo, hi, j+1, int(p.size), float(b), emd])

# ---------- save CSVs + tiny report ----------
os.makedirs(f"{BASE_DIR}/artifacts", exist_ok=True)
with open(f"{BASE_DIR}/artifacts/emd_sweep_summary.csv","w") as f:
    w = csv.writer(f); w.writerow(["lo","hi","base","n_primes","emd"])
    w.writerows(summary_rows)

with open(f"{BASE_DIR}/artifacts/scaling_summary.csv","w") as f:
    w = csv.writer(f); w.writerow(["lo","hi","slice_index","n_primes","base","emd"])
    w.writerows(scaling_rows)

# simple JSON pointers for quick reading
def best_base(rows, lo, hi):
    R = [r for r in rows if r[0]==lo and r[1]==hi]
    if not R: return None
    rmax = max(R, key=lambda r:r[4])
    return {"window":[lo,hi], "best_base":rmax[2], "best_emd":rmax[4], "n":rmax[3]}

report = {
  "windows":[[int(a),int(b)] for a,b in WINDOWS],
  "phi": phi,
  "best_per_window": [best_base(summary_rows, lo, hi) for (lo,hi) in WINDOWS],
}
with open(f"{BASE_DIR}/artifacts/lean_report.json","w") as f:
    json.dump(report, f, indent=2)

print(json.dumps(report, indent=2))
print(f"\nWrote: {BASE_DIR}/artifacts/emd_sweep_summary.csv, scaling_summary.csv, lean_report.json")
print(f"Plots: emd_curve_{WINDOWS[0][0]}_{WINDOWS[0][1]}.png and _{WINDOWS[1][0]}_{WINDOWS[1][1]}.png")
```
