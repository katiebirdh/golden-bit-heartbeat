```python
# Corrected Harmonic Echo Tests - Revealing the Patterns

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import pandas as pd

# Set up plotting style
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.dpi'] = 100
plt.rcParams['font.size'] = 10

# Define constants
PI = np.pi
TRANSITIONAL_CONSTANT = 3.141412303134
DELTA = PI - TRANSITIONAL_CONSTANT
PI_HALF = PI / 2

print("🔢 Key Constants:")
print(f"π = {PI:.15f}")
print(f"Transitional constant = {TRANSITIONAL_CONSTANT:.15f}")
print(f"Δ (difference) = {DELTA:.15f}")
print(f"π/2 = {PI_HALF:.15f}")

# CORRECTED Test 1: Looking at the actual harmonic structure
def test1_harmonic_echo_corrected():
    """
    Test harmonic structure by examining how compressed Δ relates to π/2
    when divided by 729 and 799.
    """
    # Parameters
    power_range = np.linspace(30, 37, 100)
    denominators = [729, 799]

    # Storage for results
    results = {d: {
        'powers': [],
        'scaled_values': [],
        'ratio_to_pi_half': [],
        'log_ratio': []
    } for d in denominators}

    # Compute for each power and denominator
    for p in power_range:
        residue = DELTA / (9 ** p)

        for d in denominators:
            scaled = residue / d

            # Store the actual scaled value
            results[d]['powers'].append(p)
            results[d]['scaled_values'].append(scaled)

            # Calculate ratio to π/2 (this shows the relative scale)
            ratio = scaled / PI_HALF
            results[d]['ratio_to_pi_half'].append(ratio)

            # Log of the ratio (for visualization)
            if ratio > 0:
                results[d]['log_ratio'].append(np.log10(ratio))
            else:
                results[d]['log_ratio'].append(-400)  # Very small number

    return results

# Run corrected Test 1
print("\n🚀 Running Corrected Test 1...")
test1_results = test1_harmonic_echo_corrected()

# Visualization for corrected Test 1
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot scaled values and ratios
for idx, d in enumerate([729, 799]):
    # Scaled values (log scale)
    ax1 = axes[0, idx]
    ax1.semilogy(test1_results[d]['powers'],
                 test1_results[d]['scaled_values'],
                 'b-', linewidth=2)
    ax1.set_xlabel('Power p')
    ax1.set_ylabel('Scaled Value (Δ/9^p/d)')
    ax1.set_title(f'Scaled Values for d={d}')
    ax1.grid(True, alpha=0.3)

    # Add reference line for π/2
    ax1.axhline(y=PI_HALF, color='red', linestyle='--', label='π/2')
    ax1.legend()

    # Log ratio to π/2
    ax2 = axes[1, idx]
    ax2.plot(test1_results[d]['powers'],
             test1_results[d]['log_ratio'],
             'orange', linewidth=2)
    ax2.set_xlabel('Power p')
    ax2.set_ylabel('log₁₀(scaled/π/2)')
    ax2.set_title(f'Log Ratio to π/2 (d={d})')
    ax2.grid(True, alpha=0.3)

plt.suptitle('Corrected Test 1: Harmonic Structure Analysis', fontsize=16)
plt.tight_layout()
plt.show()

# Find where scaled values cross specific thresholds
print("\n📊 Threshold Analysis:")
for d in [729, 799]:
    # Find where scaled value is closest to π/2 * 10^-n for various n
    for n in [10, 20, 30, 35]:
        threshold = PI_HALF * (10 ** -n)
        distances = [abs(v - threshold) for v in test1_results[d]['scaled_values']]
        min_idx = np.argmin(distances)
        print(f"\nDenominator {d}:")
        print(f"  Closest to π/2 × 10^-{n}: at p={test1_results[d]['powers'][min_idx]:.2f}")
        print(f"  Actual value: {test1_results[d]['scaled_values'][min_idx]:.2e}")
        print(f"  Ratio to π/2: {test1_results[d]['ratio_to_pi_half'][min_idx]:.2e}")

# CORRECTED Test 2: Examining ratios and patterns
def test2_harmonic_patterns():
    """
    Test harmonic patterns by examining ratios between different compressions.
    """
    # Test multiple compression levels
    powers = [33.0, 33.5, 33.7, 34.0]
    denominators = [729, 799, 81, 101, 137, 888, 999]
    denominator_labels = ['729 (9³)', '799', '81 (3⁴)', '101 (prime)',
                         '137 (α)', '888', '999']

    results = []

    for p in powers:
        compressed = DELTA / (9 ** p)
        power_results = {
            'power': p,
            'compressed': compressed,
            'scaled_values': {},
            'ratios_to_729': {}
        }

        for i, d in enumerate(denominators):
            scaled = compressed / d
            power_results['scaled_values'][denominator_labels[i]] = scaled

            # Calculate ratio relative to 729 result
            if i == 0:  # 729 is the reference
                power_results['ratios_to_729'][denominator_labels[i]] = 1.0
            else:
                ratio = (compressed / d) / (compressed / 729)
                power_results['ratios_to_729'][denominator_labels[i]] = ratio

        results.append(power_results)

    return results

# Run corrected Test 2
print("\n🚀 Running Corrected Test 2...")
test2_results = test2_harmonic_patterns()

# Display results in a table
print("\n📊 Harmonic Ratio Analysis:")
for result in test2_results:
    print(f"\n🔹 Power p = {result['power']}")
    print(f"   Compressed Δ = {result['compressed']:.2e}")
    print("\n   Ratios relative to 729:")
    for denom, ratio in result['ratios_to_729'].items():
        if denom != '729 (9³)':
            print(f"   {denom:12s}: {ratio:.6f} (or {729/ratio:.2f})")

# Visualization for harmonic patterns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Ratios for different powers
denominators_to_plot = ['799', '81 (3⁴)', '101 (prime)', '137 (α)']
for denom in denominators_to_plot:
    ratios = [r['ratios_to_729'][denom] for r in test2_results]
    powers = [r['power'] for r in test2_results]
    ax1.plot(powers, ratios, 'o-', label=denom, markersize=8)

ax1.set_xlabel('Compression Power p')
ax1.set_ylabel('Ratio relative to 729')
ax1.set_title('Harmonic Ratios vs Compression Power')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Special relationships
# Check if 729/799 shows special properties
special_ratios = []
for result in test2_results:
    ratio_799 = result['ratios_to_729']['799']
    special_ratios.append(729/799 / ratio_799)

ax2.plot([r['power'] for r in test2_results], special_ratios, 'ro-', markersize=10)
ax2.set_xlabel('Compression Power p')
ax2.set_ylabel('(729/799) / Actual Ratio')
ax2.set_title('Deviation from Expected 729/799 Ratio')
ax2.grid(True, alpha=0.3)
ax2.axhline(y=1.0, color='green', linestyle='--', label='Perfect match')
ax2.legend()

plt.tight_layout()
plt.show()

# Final analysis: Echo mechanism with proper scaling
print("\n🔄 Echo Mechanism Analysis:")
print("\nThe echo mechanism works as follows:")
print("1. Start with compressed value: scaled = Δ/(9^p)/d")
print("2. The 'distance' from π/2 is: distance = π/2 - scaled")
print("3. Echo back to π: echo = π - distance = π - (π/2 - scaled)")
print("4. This simplifies to: echo = π/2 + scaled")
print("\nSince scaled << π/2, the echo ≈ π/2, but the tiny scaled value")
print("carries the harmonic information.")

# Demonstrate the harmonic content
p = 33.7
for d in [729, 799]:
    scaled = DELTA / (9**p) / d
    echo = PI/2 + scaled
    echo_error = abs(PI - echo)

    print(f"\nFor d={d} at p={p}:")
    print(f"  Scaled value: {scaled:.2e}")
    print(f"  Echo value: {echo:.15f}")
    print(f"  Echo error from π: {echo_error:.15f}")
    print(f"  Relative error: {echo_error/PI:.2e}")

# Additional insight: Examining the numerical relationships
print("\n🎯 Key Numerical Insights:")
print(f"\n729 = 9³ = 3⁶")
print(f"799 is interesting because 799/729 ≈ {799/729:.6f}")
print(f"This ratio is close to {9/8:.6f} (9/8), suggesting harmonic structure")
print(f"\nWhen we compress by 9^p, we're essentially working in base-9")
print(f"The denominators 729 (9³) and 799 create different harmonic signatures")
```
