```python
import numpy as np, matplotlib.pyplot as plt
plt.figure(figsize=(5,5), facecolor="black")
inner_r, outer_r = 1.0, 2.0
inner_ang = np.linspace(0, 2*np.pi, 12, endpoint=False)
outer_ang = np.linspace(0, 2*np.pi, 18, endpoint=False)

for k in range(8):
    rot = k*np.pi/4
    x = np.concatenate([inner_r*np.cos(inner_ang+rot),
                        outer_r*np.cos(outer_ang+rot)])
    y = np.concatenate([inner_r*np.sin(inner_ang+rot),
                        outer_r*np.sin(outer_ang+rot)])
    plt.scatter(x, y, s=16, c=plt.cm.hsv(k/8), edgecolors="none")

    inner_ang = np.linspace(0, 2*np.pi, 11, endpoint=False)
outer_ang = np.linspace(0, 2*np.pi, 17, endpoint=False)


plt.gca().set_aspect("equal")
plt.axis("off")
plt.show()
```

```python
import matplotlib.animation as animation, matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(4,4), facecolor="black")
ax.set_aspect("equal"); ax.axis("off")
inner_ang = np.linspace(0, 2*np.pi, 11, endpoint=False)
outer_ang = np.linspace(0, 2*np.pi, 17, endpoint=False)
inner_r, outer_r = 1.0, 2.0
sc = ax.scatter([], [], s=24, c="white")

def update(frame):
    rot = frame*np.pi/32   # 11.25Â° per frame
    x = np.concatenate([inner_r*np.cos(inner_ang+rot),
                        outer_r*np.cos(outer_ang+rot)])
    y = np.concatenate([inner_r*np.sin(inner_ang+rot),
                        outer_r*np.sin(outer_ang+rot)])
    sc.set_offsets(np.c_[x,y]); return sc,

ani = animation.FuncAnimation(fig, update, frames=64, blit=True, interval=80)
ani.save("/content/rotating_pattern.gif", writer="pillow")
print("Saved GIF to /content/rotating_pattern.gif")
```
